## 5. å¸ƒå°”ç±»å‹ (bool)

**å¸ƒå°”ç±»å‹ (bool)** ç”¨äºè¡¨ç¤ºé€»è¾‘å€¼ï¼Œåªæœ‰ä¸¤ä¸ªå€¼ï¼š`True` å’Œ `False`ã€‚

### ç‰¹ç‚¹ï¼š
- True å’Œ False æ˜¯Pythonçš„å…³é”®å­—
- å¸ƒå°”å€¼å®é™…ä¸Šæ˜¯æ•´æ•°çš„å­ç±»ï¼ˆTrue=1, False=0ï¼‰
- åœ¨æ¡ä»¶åˆ¤æ–­å’Œé€»è¾‘è¿ç®—ä¸­å¹¿æ³›ä½¿ç”¨


```python
# å¸ƒå°”ç±»å‹çš„åˆ›å»ºå’Œä½¿ç”¨

# ç›´æ¥åˆ›å»ºå¸ƒå°”å€¼
is_true = True
is_false = False

print(f"is_true = {is_true}, ç±»å‹: {type(is_true)}")
print(f"is_false = {is_false}, ç±»å‹: {type(is_false)}")

# å¸ƒå°”å€¼æ˜¯æ•´æ•°çš„å­ç±»
print(f"\nå¸ƒå°”å€¼ä¸æ•´æ•°çš„å…³ç³»:")
print(f"True == 1: {True == 1}")  # True
print(f"False == 0: {False == 0}")  # True
print(f"True + True = {True + True}")  # 2
print(f"False + False = {False + False}")  # 0

# å¸ƒå°”ç±»å‹è½¬æ¢
print(f"\nå¸ƒå°”ç±»å‹è½¬æ¢:")
print(f"bool(1) = {bool(1)}")  # True
print(f"bool(0) = {bool(0)}")  # False
print(f"bool(-1) = {bool(-1)}")  # Trueï¼ˆéé›¶æ•°å­—éƒ½æ˜¯Trueï¼‰
print(f"bool('') = {bool('')}")  # Falseï¼ˆç©ºå­—ç¬¦ä¸²ï¼‰
print(f"bool('hello') = {bool('hello')}")  # Trueï¼ˆéç©ºå­—ç¬¦ä¸²ï¼‰
print(f"bool([]) = {bool([])}")  # Falseï¼ˆç©ºåˆ—è¡¨ï¼‰
print(f"bool([1, 2]) = {bool([1, 2])}")  # Trueï¼ˆéç©ºåˆ—è¡¨ï¼‰
print(f"bool(None) = {bool(None)}")  # False

# é€»è¾‘è¿ç®—
a, b = True, False
print(f"\né€»è¾‘è¿ç®— (a={a}, b={b}):")
print(f"and (ä¸): {a} and {b} = {a and b}")  # False
print(f"or (æˆ–): {a} or {b} = {a or b}")  # True
print(f"not (é): not {a} = {not a}")  # False
print(f"not (é): not {b} = {not b}")  # True

# æ¯”è¾ƒè¿ç®—è¿”å›å¸ƒå°”å€¼
x, y = 10, 20
print(f"\næ¯”è¾ƒè¿ç®— (x={x}, y={y}):")
print(f"x == y: {x == y}")  # False
print(f"x != y: {x != y}")  # True
print(f"x < y: {x < y}")  # True
print(f"x > y: {x > y}")  # False
print(f"x <= y: {x <= y}")  # True
print(f"x >= y: {x >= y}")  # False

# æ¡ä»¶åˆ¤æ–­ä¸­çš„å¸ƒå°”å€¼
age = 18
is_adult = age >= 18
print(f"\næ¡ä»¶åˆ¤æ–­:")
print(f"age = {age}, is_adult = {is_adult}")
if is_adult:
    print("å·²æˆå¹´")
else:
    print("æœªæˆå¹´")

# å¸ƒå°”å€¼çš„çŸ­è·¯æ±‚å€¼
def check_value():
    print("å‡½æ•°è¢«è°ƒç”¨")
    return True

print(f"\nçŸ­è·¯æ±‚å€¼:")
result1 = False and check_value()  # check_valueä¸ä¼šè¢«è°ƒç”¨
print(f"False and check_value(): {result1}")

result2 = True or check_value()  # check_valueä¸ä¼šè¢«è°ƒç”¨
print(f"True or check_value(): {result2}")
```

```python
# é›†åˆè¿ç®—ï¼ˆé›†åˆçš„æ ¸å¿ƒåŠŸèƒ½ï¼‰

set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
print(f"set1: {set1}")
print(f"set2: {set2}")

# å¹¶é›†ï¼ˆunionï¼‰- åŒ…å«ä¸¤ä¸ªé›†åˆçš„æ‰€æœ‰å…ƒç´ 
union = set1 | set2  # æˆ–ä½¿ç”¨ set1.union(set2)
print(f"\nå¹¶é›† (set1 | set2): {union}")

# äº¤é›†ï¼ˆintersectionï¼‰- åŒ…å«ä¸¤ä¸ªé›†åˆçš„å…±åŒå…ƒç´ 
intersection = set1 & set2  # æˆ–ä½¿ç”¨ set1.intersection(set2)
print(f"äº¤é›† (set1 & set2): {intersection}")

# å·®é›†ï¼ˆdifferenceï¼‰- åŒ…å«åœ¨set1ä¸­ä½†ä¸åœ¨set2ä¸­çš„å…ƒç´ 
difference = set1 - set2  # æˆ–ä½¿ç”¨ set1.difference(set2)
print(f"å·®é›† (set1 - set2): {difference}")

# å¯¹ç§°å·®é›†ï¼ˆsymmetric_differenceï¼‰- åŒ…å«åœ¨ä»»ä¸€é›†åˆä¸­ä½†ä¸åœ¨ä¸¤ä¸ªé›†åˆä¸­çš„å…ƒç´ 
symmetric_diff = set1 ^ set2  # æˆ–ä½¿ç”¨ set1.symmetric_difference(set2)
print(f"å¯¹ç§°å·®é›† (set1 ^ set2): {symmetric_diff}")

# å­é›†å’Œè¶…é›†åˆ¤æ–­
set3 = {2, 3}
print(f"\nå­é›†å’Œè¶…é›†åˆ¤æ–­ (set3={set3}):")
print(f"set3æ˜¯set1çš„å­é›†: {set3.issubset(set1)}")  # æˆ– set3 <= set1
print(f"set1æ˜¯set3çš„è¶…é›†: {set1.issuperset(set3)}")  # æˆ– set1 >= set3
print(f"set1å’Œset2æ˜¯å¦ä¸ç›¸äº¤: {set1.isdisjoint({9, 10})}")  # æ£€æŸ¥æ˜¯å¦æœ‰å…±åŒå…ƒç´ 

# é›†åˆçš„ä¿®æ”¹æ–¹æ³•ï¼ˆåŸåœ°æ“ä½œï¼‰
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

set1.update(set2)  # åŸåœ°å¹¶é›†æ“ä½œ
print(f"\nåŸåœ°æ“ä½œ:")
print(f"set1.update(set2)å: {set1}")

set1 = {1, 2, 3, 4, 5}
set1.intersection_update(set2)  # åŸåœ°äº¤é›†æ“ä½œ
print(f"set1.intersection_update(set2)å: {set1}")

set1 = {1, 2, 3, 4, 5}
set1.difference_update(set2)  # åŸåœ°å·®é›†æ“ä½œ
print(f"set1.difference_update(set2)å: {set1}")

# å†»ç»“é›†åˆï¼ˆfrozensetï¼‰- ä¸å¯å˜çš„é›†åˆ
frozen = frozenset([1, 2, 3, 4, 5])
print(f"\nå†»ç»“é›†åˆ: {frozen}, ç±»å‹: {type(frozen)}")
# frozen.add(6)  # è¿™ä¼šæŠ¥é”™ï¼Œå› ä¸ºfrozensetæ˜¯ä¸å¯å˜çš„
```

```python
# ç±»å‹æ£€æŸ¥æ–¹æ³•

# ä½¿ç”¨type()å‡½æ•°
x = 42
y = "hello"
z = [1, 2, 3]

print("ä½¿ç”¨type()å‡½æ•°:")
print(f"type(x) = {type(x)}")
print(f"type(y) = {type(y)}")
print(f"type(z) = {type(z)}")

# æ£€æŸ¥ç±»å‹æ˜¯å¦ç›¸ç­‰
print(f"\nç±»å‹æ¯”è¾ƒ:")
print(f"type(x) == int: {type(x) == int}")
print(f"type(y) == str: {type(y) == str}")
print(f"type(z) == list: {type(z) == list}")

# ä½¿ç”¨isinstance()å‡½æ•°ï¼ˆæ¨èï¼‰
print(f"\nä½¿ç”¨isinstance()å‡½æ•°ï¼ˆæ¨èï¼‰:")
print(f"isinstance(x, int): {isinstance(x, int)}")
print(f"isinstance(y, str): {isinstance(y, str)}")
print(f"isinstance(z, list): {isinstance(z, list)}")

# isinstance()å¯ä»¥æ£€æŸ¥å¤šä¸ªç±»å‹
print(f"\næ£€æŸ¥å¤šä¸ªç±»å‹:")
print(f"isinstance(x, (int, float)): {isinstance(x, (int, float))}")  # xæ˜¯intæˆ–float
print(f"isinstance(y, (int, str)): {isinstance(y, (int, str))}")  # yæ˜¯intæˆ–str

# isinstance()å’Œtype()çš„åŒºåˆ«
class Animal:
    pass

class Dog(Animal):
    pass

my_dog = Dog()

print(f"\nisinstance()å’Œtype()çš„åŒºåˆ«:")
print(f"isinstance(my_dog, Dog): {isinstance(my_dog, Dog)}")
print(f"isinstance(my_dog, Animal): {isinstance(my_dog, Animal)}")  # isinstanceè€ƒè™‘ç»§æ‰¿
print(f"type(my_dog) == Dog: {type(my_dog) == Dog}")
print(f"type(my_dog) == Animal: {type(my_dog) == Animal}")  # typeä¸è€ƒè™‘ç»§æ‰¿

# è·å–ç±»å‹åç§°
print(f"\nè·å–ç±»å‹åç§°:")
print(f"type(x).__name__: {type(x).__name__}")
print(f"type(y).__name__: {type(y).__name__}")
print(f"type(z).__name__: {type(z).__name__}")

# å®é™…åº”ç”¨ï¼šç±»å‹æ£€æŸ¥å‡½æ•°
def add_numbers(a, b):
    """åªæ¥å—æ•°å­—ç±»å‹çš„å‚æ•°"""
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("å‚æ•°å¿…é¡»æ˜¯æ•°å­—ç±»å‹")
    return a + b

print(f"\nç±»å‹æ£€æŸ¥å‡½æ•°ç¤ºä¾‹:")
print(f"add_numbers(1, 2) = {add_numbers(1, 2)}")
print(f"add_numbers(1.5, 2.5) = {add_numbers(1.5, 2.5)}")
try:
    add_numbers("1", "2")  # è¿™ä¼šæŠ¥é”™
except TypeError as e:
    print(f"add_numbers('1', '2') æŠ¥é”™: {e}")
```

### ç»ƒä¹ 6ï¼šç»¼åˆåº”ç”¨

ç»“åˆå¤šç§æ•°æ®ç±»å‹å®Œæˆä¸€ä¸ªå®Œæ•´çš„æ•°æ®å¤„ç†ä»»åŠ¡ã€‚


```python
# ç»ƒä¹ 6ï¼šç»¼åˆåº”ç”¨ - å­¦ç”Ÿæˆç»©ç®¡ç†ç³»ç»Ÿ

# ä½¿ç”¨å­—å…¸å­˜å‚¨å¤šä¸ªå­¦ç”Ÿçš„ä¿¡æ¯
students_data = {
    "S001": {
        "name": "å¼ ä¸‰",
        "age": 20,
        "scores": [85, 90, 88],
        "courses": ("æ•°å­¦", "è‹±è¯­", "ç‰©ç†")
    },
    "S002": {
        "name": "æå››",
        "age": 21,
        "scores": [92, 88, 95],
        "courses": ("æ•°å­¦", "è‹±è¯­", "åŒ–å­¦")
    },
    "S003": {
        "name": "ç‹äº”",
        "age": 19,
        "scores": [78, 82, 80],
        "courses": ("æ•°å­¦", "è‹±è¯­", "ç‰©ç†")
    },
    "S004": {
        "name": "èµµå…­",
        "age": 22,
        "scores": [96, 94, 92],
        "courses": ("æ•°å­¦", "è‹±è¯­", "ç‰©ç†")
    }
}

print("å­¦ç”Ÿæˆç»©ç®¡ç†ç³»ç»Ÿ")
print("=" * 50)

# ä»»åŠ¡1ï¼šæ˜¾ç¤ºæ‰€æœ‰å­¦ç”Ÿä¿¡æ¯
print("\næ‰€æœ‰å­¦ç”Ÿä¿¡æ¯:")
for student_id, info in students_data.items():
    avg_score = sum(info["scores"]) / len(info["scores"])
    print(f"  {student_id}: {info['name']}, å¹´é¾„: {info['age']}, "
          f"å¹³å‡åˆ†: {avg_score:.2f}")

# ä»»åŠ¡2ï¼šæ‰¾å‡ºå¹³å‡åˆ†æœ€é«˜çš„å­¦ç”Ÿ
max_avg = 0
top_student = None
for student_id, info in students_data.items():
    avg_score = sum(info["scores"]) / len(info["scores"])
    if avg_score > max_avg:
        max_avg = avg_score
        top_student = (student_id, info["name"])

print(f"\nå¹³å‡åˆ†æœ€é«˜çš„å­¦ç”Ÿ: {top_student[1]} ({top_student[0]}), å¹³å‡åˆ†: {max_avg:.2f}")

# ä»»åŠ¡3ï¼šç»Ÿè®¡æ‰€æœ‰å­¦ç”Ÿé€‰ä¿®çš„è¯¾ç¨‹ï¼ˆä½¿ç”¨é›†åˆå»é‡ï¼‰
all_courses_set = set()
for info in students_data.values():
    all_courses_set.update(info["courses"])
print(f"\næ‰€æœ‰é€‰ä¿®è¯¾ç¨‹: {sorted(all_courses_set)}")

# ä»»åŠ¡4ï¼šæ‰¾å‡ºé€‰æ‹©"ç‰©ç†"è¯¾ç¨‹çš„å­¦ç”Ÿ
physics_students = []
for student_id, info in students_data.items():
    if "ç‰©ç†" in info["courses"]:
        physics_students.append(info["name"])
print(f"\né€‰æ‹©'ç‰©ç†'è¯¾ç¨‹çš„å­¦ç”Ÿ: {', '.join(physics_students)}")

# ä»»åŠ¡5ï¼šè®¡ç®—æ¯é—¨è¯¾ç¨‹çš„å¹³å‡åˆ†
course_scores = {}
for info in students_data.values():
    courses = info["courses"]
    scores = info["scores"]
    for i, course in enumerate(courses):
        if course not in course_scores:
            course_scores[course] = []
        course_scores[course].append(scores[i])

print(f"\næ¯é—¨è¯¾ç¨‹çš„å¹³å‡åˆ†:")
for course, scores in sorted(course_scores.items()):
    avg = sum(scores) / len(scores)
    print(f"  {course}: {avg:.2f}")

# ä»»åŠ¡6ï¼šç±»å‹æ£€æŸ¥
print(f"\næ•°æ®ç±»å‹æ£€æŸ¥:")
print(f"students_dataçš„ç±»å‹: {type(students_data).__name__}")
print(f"å­¦ç”Ÿä¿¡æ¯çš„ç±»å‹: {type(list(students_data.values())[0]).__name__}")
print(f"æˆç»©åˆ—è¡¨çš„ç±»å‹: {type(list(students_data.values())[0]['scores']).__name__}")
print(f"è¯¾ç¨‹å…ƒç»„çš„ç±»å‹: {type(list(students_data.values())[0]['courses']).__name__}")
```
