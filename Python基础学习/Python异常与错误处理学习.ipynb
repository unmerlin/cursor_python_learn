## 1. 异常处理简介

**异常（Exception）**是程序运行时发生的错误。Python提供了强大的异常处理机制，允许程序在遇到错误时优雅地处理，而不是直接崩溃。

### 为什么需要异常处理？
- 提高程序的健壮性和可靠性
- 提供友好的错误信息给用户
- 允许程序在错误发生后继续执行
- 便于调试和错误定位

### 异常处理的基本结构：
```python
try:
    # 可能出错的代码
except 异常类型:
    # 处理异常的代码
```


````python
# 异常处理示例：没有异常处理的情况

# 当代码出错时，程序会直接崩溃
# result = 10 / 0  # 这行代码会引发 ZeroDivisionError，程序会中断

# 使用异常处理可以优雅地处理错误
try:
    result = 10 / 0  # 尝试除以零
except ZeroDivisionError:  # 捕获 ZeroDivisionError 异常
    print("错误：不能除以零！")  # 处理异常，程序不会崩溃

print("程序继续执行...")  # 这行代码会正常执行
````

## 2. try-except 基础

`try-except` 是异常处理的核心结构。`try` 块包含可能出错的代码，`except` 块包含处理异常的代码。


````python
# try-except 基本用法

# 示例1：处理除零错误
try:
    num1 = 10
    num2 = 0
    result = num1 / num2  # 这里会引发 ZeroDivisionError
    print(f"结果: {result}")
except ZeroDivisionError:  # 捕获 ZeroDivisionError 异常
    print("错误：除数不能为零！")

print("程序继续执行...")
````

````python
# 示例2：处理类型转换错误
try:
    user_input = "abc"  # 假设这是用户输入
    number = int(user_input)  # 尝试将字符串转换为整数
    print(f"转换后的数字: {number}")
except ValueError:  # 捕获 ValueError 异常（当转换失败时）
    print("错误：无法将字符串转换为整数！")

print("程序继续执行...")
````

````python
# 示例3：处理索引错误
try:
    my_list = [1, 2, 3]
    print(f"列表: {my_list}")
    value = my_list[10]  # 索引超出范围，会引发 IndexError
    print(f"值: {value}")
except IndexError:  # 捕获 IndexError 异常
    print("错误：列表索引超出范围！")

print("程序继续执行...")
````

````python
# 示例4：处理键错误（字典）
try:
    my_dict = {"name": "Python", "version": 3.11}
    print(f"字典: {my_dict}")
    value = my_dict["age"]  # 键不存在，会引发 KeyError
    print(f"值: {value}")
except KeyError:  # 捕获 KeyError 异常
    print("错误：字典中不存在该键！")

print("程序继续执行...")
````

## 3. 多个 except 子句

可以使用多个 `except` 子句来处理不同类型的异常，每个 `except` 处理一种特定的异常类型。


````python
# 多个 except 子句处理不同类型的异常

def divide_numbers(a, b):
    """除法函数，处理多种可能的异常"""
    try:
        result = a / b  # 可能引发 ZeroDivisionError
        return result
    except ZeroDivisionError:  # 处理除零错误
        print("错误：除数不能为零！")
        return None
    except TypeError:  # 处理类型错误（例如传入非数字类型）
        print("错误：参数必须是数字类型！")
        return None

# 测试不同的情况
print("测试1：正常除法")
print(f"10 / 2 = {divide_numbers(10, 2)}")

print("\n测试2：除零错误")
print(f"10 / 0 = {divide_numbers(10, 0)}")

print("\n测试3：类型错误")
print(f"10 / '2' = {divide_numbers(10, '2')}")
````

````python
# 多个 except 子句的优先级：按顺序匹配

def process_data(data, index):
    """处理数据，可能引发多种异常"""
    try:
        value = data[index]  # 可能引发 IndexError 或 TypeError
        result = int(value)   # 可能引发 ValueError
        return result * 2
    except IndexError:  # 先检查索引错误
        print("错误：索引超出范围！")
        return None
    except ValueError:  # 再检查值错误
        print("错误：无法转换为整数！")
        return None
    except TypeError:  # 最后检查类型错误
        print("错误：数据类型不正确！")
        return None

# 测试不同的异常情况
print("测试1：正常情况")
print(f"结果: {process_data([1, 2, 3], 0)}")

print("\n测试2：索引错误")
print(f"结果: {process_data([1, 2, 3], 10)}")

print("\n测试3：值错误")
print(f"结果: {process_data(['1', 'abc', '3'], 1)}")

print("\n测试4：类型错误")
print(f"结果: {process_data(123, 0)}")
````

## 4. except 子句的多种用法

`except` 子句有多种使用方式，可以更灵活地处理异常。


````python
# 方式1：捕获所有异常（不推荐，但有时有用）

try:
    result = 10 / 0
except:  # 捕获所有异常（不指定异常类型）
    print("发生了某种错误！")

# 注意：这种方式会捕获所有异常，包括系统退出等，通常不推荐使用
````

````python
# 方式2：捕获异常并获取异常对象

try:
    result = 10 / 0
except ZeroDivisionError as e:  # as e 获取异常对象
    print(f"错误类型: {type(e).__name__}")  # 输出异常类型名称
    print(f"错误信息: {e}")  # 输出异常信息

# 异常对象包含错误的详细信息，可以用于日志记录或调试
````

````python
# 方式3：捕获多个异常类型（使用元组）

try:
    # 可能引发多种异常的情况
    data = {"key": "value"}
    index = int(input("请输入索引（或输入非数字）: ") or "0")
    value = data[index]  # 可能引发 KeyError 或 TypeError
except (KeyError, TypeError) as e:  # 同时捕获两种异常
    print(f"发生了键错误或类型错误: {type(e).__name__}")
    print(f"错误信息: {e}")
except ValueError:  # 单独处理值错误
    print("输入的不是有效数字！")
````

````python
# 方式4：捕获所有异常并获取异常对象（推荐方式）

try:
    result = int("abc") / 0  # 可能引发多种异常
except Exception as e:  # Exception 是所有内置异常的基类
    print(f"发生了错误: {type(e).__name__}")
    print(f"错误信息: {e}")
    # 这种方式比 except: 更好，因为可以获取异常信息
````

## 5. else 子句

`else` 子句在 `try` 块没有引发异常时执行。它必须放在所有 `except` 子句之后。


````python
# else 子句的基本用法

def safe_divide(a, b):
    """安全除法，使用 else 子句"""
    try:
        result = a / b
    except ZeroDivisionError:
        print("错误：除数不能为零！")
        return None
    else:  # 如果没有异常，执行这里的代码
        print(f"计算成功：{a} / {b} = {result}")
        return result

# 测试
print("测试1：正常除法")
safe_divide(10, 2)

print("\n测试2：除零错误")
safe_divide(10, 0)
````

````python
# else 子句的典型应用：文件操作

try:
    # 尝试打开文件
    file = open("test.txt", "r")
except FileNotFoundError:
    print("错误：文件不存在！")
except PermissionError:
    print("错误：没有权限访问文件！")
else:
    # 只有在文件成功打开时才执行
    print("文件打开成功！")
    content = file.read()
    print(f"文件内容长度: {len(content)} 字符")
    file.close()  # 关闭文件
    print("文件已关闭")

# 注意：在实际应用中，应该使用 with 语句来确保文件正确关闭
````

## 6. finally 子句

`finally` 子句无论是否发生异常都会执行，通常用于清理资源（如关闭文件、释放锁等）。


````python
# finally 子句的基本用法

def demonstrate_finally():
    """演示 finally 子句的执行"""
    try:
        print("执行 try 块...")
        result = 10 / 2  # 正常执行
        print(f"结果: {result}")
    except ZeroDivisionError:
        print("捕获到异常！")
    finally:
        # 无论是否发生异常，这里的代码都会执行
        print("finally 块总是会执行！")

print("情况1：没有异常")
demonstrate_finally()

print("\n情况2：有异常")
try:
    print("执行 try 块...")
    result = 10 / 0  # 引发异常
except ZeroDivisionError:
    print("捕获到异常！")
finally:
    print("finally 块总是会执行！")
````

````python
# finally 子句的典型应用：资源清理

def read_file_with_finally(filename):
    """使用 finally 确保文件被关闭"""
    file = None
    try:
        file = open(filename, "r")
        content = file.read()
        print(f"文件内容: {content[:50]}...")  # 只显示前50个字符
        return content
    except FileNotFoundError:
        print(f"错误：文件 '{filename}' 不存在！")
        return None
    except Exception as e:
        print(f"发生错误: {e}")
        return None
    finally:
        # 无论是否发生异常，都要关闭文件
        if file:
            file.close()
            print("文件已关闭（在 finally 中）")

# 注意：在实际应用中，更推荐使用 with 语句
# with open(filename, "r") as file:
#     content = file.read()
# 这样会自动关闭文件，无需 finally
````

````python
# try-except-else-finally 完整结构

def complete_structure(value):
    """演示完整的异常处理结构"""
    try:
        # 可能出错的代码
        result = 100 / value
    except ZeroDivisionError:
        # 处理特定异常
        print("错误：除数不能为零！")
        return None
    except TypeError:
        # 处理另一种异常
        print("错误：参数必须是数字类型！")
        return None
    else:
        # 没有异常时执行
        print(f"计算成功，结果: {result}")
    finally:
        # 无论是否有异常都执行
        print("清理工作完成")

# 测试
print("测试1：正常情况")
complete_structure(2)

print("\n测试2：除零错误")
complete_structure(0)

print("\n测试3：类型错误")
complete_structure("abc")
````

## 8. 异常对象和异常信息

异常对象包含错误的详细信息，可以通过多种方式获取这些信息。


````python
# 获取异常的基本信息

try:
    result = 10 / 0
except ZeroDivisionError as e:
    # 异常对象的基本属性
    print(f"异常类型: {type(e).__name__}")  # 异常类型名称
    print(f"异常信息: {str(e)}")  # 异常的字符串表示
    print(f"异常对象: {e}")  # 直接打印异常对象
    print(f"异常参数: {e.args}")  # 异常的参数（元组）
````

````python
# 使用 traceback 模块获取详细的错误信息

import traceback

try:
    def function_a():
        return function_b()
    
    def function_b():
        return function_c()
    
    def function_c():
        return 10 / 0  # 这里引发异常
    
    result = function_a()
except ZeroDivisionError as e:
    print("异常信息:")
    print(f"类型: {type(e).__name__}")
    print(f"消息: {e}")
    
    print("\n完整堆栈跟踪:")
    traceback.print_exc()  # 打印完整的堆栈跟踪信息
    
    print("\n堆栈跟踪字符串:")
    tb_str = traceback.format_exc()
    print(tb_str)
````

````python
# 使用 sys.exc_info() 获取异常信息

import sys

try:
    result = int("abc")
except ValueError:
    # 获取当前异常的信息
    exc_type, exc_value, exc_traceback = sys.exc_info()
    
    print(f"异常类型: {exc_type}")
    print(f"异常值: {exc_value}")
    print(f"异常类型名称: {exc_type.__name__}")
    print(f"异常消息: {str(exc_value)}")
    
    # 注意：exc_info() 返回的信息只在 except 块中有效
````

## 9. 自定义异常

可以创建自定义异常类来处理特定的错误情况，使代码更加清晰和易于维护。


````python
# 创建自定义异常类

# 方式1：继承 Exception 类
class CustomError(Exception):
    """自定义异常类"""
    pass

# 使用自定义异常
try:
    raise CustomError("这是一个自定义错误")
except CustomError as e:
    print(f"捕获自定义异常: {e}")
````

````python
# 方式2：添加自定义属性和方法

class AgeError(Exception):
    """年龄错误异常"""
    def __init__(self, age, message="年龄必须在 0 到 150 之间"):
        self.age = age  # 存储年龄值
        self.message = message
        super().__init__(self.message)  # 调用父类构造函数
    
    def __str__(self):
        return f"{self.message}，当前年龄: {self.age}"

# 使用自定义异常
def validate_age(age):
    """验证年龄"""
    if age < 0 or age > 150:
        raise AgeError(age)
    return age

try:
    validate_age(200)
except AgeError as e:
    print(f"年龄验证失败: {e}")
    print(f"输入的年龄: {e.age}")
````

````python
# 方式3：创建异常类层次结构

class ValidationError(Exception):
    """验证错误基类"""
    pass

class EmailValidationError(ValidationError):
    """邮箱验证错误"""
    pass

class PasswordValidationError(ValidationError):
    """密码验证错误"""
    pass

# 使用异常类层次结构
def validate_email(email):
    """验证邮箱"""
    if "@" not in email:
        raise EmailValidationError(f"无效的邮箱格式: {email}")

def validate_password(password):
    """验证密码"""
    if len(password) < 8:
        raise PasswordValidationError("密码长度至少8个字符")

# 测试
try:
    validate_email("invalid_email")
except EmailValidationError as e:
    print(f"邮箱错误: {e}")

try:
    validate_password("123")
except PasswordValidationError as e:
    print(f"密码错误: {e}")

# 可以捕获基类来捕获所有验证错误
try:
    validate_email("test@")
    validate_password("123")
except ValidationError as e:
    print(f"验证错误: {e}")
````

````python
# 使用 raise 重新引发异常

def process_data(data):
    """处理数据，可能引发异常"""
    try:
        result = int(data)
        return result * 2
    except ValueError:
        print("在 process_data 中捕获到 ValueError")
        raise  # 重新引发异常，让调用者处理

# 测试
try:
    process_data("abc")
except ValueError as e:
    print(f"在调用者中捕获到异常: {e}")
````

````python
# 使用 raise ... from ... 链接异常

def inner_function():
    """内部函数引发异常"""
    raise ValueError("原始错误")

def outer_function():
    """外部函数捕获并重新引发异常"""
    try:
        inner_function()
    except ValueError as e:
        raise RuntimeError("处理数据时出错") from e  # 链接异常

# 测试
try:
    outer_function()
except RuntimeError as e:
    print(f"捕获到运行时错误: {e}")
    print(f"原因: {e.__cause__}")  # 访问原始异常
````

## 10. 异常处理最佳实践

遵循最佳实践可以让异常处理更加有效和清晰。


### 10.1 只捕获能处理的异常


````python
# 好的做法：只捕获能处理的异常

def divide_numbers(a, b):
    """除法函数 - 好的做法"""
    try:
        return a / b
    except ZeroDivisionError:  # 只捕获能处理的异常
        print("除数不能为零，返回 None")
        return None
    # 让其他异常（如 TypeError）向上传播

# 不好的做法：捕获所有异常但不处理
def bad_divide(a, b):
    """不好的做法"""
    try:
        return a / b
    except:  # 捕获所有异常，包括不应该捕获的
        return None  # 隐藏了真正的错误
````

### 10.2 使用具体的异常类型


````python
# 好的做法：使用具体的异常类型

def read_config(filename):
    """读取配置文件 - 好的做法"""
    try:
        with open(filename, "r") as f:
            return f.read()
    except FileNotFoundError:  # 具体处理文件不存在
        print(f"配置文件 {filename} 不存在，使用默认配置")
        return "default_config"
    except PermissionError:  # 具体处理权限错误
        print(f"没有权限读取文件 {filename}")
        return None
    except Exception as e:  # 捕获其他未知异常
        print(f"读取配置文件时发生未知错误: {e}")
        return None

# 不好的做法：使用过于宽泛的异常类型
def bad_read_config(filename):
    """不好的做法"""
    try:
        with open(filename, "r") as f:
            return f.read()
    except Exception:  # 太宽泛，无法区分不同类型的错误
        return None
````

### 10.3 使用 with 语句管理资源


````python
# 好的做法：使用 with 语句（上下文管理器）

# with 语句会自动处理资源的打开和关闭，即使发生异常也会正确关闭
try:
    with open("test.txt", "w") as f:  # 文件会自动关闭
        f.write("Hello, World!")
        # 即使这里发生异常，文件也会被正确关闭
except IOError as e:
    print(f"文件操作错误: {e}")

# 不好的做法：手动管理资源（容易忘记关闭）
try:
    f = open("test.txt", "w")
    f.write("Hello, World!")
    f.close()  # 如果上面发生异常，这行不会执行
except IOError as e:
    print(f"文件操作错误: {e}")
    # 文件可能没有被关闭
````

### 10.4 提供有意义的错误信息


````python
# 好的做法：提供有意义的错误信息

def calculate_average(numbers):
    """计算平均值 - 好的做法"""
    if not numbers:
        raise ValueError("数字列表不能为空")  # 明确的错误信息
    
    try:
        return sum(numbers) / len(numbers)
    except TypeError:
        raise TypeError("参数必须是数字列表")  # 明确的错误信息

# 不好的做法：错误信息不明确
def bad_calculate_average(numbers):
    """不好的做法"""
    try:
        return sum(numbers) / len(numbers)
    except:
        raise Exception("错误")  # 错误信息不明确，无法知道具体问题
````

### 10.5 不要使用异常控制流程


````python
# 不好的做法：使用异常控制流程

def bad_check_key(dictionary, key):
    """不好的做法：用异常检查键是否存在"""
    try:
        return dictionary[key]
    except KeyError:
        return None  # 用异常来控制流程

# 好的做法：使用正常的方式检查

def good_check_key(dictionary, key):
    """好的做法：使用正常的方式检查"""
    if key in dictionary:  # 直接检查键是否存在
        return dictionary[key]
    return None

# 或者使用 get() 方法
def better_check_key(dictionary, key):
    """更好的做法：使用 get() 方法"""
    return dictionary.get(key, None)  # 如果键不存在，返回 None

# 测试
my_dict = {"name": "Python", "version": 3.11}
print(f"使用异常控制: {bad_check_key(my_dict, 'age')}")
print(f"正常检查: {good_check_key(my_dict, 'age')}")
print(f"使用 get(): {better_check_key(my_dict, 'age')}")
````

````python
# 案例1：安全的用户输入处理

def get_user_age():
    """获取用户年龄，带错误处理"""
    while True:  # 循环直到输入有效
        try:
            age_input = input("请输入您的年龄: ")
            age = int(age_input)  # 可能引发 ValueError
            
            if age < 0:
                print("年龄不能为负数，请重新输入")
                continue
            elif age > 150:
                print("年龄不能超过150，请重新输入")
                continue
            
            return age  # 输入有效，返回年龄
        except ValueError:
            print(f"'{age_input}' 不是有效的数字，请重新输入")
        except KeyboardInterrupt:
            print("\n用户取消输入")
            return None
        except Exception as e:
            print(f"发生未知错误: {e}")
            return None

# 注意：在实际运行时会等待用户输入
# 这里只是展示代码结构
print("函数已定义，可以调用 get_user_age() 来测试")
````

### 11.2 文件操作


````python
# 案例2：安全的文件读取

def read_file_safely(filename):
    """安全地读取文件，处理各种可能的错误"""
    try:
        with open(filename, "r", encoding="utf-8") as f:
            content = f.read()
            print(f"成功读取文件 '{filename}'")
            return content
    except FileNotFoundError:
        print(f"错误：文件 '{filename}' 不存在")
        return None
    except PermissionError:
        print(f"错误：没有权限读取文件 '{filename}'")
        return None
    except UnicodeDecodeError:
        print(f"错误：文件 '{filename}' 编码不正确")
        return None
    except IOError as e:
        print(f"错误：读取文件 '{filename}' 时发生IO错误: {e}")
        return None
    except Exception as e:
        print(f"错误：读取文件时发生未知错误: {type(e).__name__}: {e}")
        return None

# 测试（注意：文件可能不存在）
result = read_file_safely("non_existent.txt")
if result is None:
    print("文件读取失败")
````

### 11.3 网络请求处理


````python
# 案例3：网络请求异常处理（模拟）

class NetworkError(Exception):
    """网络错误"""
    pass

class TimeoutError(Exception):
    """超时错误"""
    pass

def make_request(url, timeout=5):
    """模拟网络请求，带异常处理"""
    import random
    
    # 模拟可能发生的各种错误
    error_type = random.choice(["success", "timeout", "network", "value"])
    
    if error_type == "success":
        return {"status": 200, "data": "请求成功"}
    elif error_type == "timeout":
        raise TimeoutError(f"请求超时: {url}")
    elif error_type == "network":
        raise NetworkError(f"网络错误: {url}")
    else:
        raise ValueError(f"无效的URL: {url}")

# 使用异常处理进行网络请求
def safe_request(url):
    """安全地发送网络请求"""
    try:
        response = make_request(url)
        return response
    except TimeoutError as e:
        print(f"超时错误: {e}")
        return {"status": "timeout", "error": str(e)}
    except NetworkError as e:
        print(f"网络错误: {e}")
        return {"status": "network_error", "error": str(e)}
    except ValueError as e:
        print(f"参数错误: {e}")
        return {"status": "invalid", "error": str(e)}
    except Exception as e:
        print(f"未知错误: {type(e).__name__}: {e}")
        return {"status": "error", "error": str(e)}

# 测试
print("模拟网络请求（结果随机）:")
for i in range(3):
    result = safe_request("https://example.com")
    print(f"请求 {i+1}: {result}")
````

### 11.4 数据验证


````python
# 案例4：数据验证和转换

class ValidationError(Exception):
    """验证错误基类"""
    pass

class EmailFormatError(ValidationError):
    """邮箱格式错误"""
    pass

class AgeRangeError(ValidationError):
    """年龄范围错误"""
    pass

def validate_user_data(email, age):
    """验证用户数据"""
    errors = []
    
    # 验证邮箱
    if "@" not in email or "." not in email.split("@")[1]:
        errors.append(EmailFormatError(f"无效的邮箱格式: {email}"))
    
    # 验证年龄
    try:
        age_int = int(age)
        if age_int < 0 or age_int > 150:
            errors.append(AgeRangeError(f"年龄必须在0-150之间: {age_int}"))
    except ValueError:
        errors.append(ValueError(f"年龄必须是数字: {age}"))
    
    # 如果有错误，抛出第一个错误
    if errors:
        raise errors[0]
    
    return {"email": email, "age": int(age)}

# 使用异常处理验证数据
def register_user(email, age):
    """注册用户，带数据验证"""
    try:
        user_data = validate_user_data(email, age)
        print(f"用户注册成功: {user_data}")
        return user_data
    except EmailFormatError as e:
        print(f"注册失败 - 邮箱错误: {e}")
        return None
    except AgeRangeError as e:
        print(f"注册失败 - 年龄错误: {e}")
        return None
    except ValidationError as e:
        print(f"注册失败 - 验证错误: {e}")
        return None
    except Exception as e:
        print(f"注册失败 - 未知错误: {e}")
        return None

# 测试
print("测试1：有效数据")
register_user("user@example.com", "25")

print("\n测试2：无效邮箱")
register_user("invalid_email", "25")

print("\n测试3：无效年龄")
register_user("user@example.com", "200")
````

### 练习1：安全的计算器

创建一个计算器函数，能够安全地处理各种可能的错误。


````python
# 练习1：安全的计算器

def safe_calculator(operation, a, b):
    """
    安全的计算器函数
    支持操作：+, -, *, /
    参数：
        operation: 操作符（字符串）
        a, b: 两个操作数
    返回：计算结果或 None（如果出错）
    """
    try:
        # 验证操作数是否为数字
        a = float(a)  # 可能引发 ValueError
        b = float(b)  # 可能引发 ValueError
        
        # 根据操作符执行计算
        if operation == "+":
            result = a + b
        elif operation == "-":
            result = a - b
        elif operation == "*":
            result = a * b
        elif operation == "/":
            result = a / b  # 可能引发 ZeroDivisionError
        else:
            raise ValueError(f"不支持的操作符: {operation}")
        
        return result
    except ValueError as e:
        print(f"值错误: {e}")
        return None
    except ZeroDivisionError:
        print("错误：除数不能为零！")
        return None
    except Exception as e:
        print(f"未知错误: {type(e).__name__}: {e}")
        return None

# 测试
print("测试计算器:")
print(f"10 + 5 = {safe_calculator('+', 10, 5)}")
print(f"10 - 5 = {safe_calculator('-', 10, 5)}")
print(f"10 * 5 = {safe_calculator('*', 10, 5)}")
print(f"10 / 5 = {safe_calculator('/', 10, 5)}")
print(f"10 / 0 = {safe_calculator('/', 10, 0)}")
print(f"10 / 'abc' = {safe_calculator('/', 10, 'abc')}")
print(f"10 % 5 = {safe_calculator('%', 10, 5)}")
````

### 练习2：安全的列表操作

创建一个函数，安全地访问列表元素。


````python
# 练习2：安全的列表操作

def safe_list_access(my_list, index, default=None):
    """
    安全地访问列表元素
    参数：
        my_list: 列表
        index: 索引（可以是整数或字符串）
        default: 如果索引无效，返回的默认值
    返回：列表元素或默认值
    """
    try:
        # 尝试将索引转换为整数
        if isinstance(index, str):
            index = int(index)  # 可能引发 ValueError
        
        # 访问列表元素
        return my_list[index]  # 可能引发 IndexError 或 TypeError
    except (ValueError, TypeError) as e:
        print(f"索引错误: 索引必须是整数，当前为 {type(index).__name__}")
        return default
    except IndexError:
        print(f"索引错误: 索引 {index} 超出范围（列表长度为 {len(my_list)}）")
        return default
    except Exception as e:
        print(f"未知错误: {type(e).__name__}: {e}")
        return default

# 测试
my_list = [10, 20, 30, 40, 50]
print(f"列表: {my_list}")
print(f"\n安全访问:")
print(f"索引 0: {safe_list_access(my_list, 0)}")
print(f"索引 2: {safe_list_access(my_list, 2)}")
print(f"索引 10: {safe_list_access(my_list, 10)}")
print(f"索引 '2': {safe_list_access(my_list, '2')}")
print(f"索引 'abc': {safe_list_access(my_list, 'abc')}")
print(f"索引 -1: {safe_list_access(my_list, -1)}")
````

### 练习3：数据转换和验证

创建一个函数，将字符串转换为数字并进行验证。


````python
# 练习3：数据转换和验证

def convert_and_validate(value, min_value=None, max_value=None):
    """
    将值转换为整数并验证范围
    参数：
        value: 要转换的值
        min_value: 最小值（可选）
        max_value: 最大值（可选）
    返回：转换后的整数或 None
    """
    try:
        # 尝试转换为整数
        num = int(value)  # 可能引发 ValueError
        
        # 验证范围
        if min_value is not None and num < min_value:
            raise ValueError(f"值 {num} 小于最小值 {min_value}")
        
        if max_value is not None and num > max_value:
            raise ValueError(f"值 {num} 大于最大值 {max_value}")
        
        return num
    except ValueError as e:
        print(f"转换或验证失败: {e}")
        return None
    except TypeError:
        print(f"类型错误: 无法转换类型 {type(value).__name__}")
        return None
    except Exception as e:
        print(f"未知错误: {type(e).__name__}: {e}")
        return None

# 测试
print("测试数据转换和验证:")
print(f"转换 '123': {convert_and_validate('123')}")
print(f"转换 '45.67': {convert_and_validate('45.67')}")
print(f"转换 'abc': {convert_and_validate('abc')}")
print(f"转换 '50' (范围 0-100): {convert_and_validate('50', 0, 100)}")
print(f"转换 '150' (范围 0-100): {convert_and_validate('150', 0, 100)}")
print(f"转换 '-10' (范围 0-100): {convert_and_validate('-10', 0, 100)}")
````

### 练习4：异常链和上下文

创建一个函数，演示异常链和上下文信息。


````python
# 练习4：异常链和上下文

class ProcessingError(Exception):
    """处理错误"""
    pass

def step1(data):
    """步骤1：可能引发 ValueError"""
    if not isinstance(data, (int, float)):
        raise ValueError(f"数据必须是数字，当前类型: {type(data).__name__}")
    return data * 2

def step2(data):
    """步骤2：可能引发 ValueError"""
    if data < 0:
        raise ValueError(f"数据不能为负数: {data}")
    return data ** 0.5

def process_data(data):
    """处理数据，使用异常链"""
    try:
        result1 = step1(data)  # 可能引发 ValueError
        result2 = step2(result1)  # 可能引发 ValueError
        return result2
    except ValueError as e:
        # 使用 from 链接异常，保留原始异常信息
        raise ProcessingError(f"处理数据时出错: {data}") from e

# 测试异常链
print("测试异常链:")
try:
    result = process_data("abc")  # 会引发异常链
except ProcessingError as e:
    print(f"捕获到处理错误: {e}")
    print(f"原始错误: {e.__cause__}")
    print(f"原始错误类型: {type(e.__cause__).__name__}")
````

### 练习5：综合应用 - 配置文件解析器

创建一个配置文件解析器，处理各种可能的错误。


````python
# 练习5：配置文件解析器

class ConfigError(Exception):
    """配置错误基类"""
    pass

class ConfigFileError(ConfigError):
    """配置文件错误"""
    pass

class ConfigParseError(ConfigError):
    """配置解析错误"""
    pass

def parse_config_line(line):
    """解析配置文件的单行"""
    line = line.strip()
    
    # 跳过空行和注释
    if not line or line.startswith("#"):
        return None
    
    # 解析 key=value 格式
    if "=" not in line:
        raise ConfigParseError(f"无效的配置行格式: {line}")
    
    key, value = line.split("=", 1)  # 只分割第一个等号
    key = key.strip()
    value = value.strip()
    
    # 尝试转换值的类型
    try:
        # 尝试转换为整数
        value = int(value)
    except ValueError:
        try:
            # 尝试转换为浮点数
            value = float(value)
        except ValueError:
            # 保持为字符串
            pass
    
    return {key: value}

def load_config(filename):
    """加载配置文件"""
    config = {}
    
    try:
        with open(filename, "r", encoding="utf-8") as f:
            for line_num, line in enumerate(f, 1):
                try:
                    parsed = parse_config_line(line)
                    if parsed:
                        config.update(parsed)
                except ConfigParseError as e:
                    print(f"警告：第 {line_num} 行解析失败: {e}")
                    continue
    except FileNotFoundError:
        raise ConfigFileError(f"配置文件不存在: {filename}")
    except PermissionError:
        raise ConfigFileError(f"没有权限读取配置文件: {filename}")
    except Exception as e:
        raise ConfigFileError(f"读取配置文件时出错: {e}")
    
    return config

# 测试（注意：需要创建配置文件）
# 这里只是展示代码结构
print("配置文件解析器已定义")
print("使用方法：")
print("1. 创建配置文件（如 config.txt）")
print("2. 调用 load_config('config.txt')")
print("\n配置文件格式示例：")
print("name=Python")
print("version=3.11")
print("type=programming_language")
````
