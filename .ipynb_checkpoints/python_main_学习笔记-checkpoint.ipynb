## 2. åˆ›å»ºç¤ºä¾‹æ–‡ä»¶æ¥æ¼”ç¤º

```python
# åˆ›å»ºç¬¬ä¸€ä¸ªæ–‡ä»¶ï¼šcalculator.py
calculator_code = '''# calculator.py - è®¡ç®—å™¨æ¨¡å—

def add(a, b):
    """åŠ æ³•å‡½æ•°"""
    return a + b

def subtract(a, b):
    """å‡æ³•å‡½æ•°"""
    return a - b

def multiply(a, b):
    """ä¹˜æ³•å‡½æ•°"""
    return a * b

def divide(a, b):
    """é™¤æ³•å‡½æ•°"""
    if b == 0:
        return "é”™è¯¯ï¼šé™¤æ•°ä¸èƒ½ä¸º0"
    return a / b

# æ‰“å°å½“å‰æ¨¡å—å
print(f"calculator.py ä¸­çš„ __name__ = {__name__}")

# å…³é”®éƒ¨åˆ†ï¼šåˆ¤æ–­æ˜¯å¦ç›´æ¥è¿è¡Œ
if __name__ == "__main__":
    print("=" * 50)
    print("calculator.py è¢«ç›´æ¥è¿è¡Œäº†ï¼")
    print("=" * 50)
    
    # æµ‹è¯•æ‰€æœ‰å‡½æ•°
    print(f"10 + 5 = {add(10, 5)}")
    print(f"10 - 5 = {subtract(10, 5)}")
    print(f"10 Ã— 5 = {multiply(10, 5)}")
    print(f"10 Ã· 5 = {divide(10, 5)}")
else:
    print("calculator.py è¢«å¯¼å…¥äº†ï¼Œæµ‹è¯•ä»£ç ä¸ä¼šè¿è¡Œ")
'''

# å†™å…¥æ–‡ä»¶
with open('calculator.py', 'w', encoding='utf-8') as f:
    f.write(calculator_code)

print("âœ… å·²åˆ›å»º calculator.py æ–‡ä»¶")
```

```python
# åˆ›å»ºç¬¬äºŒä¸ªæ–‡ä»¶ï¼šmain_app.py
main_app_code = '''# main_app.py - ä¸»ç¨‹åº

# å¯¼å…¥ calculator æ¨¡å—
import calculator

print(f"\\nmain_app.py ä¸­çš„ __name__ = {__name__}")
print("-" * 50)

# ä½¿ç”¨ calculator æ¨¡å—çš„å‡½æ•°
result1 = calculator.add(100, 200)
result2 = calculator.multiply(7, 8)

print(f"ä½¿ç”¨å¯¼å…¥çš„å‡½æ•°ï¼š100 + 200 = {result1}")
print(f"ä½¿ç”¨å¯¼å…¥çš„å‡½æ•°ï¼š7 Ã— 8 = {result2}")

if __name__ == "__main__":
    print("\\n" + "=" * 50)
    print("main_app.py è¢«ç›´æ¥è¿è¡Œ")
    print("=" * 50)
'''

with open('main_app.py', 'w', encoding='utf-8') as f:
    f.write(main_app_code)

print("âœ… å·²åˆ›å»º main_app.py æ–‡ä»¶")
```

```python
# åˆ›å»ºä¸€ä¸ªå­—ç¬¦ä¸²å·¥å…·æ¨¡å—
string_utils_code = '''# string_utils.py - å­—ç¬¦ä¸²å·¥å…·æ¨¡å—

def reverse_string(text):
    """åè½¬å­—ç¬¦ä¸²"""
    return text[::-1]

def count_words(text):
    """ç»Ÿè®¡å•è¯æ•°é‡"""
    return len(text.split())

def to_uppercase(text):
    """è½¬æ¢ä¸ºå¤§å†™"""
    return text.upper()

if __name__ == "__main__":
    # æµ‹è¯•ä»£ç 
    test_text = "Hello Python World"
    print(f"åŸæ–‡æœ¬: {test_text}")
    print(f"åè½¬: {reverse_string(test_text)}")
    print(f"å•è¯æ•°: {count_words(test_text)}")
    print(f"å¤§å†™: {to_uppercase(test_text)}")
'''

with open('string_utils.py', 'w', encoding='utf-8') as f:
    f.write(string_utils_code)

print("âœ… å·²åˆ›å»º string_utils.py")
print("\nè¿è¡Œæµ‹è¯•ï¼š")
!python string_utils.py
```
