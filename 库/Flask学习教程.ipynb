## Flask简介

**Flask** 是一个轻量级的Python Web应用框架，使用Werkzeug WSGI工具包和Jinja2模板引擎。

### 计算机网络基础：Web应用架构

- **HTTP协议**：超文本传输协议，客户端（浏览器）和服务器之间的通信协议
- **WSGI**：Web Server Gateway Interface，Python Web服务器和Web应用之间的标准接口
- **MVC模式**：Model-View-Controller，Flask遵循这种设计模式
  - Model：数据模型
  - View：视图（HTML模板）
  - Controller：控制器（路由处理函数）

### Flask的特点
- 轻量级、灵活
- 易于扩展
- 内置开发服务器
- 支持RESTful API开发

## 第一个Flask应用

### 基础概念

**Flask应用对象**：`Flask(__name__)` 创建一个Flask应用实例
- `__name__` 参数帮助Flask定位资源文件（模板、静态文件等）

**路由（Route）**：URL路径到Python函数的映射
- 当用户访问某个URL时，Flask会调用对应的视图函数

**视图函数（View Function）**：处理HTTP请求并返回响应的Python函数


````python
from flask import Flask

# 创建Flask应用实例
# __name__ 是当前模块名，Flask用它来定位资源文件
app = Flask(__name__)

# 定义路由：使用装饰器 @app.route() 将URL路径映射到函数
# '/' 表示网站根路径（如 http://localhost:5000/）
@app.route('/')
def hello():
    """
    视图函数：处理HTTP GET请求并返回响应
    返回的字符串会被Flask转换为HTTP响应
    """
    return '<h1>Hello, Flask!</h1><p>这是我的第一个Flask应用</p>'

# 运行应用（仅在直接执行此脚本时运行）
# debug=True 启用调试模式，代码修改后自动重载
if __name__ == '__main__':
    app.run(debug=True, host='0.0.0.0', port=5000)
    # debug=True: 开发模式，显示详细错误信息
    # host='0.0.0.0': 允许外部访问
    # port=5000: 指定端口号
````

### 计算机网络知识：HTTP请求和响应

**HTTP请求组成**：
- 请求行：方法（GET/POST） + URL + 协议版本
- 请求头：包含客户端信息（User-Agent、Accept等）
- 请求体：POST请求的数据

**HTTP响应组成**：
- 状态行：状态码（200 OK、404 Not Found等）
- 响应头：服务器信息（Content-Type、Content-Length等）
- 响应体：实际内容（HTML、JSON等）

**常见HTTP方法**：
- GET：获取资源（查询参数在URL中）
- POST：提交数据（数据在请求体中）
- PUT：更新资源
- DELETE：删除资源


## 路由和视图函数

### 动态路由

Flask支持在URL中使用变量，实现动态路由。


````python
from flask import Flask, request

app = Flask(__name__)

# 静态路由：固定URL路径
@app.route('/')
def index():
    return '<h1>首页</h1>'

# 动态路由：使用 <变量名> 定义URL变量
# 访问 /user/张三 时，name变量会被赋值为"张三"
@app.route('/user/<name>')
def show_user(name):
    """
    动态路由示例
    :param name: URL中的用户名参数
    """
    return f'<h1>欢迎, {name}!</h1>'

# 指定变量类型：<int:id> 只接受整数
# 如果访问 /post/abc 会返回404错误
@app.route('/post/<int:post_id>')
def show_post(post_id):
    """
    类型转换：Flask会自动将URL中的字符串转换为int类型
    支持的类型：int, float, path（包含斜杠的路径）
    """
    return f'<h1>文章ID: {post_id}</h1>'

# 多个动态参数
@app.route('/user/<name>/post/<int:post_id>')
def user_post(name, post_id):
    return f'<p>用户 {name} 的文章 {post_id}</p>'

# 指定HTTP方法
@app.route('/login', methods=['GET', 'POST'])
def login():
    """
    methods参数指定允许的HTTP方法
    默认只允许GET请求
    """
    if request.method == 'POST':
        return '<p>处理登录POST请求</p>'
    return '<p>显示登录表单（GET请求）</p>'

if __name__ == '__main__':
    app.run(debug=True)
````

## 模板渲染（HTML基础）

### Jinja2模板引擎

Flask使用Jinja2模板引擎来渲染HTML。模板文件通常放在 `templates` 文件夹中。

### HTML基础知识

**HTML结构**：
- `<html>`：根元素
- `<head>`：文档头部（元数据、CSS、JS引用）
- `<body>`：文档主体（可见内容）
- 标签通常是成对的：`<tag>内容</tag>`
- 自闭合标签：`<img />`, `<br />`, `<hr />`

**常用HTML标签**：
- `<h1>` 到 `<h6>`：标题（h1最大）
- `<p>`：段落
- `<div>`：块级容器
- `<span>`：行内容器
- `<a href="url">`：超链接
- `<img src="path" alt="text">`：图片
- `<form>`：表单
- `<input>`：输入框
- `<button>`：按钮


````python
from flask import Flask, render_template

app = Flask(__name__)

# 渲染模板：使用 render_template() 函数
# Flask会在 templates 文件夹中查找模板文件
@app.route('/')
def index():
    """
    render_template() 函数：
    1. 查找 templates/index.html 文件
    2. 使用Jinja2引擎渲染模板
    3. 返回HTML字符串作为HTTP响应
    """
    return render_template('index.html')

# 传递变量到模板
@app.route('/user/<name>')
def user(name):
    """
    模板变量：通过关键字参数传递数据到模板
    模板中可以使用 {{ name }} 来显示变量值
    """
    return render_template('user.html', name=name, age=25)

# 传递复杂数据结构
@app.route('/products')
def products():
    """
    传递列表和字典到模板
    """
    products_list = [
        {'name': '苹果', 'price': 5.5},
        {'name': '香蕉', 'price': 3.2},
        {'name': '橙子', 'price': 4.8}
    ]
    return render_template('products.html', products=products_list)

if __name__ == '__main__':
    app.run(debug=True)
````

### Jinja2模板语法

**变量输出**：`{{ variable }}` - 输出变量值，自动转义HTML特殊字符

**控制结构**：
- `{% if condition %}...{% endif %}` - 条件判断
- `{% for item in list %}...{% endfor %}` - 循环
- `{% block name %}...{% endblock %}` - 模板继承

**过滤器**：`{{ variable|filter }}`
- `|upper` - 转大写
- `|lower` - 转小写
- `|length` - 获取长度
- `|default('默认值')` - 默认值

**注释**：`{# 这是注释 #}`

## 请求处理

### 获取请求数据

Flask提供了 `request` 对象来访问HTTP请求数据。


````python
from flask import Flask, request, jsonify

app = Flask(__name__)

# 获取GET请求参数（查询字符串）
# 访问：http://localhost:5000/search?q=flask&page=1
@app.route('/search')
def search():
    """
    request.args：获取URL查询参数（GET请求）
    - request.args.get('key')：获取单个参数，不存在返回None
    - request.args.get('key', 'default')：获取参数，不存在返回默认值
    - request.args['key']：直接访问，不存在会抛出KeyError
    """
    query = request.args.get('q', '')  # 获取查询关键词
    page = request.args.get('page', '1', type=int)  # 类型转换
    
    return f'<h1>搜索结果</h1><p>关键词：{query}</p><p>页码：{page}</p>'

# 获取POST请求数据
@app.route('/submit', methods=['GET', 'POST'])
def submit():
    """
    request.method：获取HTTP方法（'GET', 'POST'等）
    request.form：获取表单数据（POST请求，Content-Type: application/x-www-form-urlencoded）
    request.json：获取JSON数据（POST请求，Content-Type: application/json）
    """
    if request.method == 'POST':
        # 获取表单数据
        name = request.form.get('name')
        email = request.form.get('email')
        
        # 获取JSON数据（如果请求是JSON格式）
        # data = request.get_json()
        
        return f'<h1>提交成功</h1><p>姓名：{name}</p><p>邮箱：{email}</p>'
    
    # GET请求：显示表单
    return '''
    <form method="POST">
        <p>姓名：<input type="text" name="name"></p>
        <p>邮箱：<input type="email" name="email"></p>
        <p><button type="submit">提交</button></p>
    </form>
    '''

# 获取请求头信息
@app.route('/headers')
def show_headers():
    """
    request.headers：获取HTTP请求头
    常用请求头：
    - User-Agent：客户端信息
    - Referer：来源页面
    - Accept：接受的内容类型
    """
    user_agent = request.headers.get('User-Agent', 'Unknown')
    return f'<p>User-Agent: {user_agent}</p>'

# 处理JSON API请求
@app.route('/api/data', methods=['POST'])
def api_data():
    """
    RESTful API示例
    - 接收JSON数据
    - 返回JSON响应
    """
    # 检查请求是否为JSON格式
    if not request.is_json:
        return jsonify({'error': '请求必须是JSON格式'}), 400
    
    # 获取JSON数据
    data = request.get_json()
    
    # 处理数据（示例）
    result = {
        'status': 'success',
        'received_data': data,
        'message': '数据接收成功'
    }
    
    # jsonify()：将Python字典转换为JSON响应
    return jsonify(result), 200

# 获取文件上传
@app.route('/upload', methods=['GET', 'POST'])
def upload_file():
    """
    request.files：获取上传的文件
    - request.files['key']：获取文件对象
    - file.save(path)：保存文件
    """
    if request.method == 'POST':
        if 'file' not in request.files:
            return '<p>没有文件被上传</p>'
        
        file = request.files['file']
        if file.filename == '':
            return '<p>没有选择文件</p>'
        
        # 保存文件（实际应用中需要验证文件类型和大小）
        # file.save(f'uploads/{file.filename}')
        return f'<p>文件 {file.filename} 上传成功</p>'
    
    return '''
    <form method="POST" enctype="multipart/form-data">
        <p><input type="file" name="file"></p>
        <p><button type="submit">上传</button></p>
    </form>
    '''

if __name__ == '__main__':
    app.run(debug=True)
````

### 计算机网络知识：HTTP请求格式

**请求行**：
```
GET /search?q=flask HTTP/1.1
```

**请求头示例**：
```
Host: localhost:5000
User-Agent: Mozilla/5.0...
Accept: text/html,application/json
Content-Type: application/x-www-form-urlencoded
```

**请求体**（POST请求）：

**Content-Type类型**：
- `application/x-www-form-urlencoded`：表单数据
- `application/json`：JSON数据
- `multipart/form-data`：文件上传


## 表单处理

### HTML表单基础

**表单元素**：
- `<form>`：表单容器
  - `action`：提交目标URL
  - `method`：HTTP方法（GET/POST）
  - `enctype`：编码类型（文件上传需要multipart/form-data）
- `<input>`：输入框
  - `type="text"`：文本输入
  - `type="email"`：邮箱输入
  - `type="password"`：密码输入
  - `type="submit"`：提交按钮
- `<textarea>`：多行文本输入
- `<select>`：下拉选择框
- `<label>`：标签（提高可访问性）


````python
from flask import Flask, request, render_template, redirect, url_for, flash

app = Flask(__name__)
app.secret_key = 'your-secret-key-here'  # 用于flash消息的加密

# 简单的数据存储（实际应用中使用数据库）
users = []

@app.route('/register', methods=['GET', 'POST'])
def register():
    """
    用户注册表单处理
    演示完整的表单验证和处理流程
    """
    if request.method == 'POST':
        # 获取表单数据
        username = request.form.get('username', '').strip()
        email = request.form.get('email', '').strip()
        password = request.form.get('password', '')
        confirm_password = request.form.get('confirm_password', '')
        
        # 表单验证
        errors = []
        
        if not username:
            errors.append('用户名不能为空')
        elif len(username) < 3:
            errors.append('用户名至少3个字符')
        
        if not email:
            errors.append('邮箱不能为空')
        elif '@' not in email:
            errors.append('邮箱格式不正确')
        
        if not password:
            errors.append('密码不能为空')
        elif len(password) < 6:
            errors.append('密码至少6个字符')
        
        if password != confirm_password:
            errors.append('两次输入的密码不一致')
        
        # 检查用户名是否已存在
        if any(user['username'] == username for user in users):
            errors.append('用户名已存在')
        
        # 如果有错误，返回错误信息
        if errors:
            return render_template('register.html', errors=errors, 
                                 username=username, email=email)
        
        # 保存用户（实际应用中保存到数据库）
        user = {
            'username': username,
            'email': email,
            'password': password  # 实际应用中应该加密存储
        }
        users.append(user)
        
        # 使用flash消息（需要session支持）
        flash(f'注册成功！欢迎 {username}')
        
        # 重定向到登录页面
        return redirect(url_for('login'))
    
    # GET请求：显示注册表单
    return render_template('register.html')

@app.route('/login', methods=['GET', 'POST'])
def login():
    """
    用户登录表单处理
    """
    if request.method == 'POST':
        username = request.form.get('username', '').strip()
        password = request.form.get('password', '')
        
        # 验证用户
        user = next((u for u in users if u['username'] == username and u['password'] == password), None)
        
        if user:
            flash(f'登录成功！欢迎回来，{username}')
            return redirect(url_for('dashboard', username=username))
        else:
            return render_template('login.html', error='用户名或密码错误')
    
    return render_template('login.html')

@app.route('/dashboard/<username>')
def dashboard(username):
    """用户仪表板"""
    return f'''
    <h1>欢迎，{username}！</h1>
    <p>这是您的个人仪表板</p>
    <a href="/register">注册新用户</a> | 
    <a href="/login">重新登录</a>
    '''

if __name__ == '__main__':
    app.run(debug=True)
````

## 会话管理

### Session和Cookie

**Cookie**：存储在客户端（浏览器）的小型数据
- 由服务器通过`Set-Cookie`响应头设置
- 浏览器自动在后续请求中通过`Cookie`请求头发送
- 用于保持用户状态（如登录状态）

**Session**：存储在服务器端的用户会话数据
- Flask使用加密的Cookie来存储Session ID
- 实际数据存储在服务器内存或数据库中
- 比Cookie更安全（敏感数据不存储在客户端）

### 计算机网络知识：Cookie机制

1. **服务器设置Cookie**：
   ```
   HTTP/1.1 200 OK
   Set-Cookie: session_id=abc123; Path=/; HttpOnly
   ```

2. **客户端发送Cookie**：
   ```
   GET /dashboard HTTP/1.1
   Cookie: session_id=abc123
   ```

3. **Cookie属性**：
   - `Path`：Cookie生效的路径
   - `Domain`：Cookie生效的域名
   - `HttpOnly`：防止JavaScript访问（安全）
   - `Secure`：仅通过HTTPS传输
   - `Expires/Max-Age`：过期时间


````python
from flask import Flask, session, redirect, url_for, request, render_template

app = Flask(__name__)
# secret_key用于加密session cookie，必须设置且保持机密
app.secret_key = 'your-secret-key-change-this-in-production'

# 模拟用户数据库
users_db = {
    'admin': {'password': 'admin123', 'name': '管理员'},
    'user1': {'password': 'user123', 'name': '用户1'}
}

@app.route('/')
def index():
    """首页：检查用户是否已登录"""
    if 'username' in session:
        # session是字典，可以存储任意数据
        return f'''
        <h1>欢迎回来，{session["username"]}！</h1>
        <p>您的姓名：{session.get("name", "未知")}</p>
        <p><a href="/logout">退出登录</a></p>
        '''
    return '''
    <h1>欢迎访问</h1>
    <p><a href="/login">请先登录</a></p>
    '''

@app.route('/login', methods=['GET', 'POST'])
def login():
    """登录处理：设置session"""
    if request.method == 'POST':
        username = request.form.get('username')
        password = request.form.get('password')
        
        # 验证用户
        if username in users_db and users_db[username]['password'] == password:
            # 设置session数据
            session['username'] = username
            session['name'] = users_db[username]['name']
            session['login_time'] = str(datetime.now())
            
            # 重定向到首页
            return redirect(url_for('index'))
        else:
            return render_template('session_login.html', error='用户名或密码错误')
    
    return render_template('session_login.html')

@app.route('/logout')
def logout():
    """退出登录：清除session"""
    # 清除session数据
    session.pop('username', None)
    session.pop('name', None)
    session.pop('login_time', None)
    
    # 或者使用 session.clear() 清除所有session数据
    
    return redirect(url_for('index'))

@app.route('/profile')
def profile():
    """用户资料页面：需要登录才能访问"""
    if 'username' not in session:
        return redirect(url_for('login'))
    
    return f'''
    <h1>用户资料</h1>
    <p>用户名：{session["username"]}</p>
    <p>姓名：{session.get("name", "未知")}</p>
    <p>登录时间：{session.get("login_time", "未知")}</p>
    <p><a href="/">返回首页</a> | <a href="/logout">退出</a></p>
    '''

# 使用装饰器实现登录检查
from functools import wraps

def login_required(f):
    """
    装饰器：检查用户是否已登录
    如果未登录，重定向到登录页面
    """
    @wraps(f)
    def decorated_function(*args, **kwargs):
        if 'username' not in session:
            return redirect(url_for('login', next=request.url))
        return f(*args, **kwargs)
    return decorated_function

@app.route('/protected')
@login_required  # 使用装饰器保护路由
def protected():
    """受保护的页面：只有登录用户才能访问"""
    return f'<h1>这是受保护的页面</h1><p>只有登录用户才能看到这个内容</p>'

if __name__ == '__main__':
    from datetime import datetime
    app.run(debug=True)
````

````python
# 练习2：RESTful API实现
from flask import Flask, jsonify, request

app = Flask(__name__)

# 任务数据存储（实际应用中使用数据库）
tasks = [
    {'id': 1, 'title': '学习Flask', 'completed': False},
    {'id': 2, 'title': '完成项目', 'completed': True}
]

# 获取所有任务
@app.route('/api/tasks', methods=['GET'])
def get_tasks():
    """
    GET /api/tasks
    返回所有任务列表
    """
    return jsonify({
        'status': 'success',
        'count': len(tasks),
        'tasks': tasks
    }), 200

# 获取单个任务
@app.route('/api/tasks/<int:task_id>', methods=['GET'])
def get_task(task_id):
    """
    GET /api/tasks/<id>
    返回指定ID的任务
    """
    task = next((t for t in tasks if t['id'] == task_id), None)
    if not task:
        return jsonify({
            'status': 'error',
            'message': '任务不存在'
        }), 404
    
    return jsonify({
        'status': 'success',
        'task': task
    }), 200

# 创建新任务
@app.route('/api/tasks', methods=['POST'])
def create_task():
    """
    POST /api/tasks
    创建新任务
    请求体：{"title": "任务标题"}
    """
    if not request.is_json:
        return jsonify({
            'status': 'error',
            'message': '请求必须是JSON格式'
        }), 400
    
    data = request.get_json()
    title = data.get('title', '').strip()
    
    if not title:
        return jsonify({
            'status': 'error',
            'message': '任务标题不能为空'
        }), 400
    
    # 创建新任务
    new_task = {
        'id': max([t['id'] for t in tasks], default=0) + 1,
        'title': title,
        'completed': False
    }
    tasks.append(new_task)
    
    return jsonify({
        'status': 'success',
        'message': '任务创建成功',
        'task': new_task
    }), 201  # 201 Created

# 更新任务
@app.route('/api/tasks/<int:task_id>', methods=['PUT'])
def update_task(task_id):
    """
    PUT /api/tasks/<id>
    更新任务
    请求体：{"title": "新标题", "completed": true}
    """
    task = next((t for t in tasks if t['id'] == task_id), None)
    if not task:
        return jsonify({
            'status': 'error',
            'message': '任务不存在'
        }), 404
    
    if not request.is_json:
        return jsonify({
            'status': 'error',
            'message': '请求必须是JSON格式'
        }), 400
    
    data = request.get_json()
    
    # 更新任务属性
    if 'title' in data:
        task['title'] = data['title'].strip()
    if 'completed' in data:
        task['completed'] = bool(data['completed'])
    
    return jsonify({
        'status': 'success',
        'message': '任务更新成功',
        'task': task
    }), 200

# 删除任务
@app.route('/api/tasks/<int:task_id>', methods=['DELETE'])
def delete_task(task_id):
    """
    DELETE /api/tasks/<id>
    删除任务
    """
    global tasks
    task = next((t for t in tasks if t['id'] == task_id), None)
    if not task:
        return jsonify({
            'status': 'error',
            'message': '任务不存在'
        }), 404
    
    tasks = [t for t in tasks if t['id'] != task_id]
    
    return jsonify({
        'status': 'success',
        'message': '任务删除成功'
    }), 200

# API文档页面
@app.route('/api')
def api_docs():
    """API文档"""
    return '''
    <h1>任务管理API文档</h1>
    <h2>端点列表：</h2>
    <ul>
        <li><strong>GET /api/tasks</strong> - 获取所有任务</li>
        <li><strong>GET /api/tasks/&lt;id&gt;</strong> - 获取单个任务</li>
        <li><strong>POST /api/tasks</strong> - 创建新任务（需要JSON：{"title": "任务标题"}）</li>
        <li><strong>PUT /api/tasks/&lt;id&gt;</strong> - 更新任务（需要JSON：{"title": "新标题", "completed": true}）</li>
        <li><strong>DELETE /api/tasks/&lt;id&gt;</strong> - 删除任务</li>
    </ul>
    <h2>测试示例：</h2>
    <pre>
# 使用curl测试（在命令行中执行）
curl http://localhost:5000/api/tasks
curl -X POST http://localhost:5000/api/tasks -H "Content-Type: application/json" -d '{"title":"新任务"}'
curl -X PUT http://localhost:5000/api/tasks/1 -H "Content-Type: application/json" -d '{"completed":true}'
curl -X DELETE http://localhost:5000/api/tasks/1
    </pre>
    '''

if __name__ == '__main__':
    app.run(debug=True)
````

````python
# 练习3：错误处理
from flask import Flask, render_template, abort

app = Flask(__name__)

@app.route('/')
def index():
    return '<h1>首页</h1><p><a href="/error/404">触发404错误</a> | <a href="/error/500">触发500错误</a></p>'

@app.route('/error/<int:code>')
def trigger_error(code):
    """触发不同类型的错误"""
    if code == 404:
        abort(404)  # 主动触发404错误
    elif code == 500:
        abort(500)  # 主动触发500错误
    else:
        return f'<p>错误代码：{code}</p>'

# 自定义错误处理函数
@app.errorhandler(404)
def page_not_found(error):
    """
    404错误处理
    当访问不存在的页面时，会调用这个函数
    """
    return render_template('error.html', 
                         error_code=404,
                         error_message='页面未找到',
                         error_description='您访问的页面不存在，请检查URL是否正确。'), 404

@app.errorhandler(500)
def internal_server_error(error):
    """
    500错误处理
    当服务器内部错误时，会调用这个函数
    """
    return render_template('error.html',
                         error_code=500,
                         error_message='服务器内部错误',
                         error_description='服务器遇到了一个错误，请稍后再试。'), 500

@app.errorhandler(403)
def forbidden(error):
    """403禁止访问错误"""
    return render_template('error.html',
                         error_code=403,
                         error_message='禁止访问',
                         error_description='您没有权限访问此资源。'), 403

if __name__ == '__main__':
    app.run(debug=True)
````

## 总结

### Flask核心概念回顾

1. **应用实例**：`Flask(__name__)` 创建应用
2. **路由装饰器**：`@app.route()` 定义URL路径
3. **视图函数**：处理请求并返回响应
4. **模板渲染**：使用Jinja2引擎渲染HTML
5. **静态文件**：CSS、JS、图片等放在`static`文件夹
6. **请求对象**：`request` 访问HTTP请求数据
7. **响应对象**：`jsonify()` 返回JSON，`redirect()` 重定向
8. **会话管理**：`session` 存储用户会话数据

### 计算机网络知识总结

**HTTP协议**：
- 无状态协议：每次请求独立，不保留之前的状态
- 请求-响应模式：客户端发送请求，服务器返回响应
- 状态码：200（成功）、404（未找到）、500（服务器错误）

**Web应用架构**：
- 客户端（浏览器） ↔ HTTP协议 ↔ 服务器（Flask应用）
- WSGI：Python Web应用和服务器之间的接口标准
- RESTful API：使用HTTP方法表示操作，URL表示资源
