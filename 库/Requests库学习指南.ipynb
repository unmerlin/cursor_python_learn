## Requests简介

**Requests** 是Python中最受欢迎的HTTP库，它让HTTP请求变得简单优雅。

### 主要特点：
- **人性化API**：简洁直观的接口设计
- **自动内容解码**：自动处理响应编码
- **连接池**：自动管理连接，提高性能
- **SSL/TLS支持**：内置HTTPS支持
- **会话管理**：支持Cookie和会话保持
- **文件上传**：简单易用的文件上传功能

### 为什么选择Requests？
相比Python内置的`urllib`库，Requests提供了更加Pythonic的API，代码更简洁，功能更强大。


## HTTP基础知识

在深入学习Requests之前，我们需要理解HTTP的基本概念。

### HTTP是什么？
**HTTP（HyperText Transfer Protocol）** 是互联网上应用最广泛的网络协议，用于客户端和服务器之间的通信。

### HTTP请求流程（形象化理解）

想象一下**点外卖**的过程：

1. **客户端（你）** → 发送请求（点餐）
   - 告诉服务器：我要什么（请求方法：GET/POST等）
   - 告诉服务器：我要什么菜（URL地址）
   - 告诉服务器：我的要求（请求头：User-Agent、Cookie等）
   - 告诉服务器：特殊需求（请求体：POST数据）

2. **服务器（餐厅）** → 处理请求（准备餐食）
   - 查看你的订单（解析请求）
   - 准备餐食（处理业务逻辑）
   - 打包配送（生成响应）

3. **服务器** → 返回响应（送餐）
   - 状态码：200（成功）、404（没找到）、500（服务器错误）
   - 响应头：内容类型、大小等
   - 响应体：实际的餐食（数据内容）

### 常见HTTP方法

| 方法 | 说明 | 形象比喻 |
|------|------|----------|
| GET | 获取资源 | 查看菜单 |
| POST | 提交数据 | 下单点餐 |
| PUT | 更新资源 | 修改订单 |
| DELETE | 删除资源 | 取消订单 |

### HTTP状态码分类

| 状态码范围 | 类别 | 常见示例 | 形象比喻 |
|-----------|------|---------|----------|
| 200-299 | 成功 | 200 OK | 餐食送达成功 |
| 300-399 | 重定向 | 301 永久移动 | 餐厅搬家了，新地址在... |
| 400-499 | 客户端错误 | 404 未找到 | 菜单上没有这道菜 |
| 500-599 | 服务器错误 | 500 内部错误 | 厨房出故障了 |

## 发送GET请求

GET请求用于**获取**资源，是最常用的HTTP方法。

### GET请求的特点：
- 参数通过URL传递（查询字符串）
- 请求体通常为空
- 可以被缓存
- 不应该用于敏感数据传递

### 形象化理解：
GET请求就像**查看菜单**，你只是查看信息，不会改变任何东西。


````python
# 示例1：最简单的GET请求
# 使用httpbin.org作为测试API（这是一个专门用于测试HTTP请求的网站）

response = requests.get('https://httpbin.org/get')
print("状态码:", response.status_code)
print("响应内容:")
print(response.text[:200])  # 只打印前200个字符
````

````python
# 示例2：带参数的GET请求
# 方式1：直接在URL中添加查询参数
response1 = requests.get('https://httpbin.org/get?name=张三&age=25')
print("方式1 - URL中直接添加参数:")
print(response1.json())  # json()方法将响应解析为Python字典
print()

# 方式2：使用params参数（推荐方式，更清晰）
# params参数会自动构建查询字符串，并处理特殊字符编码
params = {
    'name': '李四',
    'age': 30,
    'city': '北京'
}
response2 = requests.get('https://httpbin.org/get', params=params)
print("方式2 - 使用params参数:")
print(response2.json())
````

## 发送POST请求

POST请求用于**提交数据**到服务器，通常用于表单提交、API调用等。

### POST请求的特点：
- 数据通过请求体传递（更安全）
- 可以发送大量数据
- 不会被缓存
- 适合敏感数据传递

### 形象化理解：
POST请求就像**下单点餐**，你需要告诉餐厅你的选择（提交数据），餐厅会根据你的选择进行处理。


````python
# 示例1：发送JSON数据（最常用）
# 很多现代API都使用JSON格式进行数据交换

data = {
    'username': 'test_user',
    'email': 'test@example.com',
    'age': 25
}

# 使用json参数，requests会自动：
# 1. 将字典转换为JSON字符串
# 2. 设置Content-Type为application/json
response = requests.post('https://httpbin.org/post', json=data)
print("发送JSON数据:")
print(response.json())
print()

# 注意：json参数和data参数的区别
# - json参数：自动序列化为JSON，Content-Type为application/json
# - data参数：发送表单数据，Content-Type为application/x-www-form-urlencoded
````

````python
# 示例2：发送表单数据
# 模拟HTML表单提交

form_data = {
    'username': 'admin',
    'password': 'secret123',  # 注意：实际应用中密码应该加密传输
    'remember_me': 'true'
}

response = requests.post('https://httpbin.org/post', data=form_data)
print("发送表单数据:")
print(response.json())
print()

# 合规提醒：
# 1. 密码等敏感信息应该使用HTTPS加密传输
# 2. 不要在代码中硬编码密码，应该使用环境变量或配置文件
# 3. 考虑使用OAuth、JWT等更安全的认证方式
````

## 请求参数与查询字符串

### 查询字符串（Query String）形象化理解

查询字符串就像**点餐时的备注**：
- URL: `https://example.com/menu?dish=宫保鸡丁&spicy=中辣&rice=大份`
- `?` 后面就是查询字符串
- `&` 连接多个参数
- `=` 连接参数名和参数值

### 参数编码的重要性

URL中只能包含ASCII字符，中文字符和特殊字符需要编码（URL编码）。
Requests会自动处理编码，但理解原理很重要。


````python
# 创建示例表格展示不同参数传递方式
import pandas as pd

# 创建对比表格
comparison_data = {
    '方式': ['直接在URL中', '使用params参数', '手动编码'],
    '代码示例': [
        "requests.get('url?key=value')",
        "requests.get('url', params={'key': 'value'})",
        "requests.get('url', params=urllib.parse.urlencode({'key': 'value'}))"
    ],
    '优点': [
        '简单直接',
        '自动编码，推荐使用',
        '完全控制'
    ],
    '缺点': [
        '需要手动编码特殊字符',
        '无',
        '代码复杂'
    ],
    '适用场景': [
        '简单参数',
        '大多数情况（推荐）',
        '特殊需求'
    ]
}

df = pd.DataFrame(comparison_data)
print("参数传递方式对比表：")
print(df.to_string(index=False))
print()

# 实际示例：展示自动编码的好处
print("=" * 60)
print("实际示例：中文字符自动编码")
print("=" * 60)

# 使用params参数，自动处理中文编码
params = {
    'keyword': 'Python学习',
    'category': '编程教程',
    'page': 1
}

response = requests.get('https://httpbin.org/get', params=params)
result = response.json()

print("发送的参数（原始）:")
for key, value in params.items():
    print(f"  {key}: {value}")

print("\n实际发送的URL（已编码）:")
print(f"  {response.url}")

# 注意：URL编码将中文字符转换为%XX格式
# 例如："Python学习" 会被编码为 "Python%E5%AD%A6%E4%B9%A0"
# Requests会自动处理这个过程，我们不需要手动编码
````

## 请求头设置

### 请求头（Headers）形象化理解

请求头就像**点餐时的身份信息**：
- **User-Agent**：告诉服务器你是什么"设备"（浏览器、手机App等）
- **Accept**：告诉服务器你能接受什么格式的数据（JSON、HTML等）
- **Authorization**：身份认证信息（用户名密码、Token等）
- **Content-Type**：告诉服务器你发送的数据是什么格式

### 为什么需要设置请求头？

1. **反爬虫机制**：很多网站会检查User-Agent，没有的话可能被拒绝
2. **API要求**：很多API要求特定的请求头
3. **内容协商**：告诉服务器你希望接收什么格式的数据


````python
# 示例1：设置单个请求头
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
}

response = requests.get('https://httpbin.org/headers', headers=headers)
print("设置的请求头:")
print(response.json())
print()

# 合规提醒：
# 1. 使用真实的User-Agent是合法的，但不要伪装成其他软件进行恶意行为
# 2. 遵守网站的robots.txt协议
# 3. 不要过于频繁地请求，避免对服务器造成压力
````

````python
# 示例2：设置多个请求头（常见场景）
# 创建请求头对比表格

headers_examples = {
    '场景': [
        '访问普通网站',
        '调用REST API',
        '上传文件',
        '身份认证'
    ],
    '常用请求头': [
        'User-Agent',
        'Content-Type, Accept',
        'Content-Type: multipart/form-data',
        'Authorization'
    ],
    '说明': [
        '标识客户端类型',
        '指定数据格式',
        '文件上传格式',
        '认证令牌'
    ]
}

df_headers = pd.DataFrame(headers_examples)
print("不同场景的请求头设置：")
print(df_headers.to_string(index=False))
print()

# 实际示例：API调用时的请求头设置
api_headers = {
    'User-Agent': 'MyApp/1.0',
    'Accept': 'application/json',  # 告诉服务器我希望接收JSON格式
    'Content-Type': 'application/json',  # 告诉服务器我发送的是JSON格式
    'Authorization': 'Bearer your_token_here'  # 身份认证令牌
}

# 注意：Authorization有多种方式
# 1. Bearer Token: 'Bearer your_token'
# 2. Basic Auth: 'Basic base64(username:password)'
# 3. API Key: 'ApiKey your_api_key'

print("API调用示例请求头:")
for key, value in api_headers.items():
    print(f"  {key}: {value}")
````

## 状态码处理

### HTTP状态码形象化理解

状态码就像**外卖订单的状态**：
- **200系列**：订单成功，餐食正在配送
- **300系列**：餐厅搬家了，新地址是...
- **400系列**：你的订单有问题（地址错误、支付失败等）
- **500系列**：餐厅出故障了（厨房设备坏了等）

### 常见状态码速查表

| 状态码 | 名称 | 含义 | 处理建议 |
|--------|------|------|----------|
| 200 | OK | 请求成功 | 正常处理响应 |
| 201 | Created | 资源创建成功 | 通常用于POST请求 |
| 301 | Moved Permanently | 永久重定向 | 更新URL |
| 302 | Found | 临时重定向 | 跟随重定向 |
| 400 | Bad Request | 请求错误 | 检查请求参数 |
| 401 | Unauthorized | 未授权 | 检查认证信息 |
| 403 | Forbidden | 禁止访问 | 检查权限 |
| 404 | Not Found | 资源不存在 | 检查URL |
| 500 | Internal Server Error | 服务器错误 | 稍后重试 |
| 503 | Service Unavailable | 服务不可用 | 稍后重试 |


````python
# 示例1：检查状态码
response = requests.get('https://httpbin.org/status/200')

# 方式1：直接检查状态码
if response.status_code == 200:
    print("请求成功！")
else:
    print(f"请求失败，状态码: {response.status_code}")

# 方式2：使用response.ok属性（200-299范围内为True）
if response.ok:
    print("请求成功（使用ok属性）")
else:
    print("请求失败")

# 方式3：使用raise_for_status()方法（推荐）
# 如果状态码表示错误，会自动抛出异常
try:
    response = requests.get('https://httpbin.org/status/404')
    response.raise_for_status()  # 如果是4xx或5xx，会抛出HTTPError异常
except requests.exceptions.HTTPError as e:
    print(f"HTTP错误: {e}")
````

````python
# 示例2：不同状态码的处理策略
# 创建状态码处理策略表格

status_handling = {
    '状态码范围': [
        '200-299',
        '300-399',
        '400-499',
        '500-599'
    ],
    '含义': [
        '成功',
        '重定向',
        '客户端错误',
        '服务器错误'
    ],
    '处理策略': [
        '正常处理响应数据',
        '跟随重定向（Requests默认自动处理）',
        '检查请求参数、认证信息等',
        '记录错误，稍后重试'
    ],
    '代码示例': [
        "if response.ok: process_data()",
        "response = requests.get(url, allow_redirects=True)",
        "检查参数、重新认证",
        "time.sleep(5); retry()"
    ]
}

df_status = pd.DataFrame(status_handling)
print("状态码处理策略：")
print(df_status.to_string(index=False))
print()

# 实际示例：完整的错误处理
def make_request_with_retry(url, max_retries=3):
    """
    带重试机制的请求函数
    
    参数:
        url: 请求的URL
        max_retries: 最大重试次数
    """
    for attempt in range(max_retries):
        try:
            response = requests.get(url, timeout=5)
            response.raise_for_status()  # 如果是错误状态码，抛出异常
            return response
        except requests.exceptions.Timeout:
            print(f"请求超时，尝试 {attempt + 1}/{max_retries}")
        except requests.exceptions.HTTPError as e:
            if response.status_code >= 500:  # 服务器错误，可以重试
                print(f"服务器错误 {response.status_code}，尝试 {attempt + 1}/{max_retries}")
            else:  # 客户端错误，不应该重试
                print(f"客户端错误: {e}")
                raise
        except requests.exceptions.RequestException as e:
            print(f"请求异常: {e}，尝试 {attempt + 1}/{max_retries}")
        
        if attempt < max_retries - 1:
            import time
            time.sleep(2 ** attempt)  # 指数退避：2秒、4秒、8秒...
    
    raise Exception("请求失败，已达到最大重试次数")

# 测试（使用一个会返回500错误的URL）
# response = make_request_with_retry('https://httpbin.org/status/500')
````

## 异常处理

### Requests异常体系形象化理解

异常就像**点外卖时遇到的各种问题**：
- **ConnectionError**：网络断了，无法连接到餐厅
- **Timeout**：等太久了，超时了
- **HTTPError**：餐厅回复了，但是说订单有问题（4xx、5xx状态码）
- **RequestException**：所有请求异常的基类

### 异常处理的重要性

良好的异常处理可以让程序更加健壮，避免因为网络问题导致程序崩溃。


````python
# 创建异常类型说明表格
exceptions_info = {
    '异常类型': [
        'RequestException',
        'ConnectionError',
        'HTTPError',
        'Timeout',
        'TooManyRedirects',
        'URLRequired',
        'InvalidURL'
    ],
    '说明': [
        '所有请求异常的基类',
        '连接错误（网络问题、DNS解析失败等）',
        'HTTP错误（4xx、5xx状态码）',
        '请求超时',
        '重定向次数过多',
        '缺少必需的URL',
        '无效的URL格式'
    ],
    '常见原因': [
        '基类，不直接使用',
        '网络断开、服务器不可达',
        '服务器返回错误状态码',
        '服务器响应太慢',
        '重定向循环',
        '代码错误',
        'URL格式错误'
    ]
}

df_exceptions = pd.DataFrame(exceptions_info)
print("Requests异常类型说明：")
print(df_exceptions.to_string(index=False))
print()

# 实际示例：完整的异常处理
def safe_request(url):
    """
    安全的请求函数，包含完整的异常处理
    """
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # 检查HTTP错误
        return response.json()
    
    except requests.exceptions.Timeout:
        print("错误：请求超时，服务器响应太慢")
        return None
    
    except requests.exceptions.ConnectionError:
        print("错误：无法连接到服务器，请检查网络连接")
        return None
    
    except requests.exceptions.HTTPError as e:
        print(f"错误：HTTP错误 {e.response.status_code}")
        return None
    
    except requests.exceptions.RequestException as e:
        print(f"错误：请求异常 {e}")
        return None
    
    except ValueError:  # json()解析失败
        print("错误：响应不是有效的JSON格式")
        return None

# 测试
print("=" * 60)
print("异常处理测试")
print("=" * 60)
result = safe_request('https://httpbin.org/json')
if result:
    print("请求成功！")
    print(result)
````

## 会话管理

### Session对象形象化理解

Session就像**在餐厅办会员卡**：
- 第一次访问时，服务器给你一张"会员卡"（Session ID）
- 之后的每次请求都带上这张"会员卡"
- 服务器通过"会员卡"识别你，保持你的登录状态、购物车等

### Session vs 普通请求的区别

| 特性 | 普通请求 | Session |
|------|----------|---------|
| Cookie管理 | 每次请求独立 | 自动保持Cookie |
| 连接复用 | 每次新建连接 | 复用连接（性能更好）|
| 适用场景 | 单次请求 | 多次相关请求 |

### 使用Session的好处：
1. **自动管理Cookie**：登录后自动保持会话
2. **连接池**：复用TCP连接，提高性能
3. **默认参数**：可以为所有请求设置默认的headers、auth等


````python
# 示例1：使用Session保持Cookie
# 创建Session对象
session = requests.Session()

# 第一次请求：设置Cookie
response1 = session.get('https://httpbin.org/cookies/set/session_id/abc123')
print("第一次请求后的Cookie:")
print(session.cookies.get_dict())
print()

# 第二次请求：自动携带Cookie
response2 = session.get('https://httpbin.org/cookies')
print("第二次请求，服务器收到的Cookie:")
print(response2.json())
print()

# 注意：Session会自动管理Cookie，不需要手动设置
# 这在登录网站后保持会话状态非常有用
````

````python
# 示例2：Session设置默认参数
# 创建Session并设置默认请求头和认证信息

session = requests.Session()

# 设置默认请求头（所有通过这个Session的请求都会自动包含这些头）
session.headers.update({
    'User-Agent': 'MyApp/1.0',
    'Accept': 'application/json'
})

# 设置默认认证（如果需要）
# session.auth = ('username', 'password')  # Basic Auth
# 或者
# session.headers['Authorization'] = 'Bearer your_token'

# 现在所有请求都会自动包含这些默认设置
response = session.get('https://httpbin.org/headers')
print("请求头（包含默认设置）:")
print(response.json()['headers'])
print()

# 创建对比表格
session_comparison = {
    '场景': [
        '单次请求',
        '多次独立请求',
        '需要保持会话',
        '需要设置默认参数'
    ],
    '使用方式': [
        'requests.get()',
        '多次调用requests.get()',
        '使用Session',
        '使用Session'
    ],
    '原因': [
        '简单直接',
        '每次请求独立，互不影响',
        '自动管理Cookie和连接',
        '避免重复设置参数'
    ]
}

df_session = pd.DataFrame(session_comparison)
print("Session使用场景对比：")
print(df_session.to_string(index=False))
````

````python
# 示例2：文件下载
# 下载一个小文件（httpbin.org返回的JSON数据）

url = 'https://httpbin.org/json'
response = requests.get(url)

# 方式1：保存为文本文件
with open('downloaded_data.json', 'w', encoding='utf-8') as f:
    f.write(response.text)
print("文件已保存为 downloaded_data.json")

# 方式2：保存二进制文件（如图片）
# 使用content属性获取原始字节
image_url = 'https://httpbin.org/image/png'
image_response = requests.get(image_url)

# 注意：对于大文件，应该使用流式下载
def download_file_streaming(url, filename):
    """
    流式下载大文件，避免内存占用过大
    """
    response = requests.get(url, stream=True)  # stream=True启用流式下载
    
    # 检查响应状态
    response.raise_for_status()
    
    # 分块写入文件
    with open(filename, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8192):  # 每次读取8KB
            if chunk:  # 过滤掉keep-alive的空块
                f.write(chunk)
    
    print(f"文件已下载: {filename}")

# 创建文件操作对比表格
file_ops = {
    '操作': [
        '上传小文件',
        '上传大文件',
        '下载小文件',
        '下载大文件'
    ],
    '方法': [
        'requests.post(url, files=files)',
        '使用stream=True，分块上传',
        'response.content 或 response.text',
        '使用stream=True，iter_content()'
    ],
    '注意事项': [
        '检查文件大小和类型',
        '显示上传进度，处理超时',
        '直接保存到内存',
        '分块保存，避免内存溢出'
    ]
}

df_files = pd.DataFrame(file_ops)
print("\n文件操作方式对比：")
print(df_files.to_string(index=False))
````

## 超时设置

### 超时设置形象化理解

超时就像**点外卖的等待时间限制**：
- 如果餐厅30分钟还没送到，你可能就取消订单了
- 超时设置可以防止程序无限等待，提高用户体验

### 超时参数说明：
- **timeout=5**：连接和读取总共5秒超时
- **timeout=(2, 5)**：连接2秒超时，读取5秒超时

### 为什么需要设置超时？
1. **避免程序挂起**：网络问题时程序不会无限等待
2. **提高用户体验**：快速失败，给用户反馈
3. **资源管理**：释放连接资源


````python
# 示例1：设置简单超时
try:
    # 设置5秒超时
    response = requests.get('https://httpbin.org/delay/2', timeout=5)
    print("请求成功（2秒延迟，5秒超时）")
except requests.exceptions.Timeout:
    print("请求超时")

# 示例2：分别设置连接和读取超时
try:
    # (连接超时, 读取超时)
    # 连接超时：建立连接的最大时间
    # 读取超时：服务器响应数据的最大时间
    response = requests.get('https://httpbin.org/delay/1', timeout=(2, 5))
    print("请求成功")
except requests.exceptions.ConnectTimeout:
    print("连接超时：无法在2秒内建立连接")
except requests.exceptions.ReadTimeout:
    print("读取超时：服务器在5秒内没有响应数据")

# 创建超时设置对比表格
timeout_comparison = {
    '设置方式': [
        'timeout=5',
        'timeout=(2, 5)',
        'timeout=None'
    ],
    '含义': [
        '连接+读取总共5秒',
        '连接2秒，读取5秒',
        '不设置超时（不推荐）'
    ],
    '适用场景': [
        '大多数情况',
        '需要精确控制',
        '仅用于测试（生产环境不推荐）'
    ],
    '注意事项': [
        '简单易用',
        '更灵活的控制',
        '可能导致程序挂起'
    ]
}

df_timeout = pd.DataFrame(timeout_comparison)
print("超时设置方式对比：")
print(df_timeout.to_string(index=False))
print()

# 最佳实践：始终设置超时
print("最佳实践：")
print("1. 始终设置合理的超时时间（通常5-30秒）")
print("2. 对于可能较慢的操作，设置更长的超时")
print("3. 对于关键操作，实现重试机制")
print("4. 监控超时情况，优化网络或服务器性能")
````

## 实践练习

### 综合练习：构建一个简单的API客户端

让我们通过一个完整的例子，综合运用前面学到的知识：
1. 使用Session管理会话
2. 设置请求头和超时
3. 处理异常和状态码
4. 解析JSON响应


````python
# 综合示例：创建一个简单的API客户端类
class SimpleAPIClient:
    """
    简单的API客户端示例
    演示如何使用Requests库构建一个实用的API客户端
    """
    
    def __init__(self, base_url, timeout=10):
        """
        初始化API客户端
        
        参数:
            base_url: API的基础URL
            timeout: 默认超时时间（秒）
        """
        self.base_url = base_url.rstrip('/')  # 移除末尾的斜杠
        self.timeout = timeout
        
        # 创建Session，复用连接
        self.session = requests.Session()
        
        # 设置默认请求头
        self.session.headers.update({
            'User-Agent': 'SimpleAPIClient/1.0',
            'Accept': 'application/json',
            'Content-Type': 'application/json'
        })
    
    def get(self, endpoint, params=None, **kwargs):
        """
        发送GET请求
        
        参数:
            endpoint: API端点（相对于base_url）
            params: 查询参数
            **kwargs: 其他requests参数
        """
        url = f"{self.base_url}/{endpoint.lstrip('/')}"
        
        try:
            response = self.session.get(
                url, 
                params=params, 
                timeout=self.timeout,
                **kwargs
            )
            response.raise_for_status()  # 检查HTTP错误
            return response.json()  # 返回JSON数据
        except requests.exceptions.RequestException as e:
            print(f"请求失败: {e}")
            return None
    
    def post(self, endpoint, data=None, json_data=None, **kwargs):
        """
        发送POST请求
        
        参数:
            endpoint: API端点
            data: 表单数据
            json_data: JSON数据
            **kwargs: 其他requests参数
        """
        url = f"{self.base_url}/{endpoint.lstrip('/')}"
        
        try:
            if json_data:
                response = self.session.post(
                    url,
                    json=json_data,
                    timeout=self.timeout,
                    **kwargs
                )
            else:
                response = self.session.post(
                    url,
                    data=data,
                    timeout=self.timeout,
                    **kwargs
                )
            response.raise_for_status()
            return response.json()
        except requests.exceptions.RequestException as e:
            print(f"请求失败: {e}")
            return None
    
    def close(self):
        """关闭Session，释放资源"""
        self.session.close()

# 使用示例
print("=" * 60)
print("API客户端使用示例")
print("=" * 60)

# 创建客户端实例
client = SimpleAPIClient('https://httpbin.org', timeout=5)

# GET请求示例
print("\n1. GET请求示例：")
result = client.get('get', params={'name': '测试', 'page': 1})
if result:
    print(f"请求成功，URL: {result.get('url')}")

# POST请求示例
print("\n2. POST请求示例：")
result = client.post('post', json_data={'message': 'Hello, API!'})
if result:
    print(f"请求成功，数据: {result.get('json')}")

# 关闭客户端
client.close()
print("\n客户端已关闭")
````
