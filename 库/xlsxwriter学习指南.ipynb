## æ ¼å¼åŒ–ï¼šç¾åŒ–ä½ çš„è¡¨æ ¼

### æ ¼å¼åŒ–å¯¹è±¡ï¼ˆFormatï¼‰

æ ¼å¼åŒ–æ˜¯XlsxWriterçš„æ ¸å¿ƒä¼˜åŠ¿ä¹‹ä¸€ï¼Œå¯ä»¥ç²¾ç¡®æ§åˆ¶ï¼š
- å­—ä½“ï¼ˆå¤§å°ã€é¢œè‰²ã€ç²—ä½“ã€æ–œä½“ï¼‰
- å¯¹é½æ–¹å¼ï¼ˆæ°´å¹³ã€å‚ç›´ï¼‰
- è¾¹æ¡†ï¼ˆæ ·å¼ã€é¢œè‰²ï¼‰
- èƒŒæ™¯è‰²å’Œå›¾æ¡ˆ
- æ•°å­—æ ¼å¼ï¼ˆè´§å¸ã€ç™¾åˆ†æ¯”ã€æ—¥æœŸç­‰ï¼‰


````python
# åˆ›å»ºæ ¼å¼åŒ–æ ·å¼

workbook = xlsxwriter.Workbook('ç¤ºä¾‹4_æ ¼å¼åŒ–è¡¨æ ¼.xlsx')
worksheet = workbook.add_worksheet('æ ¼å¼åŒ–ç¤ºä¾‹')

# å®šä¹‰å„ç§æ ¼å¼
header_format = workbook.add_format({
    'bold': True,              # ç²—ä½“
    'bg_color': '#366092',     # èƒŒæ™¯è‰²ï¼ˆè“è‰²ï¼‰
    'font_color': '#FFFFFF',   # å­—ä½“é¢œè‰²ï¼ˆç™½è‰²ï¼‰
    'align': 'center',         # æ°´å¹³å±…ä¸­
    'valign': 'vcenter',       # å‚ç›´å±…ä¸­
    'border': 1,               # è¾¹æ¡†
    'font_size': 12,           # å­—ä½“å¤§å°
})

currency_format = workbook.add_format({
    'num_format': 'Â¥#,##0.00',  # è´§å¸æ ¼å¼
    'align': 'right',
})

percent_format = workbook.add_format({
    'num_format': '0.00%',      # ç™¾åˆ†æ¯”æ ¼å¼
    'align': 'right',
})

date_format = workbook.add_format({
    'num_format': 'yyyy-mm-dd', # æ—¥æœŸæ ¼å¼
    'align': 'center',
})

# å†™å…¥è¡¨å¤´
headers = ['æ—¥æœŸ', 'äº§å“', 'é”€å”®é¢', 'å¢é•¿ç‡']
worksheet.write_row(0, 0, headers, header_format)

# å†™å…¥æ•°æ®
data = [
    ['2024-01-15', 'äº§å“A', 50000, 0.15],
    ['2024-01-16', 'äº§å“B', 75000, 0.23],
    ['2024-01-17', 'äº§å“C', 60000, 0.18],
]

for row_num, row_data in enumerate(data, start=1):
    worksheet.write(row_num, 0, row_data[0], date_format)
    worksheet.write(row_num, 1, row_data[1])
    worksheet.write(row_num, 2, row_data[2], currency_format)
    worksheet.write(row_num, 3, row_data[3], percent_format)

# è®¾ç½®åˆ—å®½
worksheet.set_column('A:A', 12)  # æ—¥æœŸåˆ—
worksheet.set_column('B:B', 10)  # äº§å“åˆ—
worksheet.set_column('C:C', 12)  # é”€å”®é¢åˆ—
worksheet.set_column('D:D', 10)  # å¢é•¿ç‡åˆ—

workbook.close()
print("âœ… æ ¼å¼åŒ–è¡¨æ ¼åˆ›å»ºæˆåŠŸï¼")
````

### çŸ¥è¯†ç‚¹4ï¼šæ ¼å¼åŒ–è¯¦è§£

**æ ¼å¼åŒ–å¯¹è±¡çš„ç‰¹ç‚¹**ï¼š
- ğŸ¨ **å¯å¤ç”¨**ï¼šåˆ›å»ºä¸€ä¸ªæ ¼å¼å¯¹è±¡ï¼Œå¯ä»¥åœ¨å¤šä¸ªå•å…ƒæ ¼ä¸­ä½¿ç”¨
- âš¡ **é«˜æ•ˆ**ï¼šæ ¼å¼å¯¹è±¡åœ¨å†…éƒ¨è¢«ç¼“å­˜ï¼Œä¸ä¼šé‡å¤åˆ›å»º

| å±æ€§ | è¯´æ˜ | ç¤ºä¾‹å€¼ |
|------|------|--------|

| `align` | æ°´å¹³å¯¹é½ | `'left'`, `'center'`, `'right'` |

| `border` | è¾¹æ¡† | `1` (æœ‰è¾¹æ¡†) / `0` (æ— è¾¹æ¡†) |
| `num_format` | æ•°å­—æ ¼å¼ | `'Â¥#,##0.00'` (è´§å¸) |

**ç›¸æ¯”Excelçš„ä¼˜åŠ¿**ï¼š
- åœ¨Excelä¸­ï¼Œéœ€è¦é€ä¸ªå•å…ƒæ ¼è®¾ç½®æ ¼å¼ï¼Œè€—æ—¶ä¸”å®¹æ˜“å‡ºé”™
- ä½¿ç”¨ä»£ç å¯ä»¥ä¸€æ¬¡æ€§åº”ç”¨æ ¼å¼åˆ°æ•´è¡Œ/æ•´åˆ—ï¼Œä¿è¯æ ¼å¼ä¸€è‡´æ€§


````python
# é«˜çº§æ ¼å¼åŒ–ï¼šåˆå¹¶å•å…ƒæ ¼ã€è®¾ç½®è¡Œé«˜

workbook = xlsxwriter.Workbook('ç¤ºä¾‹5_é«˜çº§æ ¼å¼åŒ–.xlsx')
worksheet = workbook.add_worksheet('é«˜çº§æ ¼å¼')

# åˆ›å»ºæ ¼å¼
title_format = workbook.add_format({
    'bold': True,
    'font_size': 16,
    'align': 'center',
    'valign': 'vcenter',
    'bg_color': '#4472C4',
    'font_color': '#FFFFFF',
})

section_format = workbook.add_format({
    'bold': True,
    'bg_color': '#D9E1F2',
    'border': 1,
})

# åˆå¹¶å•å…ƒæ ¼å¹¶å†™å…¥æ ‡é¢˜
worksheet.merge_range('A1:D1', '2024å¹´ç¬¬ä¸€å­£åº¦é”€å”®æŠ¥è¡¨', title_format)

# å†™å…¥è¡¨å¤´
headers = ['æœˆä»½', 'é”€å”®é¢', 'æˆæœ¬', 'åˆ©æ¶¦']
header_format = workbook.add_format({
    'bold': True,
    'bg_color': '#B4C6E7',
    'border': 1,
    'align': 'center',
})
worksheet.write_row(2, 0, headers, header_format)

# å†™å…¥æ•°æ®
data = [
    ['1æœˆ', 500000, 300000, 200000],
    ['2æœˆ', 550000, 320000, 230000],
    ['3æœˆ', 600000, 350000, 250000],
]

data_format = workbook.add_format({
    'border': 1,
    'num_format': '#,##0',
})

for row_num, row_data in enumerate(data, start=3):
    for col_num, value in enumerate(row_data):
        if col_num == 0:
            worksheet.write(row_num, col_num, value, section_format)
        else:
            worksheet.write(row_num, col_num, value, data_format)

# è®¾ç½®è¡Œé«˜å’Œåˆ—å®½
worksheet.set_row(0, 30)  # æ ‡é¢˜è¡Œé«˜åº¦
worksheet.set_row(2, 25)  # è¡¨å¤´è¡Œé«˜åº¦
worksheet.set_column('A:D', 15)

workbook.close()
print("âœ… é«˜çº§æ ¼å¼åŒ–è¡¨æ ¼åˆ›å»ºæˆåŠŸï¼")
````

## æ‰¹é‡æ“ä½œï¼šé«˜æ•ˆå¤„ç†å¤§é‡æ•°æ®

XlsxWriterçš„**æ‰¹é‡æ“ä½œèƒ½åŠ›**æ˜¯å…¶æ ¸å¿ƒä¼˜åŠ¿ä¹‹ä¸€ï¼Œç‰¹åˆ«é€‚åˆå¤„ç†å¤§é‡æ•°æ®å’Œæ‰¹é‡ç”ŸæˆæŠ¥è¡¨ã€‚


````python
# æ‰¹é‡ç”Ÿæˆå¤šä¸ªæœˆä»½çš„æŠ¥è¡¨

import calendar
from datetime import datetime

workbook = xlsxwriter.Workbook('ç¤ºä¾‹9_æ‰¹é‡ç”ŸæˆæŠ¥è¡¨.xlsx')

# å®šä¹‰æ ¼å¼
header_format = workbook.add_format({
    'bold': True,
    'bg_color': '#4472C4',
    'font_color': '#FFFFFF',
    'align': 'center',
    'border': 1,
})

data_format = workbook.add_format({
    'border': 1,
    'num_format': '#,##0',
})

# æ¨¡æ‹Ÿé”€å”®æ•°æ®ï¼ˆå®é™…åº”ç”¨ä¸­ä»æ•°æ®åº“æˆ–APIè·å–ï¼‰
def generate_monthly_data(month):
    """ç”ŸæˆæŸä¸ªæœˆä»½çš„æ¨¡æ‹Ÿæ•°æ®"""
    base_sales = 50000
    days = calendar.monthrange(2024, month)[1]  # è·å–è¯¥æœˆå¤©æ•°
    data = []
    for day in range(1, days + 1):
        # æ¨¡æ‹Ÿæ¯æ—¥é”€å”®é¢ï¼ˆæœ‰æ³¢åŠ¨ï¼‰
        sales = base_sales + (day % 7) * 5000 - 10000
        data.append([f'2024-{month:02d}-{day:02d}', sales])
    return data

# ä¸ºæ¯ä¸ªæœˆåˆ›å»ºä¸€ä¸ªå·¥ä½œè¡¨
for month in range(1, 4):  # ç”Ÿæˆ1-3æœˆçš„æŠ¥è¡¨
    month_name = f'{month}æœˆ'
    worksheet = workbook.add_worksheet(month_name)
    
    # å†™å…¥æ ‡é¢˜
    title_format = workbook.add_format({
        'bold': True,
        'font_size': 14,
        'align': 'center',
    })
    worksheet.merge_range(0, 0, 0, 2, f'2024å¹´{month_name}é”€å”®æ—¥æŠ¥', title_format)
    
    # å†™å…¥è¡¨å¤´
    headers = ['æ—¥æœŸ', 'é”€å”®é¢', 'ç´¯è®¡']
    worksheet.write_row(2, 0, headers, header_format)
    
    # å†™å…¥æ•°æ®
    monthly_data = generate_monthly_data(month)
    cumulative = 0
    for row_num, (date, sales) in enumerate(monthly_data, start=3):
        cumulative += sales
        worksheet.write(row_num, 0, date)
        worksheet.write(row_num, 1, sales, data_format)
        worksheet.write(row_num, 2, cumulative, data_format)
    
    # å†™å…¥æ€»è®¡
    total_row = len(monthly_data) + 3
    total_format = workbook.add_format({
        'bold': True,
        'bg_color': '#FFC000',
        'border': 1,
    })
    worksheet.write(total_row, 0, 'æ€»è®¡', total_format)
    worksheet.write_formula(total_row, 1, f'=SUM(B4:B{total_row})', total_format)
    worksheet.write(total_row, 2, cumulative, total_format)
    
    # è®¾ç½®åˆ—å®½
    worksheet.set_column('A:A', 12)
    worksheet.set_column('B:C', 15)

workbook.close()
print("âœ… æ‰¹é‡ç”Ÿæˆå¤šä¸ªæœˆä»½æŠ¥è¡¨æˆåŠŸï¼")
````

````python
# æ¼”ç¤ºï¼šPandas + XlsxWriter ç»„åˆä½¿ç”¨ï¼ˆæœ€ä½³å®è·µï¼‰

import pandas as pd
import xlsxwriter

# ä½¿ç”¨Pandaså¤„ç†æ•°æ®
data = {
    'äº§å“': ['äº§å“A', 'äº§å“B', 'äº§å“C', 'äº§å“D'],
    'é”€å”®é¢': [150000, 120000, 100000, 50000],
    'æˆæœ¬': [90000, 72000, 60000, 30000],
}

df = pd.DataFrame(data)
df['åˆ©æ¶¦'] = df['é”€å”®é¢'] - df['æˆæœ¬']
df['åˆ©æ¶¦ç‡'] = df['åˆ©æ¶¦'] / df['é”€å”®é¢']

print("Pandaså¤„ç†åçš„æ•°æ®ï¼š")
print(df)
print()

# ä½¿ç”¨XlsxWriteråˆ›å»ºæ ¼å¼åŒ–çš„Excelæ–‡ä»¶
workbook = xlsxwriter.Workbook('ç¤ºä¾‹13_Pandas+XlsxWriter.xlsx')
worksheet = workbook.add_worksheet('é”€å”®åˆ†æ')

# å®šä¹‰æ ¼å¼
header_format = workbook.add_format({
    'bold': True,
    'bg_color': '#4472C4',
    'font_color': '#FFFFFF',
    'align': 'center',
    'border': 1,
})

currency_format = workbook.add_format({
    'num_format': 'Â¥#,##0',
    'border': 1,
})

percent_format = workbook.add_format({
    'num_format': '0.00%',
    'border': 1,
})

# å†™å…¥è¡¨å¤´
worksheet.write_row(0, 0, df.columns.tolist(), header_format)

# å†™å…¥æ•°æ®
for row_num, (index, row) in enumerate(df.iterrows(), start=1):
    worksheet.write(row_num, 0, row['äº§å“'])
    worksheet.write(row_num, 1, row['é”€å”®é¢'], currency_format)
    worksheet.write(row_num, 2, row['æˆæœ¬'], currency_format)
    worksheet.write(row_num, 3, row['åˆ©æ¶¦'], currency_format)
    worksheet.write(row_num, 4, row['åˆ©æ¶¦ç‡'], percent_format)

# æ·»åŠ æ€»è®¡è¡Œ
total_row = len(df) + 1
total_format = workbook.add_format({
    'bold': True,
    'bg_color': '#FFC000',
    'border': 1,
})

worksheet.write(total_row, 0, 'æ€»è®¡', total_format)
worksheet.write_formula(total_row, 1, f'=SUM(B2:B{total_row})', total_format)
worksheet.write_formula(total_row, 2, f'=SUM(C2:C{total_row})', total_format)
worksheet.write_formula(total_row, 3, f'=SUM(D2:D{total_row})', total_format)

# è®¾ç½®åˆ—å®½
worksheet.set_column('A:E', 15)

workbook.close()
print("âœ… Pandasæ•°æ®å¤„ç† + XlsxWriteræ ¼å¼åŒ–è¾“å‡ºå®Œæˆï¼")
````

````python
# å®æˆ˜æ¡ˆä¾‹ï¼šç”Ÿæˆæœˆåº¦é”€å”®åˆ†ææŠ¥è¡¨

import xlsxwriter
from datetime import datetime, timedelta
import random

# æ¨¡æ‹Ÿç”Ÿæˆé”€å”®æ•°æ®
def generate_sales_data(year, month):
    """ç”ŸæˆæŒ‡å®šå¹´æœˆçš„é”€å”®æ•°æ®"""
    days = (datetime(year, month+1, 1) - timedelta(days=1)).day
    data = []
    base_sales = 50000
    
    for day in range(1, days + 1):
        # æ¨¡æ‹Ÿæ¯æ—¥é”€å”®æ•°æ®ï¼ˆæœ‰æ³¢åŠ¨ï¼‰
        sales = base_sales + random.randint(-10000, 15000)
        cost = int(sales * 0.6)
        profit = sales - cost
        data.append({
            'date': f'{year}-{month:02d}-{day:02d}',
            'sales': sales,
            'cost': cost,
            'profit': profit,
            'customers': random.randint(50, 150),
        })
    return data

# ç”Ÿæˆ2024å¹´1æœˆçš„æ•°æ®
sales_data = generate_sales_data(2024, 1)

# åˆ›å»ºExcelå·¥ä½œç°¿
workbook = xlsxwriter.Workbook('å®æˆ˜æ¡ˆä¾‹_æœˆåº¦é”€å”®æŠ¥è¡¨.xlsx')

# ========== åˆ›å»ºæ•°æ®æ˜ç»†è¡¨ ==========
detail_sheet = workbook.add_worksheet('æ¯æ—¥æ˜ç»†')

# å®šä¹‰æ ¼å¼
title_format = workbook.add_format({
    'bold': True,
    'font_size': 16,
    'align': 'center',
    'valign': 'vcenter',
    'bg_color': '#4472C4',
    'font_color': '#FFFFFF',
})

header_format = workbook.add_format({
    'bold': True,
    'bg_color': '#B4C6E7',
    'border': 1,
    'align': 'center',
})

currency_format = workbook.add_format({
    'num_format': 'Â¥#,##0',
    'border': 1,
})

number_format = workbook.add_format({
    'num_format': '#,##0',
    'border': 1,
})

# å†™å…¥æ ‡é¢˜
detail_sheet.merge_range('A1:F1', '2024å¹´1æœˆé”€å”®æ˜ç»†è¡¨', title_format)
detail_sheet.set_row(0, 30)

# å†™å…¥è¡¨å¤´
headers = ['æ—¥æœŸ', 'é”€å”®é¢', 'æˆæœ¬', 'åˆ©æ¶¦', 'å®¢æˆ·æ•°', 'å®¢å•ä»·']
detail_sheet.write_row(2, 0, headers, header_format)

# å†™å…¥æ•°æ®
for row_num, record in enumerate(sales_data, start=3):
    detail_sheet.write(row_num, 0, record['date'])
    detail_sheet.write(row_num, 1, record['sales'], currency_format)
    detail_sheet.write(row_num, 2, record['cost'], currency_format)
    detail_sheet.write(row_num, 3, record['profit'], currency_format)
    detail_sheet.write(row_num, 4, record['customers'], number_format)
    # è®¡ç®—å®¢å•ä»·
    detail_sheet.write_formula(row_num, 5, f'=B{row_num+1}/E{row_num+1}', currency_format)

# æ·»åŠ æ¡ä»¶æ ¼å¼ï¼šåˆ©æ¶¦å¤§äº20000çš„æ˜¾ç¤ºç»¿è‰²
detail_sheet.conditional_format('D4:D34', {
    'type': 'cell',
    'criteria': '>=',
    'value': 20000,
    'format': workbook.add_format({'bg_color': '#C6EFCE'})
})

# æ·»åŠ æ€»è®¡è¡Œ
total_row = len(sales_data) + 3
total_format = workbook.add_format({
    'bold': True,
    'bg_color': '#FFC000',
    'border': 1,
    'num_format': 'Â¥#,##0',
})

detail_sheet.write(total_row, 0, 'æ€»è®¡', total_format)
for col in range(1, 4):
    detail_sheet.write_formula(total_row, col, f'=SUM({chr(65+col)}{4}:{chr(65+col)}{total_row})', total_format)
detail_sheet.write_formula(total_row, 4, f'=SUM(E4:E{total_row})', total_format)
detail_sheet.write_formula(total_row, 5, f'=B{total_row+1}/E{total_row+1}', total_format)

# è®¾ç½®åˆ—å®½
detail_sheet.set_column('A:A', 12)
detail_sheet.set_column('B:F', 15)

# ========== åˆ›å»ºæ±‡æ€»åˆ†æè¡¨ ==========
summary_sheet = workbook.add_worksheet('æ±‡æ€»åˆ†æ')

# å†™å…¥æ ‡é¢˜
summary_sheet.merge_range('A1:D1', 'æœˆåº¦æ±‡æ€»åˆ†æ', title_format)
summary_sheet.set_row(0, 30)

# è®¡ç®—æ±‡æ€»æ•°æ®
total_sales = sum(r['sales'] for r in sales_data)
total_cost = sum(r['cost'] for r in sales_data)
total_profit = sum(r['profit'] for r in sales_data)
avg_daily_sales = total_sales / len(sales_data)
max_sales = max(r['sales'] for r in sales_data)
min_sales = min(r['sales'] for r in sales_data)

# å†™å…¥æ±‡æ€»æŒ‡æ ‡
summary_sheet.write(2, 0, 'æŒ‡æ ‡', header_format)
summary_sheet.write(2, 1, 'æ•°å€¼', header_format)
summary_sheet.write(2, 2, 'è¯´æ˜', header_format)

metrics = [
    ['æ€»é”€å”®é¢', total_sales, 'æœ¬æœˆæ‰€æœ‰é”€å”®é¢ä¹‹å’Œ'],
    ['æ€»æˆæœ¬', total_cost, 'æœ¬æœˆæ‰€æœ‰æˆæœ¬ä¹‹å’Œ'],
    ['æ€»åˆ©æ¶¦', total_profit, 'é”€å”®é¢å‡å»æˆæœ¬'],
    ['å¹³å‡æ—¥é”€å”®é¢', avg_daily_sales, 'æ€»é”€å”®é¢é™¤ä»¥å¤©æ•°'],
    ['æœ€é«˜æ—¥é”€å”®é¢', max_sales, 'å•æ—¥æœ€é«˜é”€å”®é¢'],
    ['æœ€ä½æ—¥é”€å”®é¢', min_sales, 'å•æ—¥æœ€ä½é”€å”®é¢'],
    ['åˆ©æ¶¦ç‡', total_profit / total_sales, 'åˆ©æ¶¦å é”€å”®é¢çš„æ¯”ä¾‹'],
]

for row_num, (metric, value, desc) in enumerate(metrics, start=3):
    summary_sheet.write(row_num, 0, metric, number_format)
    if 'ç‡' in metric:
        summary_sheet.write(row_num, 1, value, workbook.add_format({'num_format': '0.00%', 'border': 1}))
    else:
        summary_sheet.write(row_num, 1, value, currency_format)
    summary_sheet.write(row_num, 2, desc)

summary_sheet.set_column('A:A', 15)
summary_sheet.set_column('B:B', 20)
summary_sheet.set_column('C:C', 30)

# ========== åˆ›å»ºå›¾è¡¨ ==========
chart_sheet = workbook.add_worksheet('å›¾è¡¨åˆ†æ')

# åˆ›å»ºé”€å”®é¢è¶‹åŠ¿æŠ˜çº¿å›¾
chart1 = workbook.add_chart({'type': 'line'})
chart1.add_series({
    'name': 'æ¯æ—¥é”€å”®é¢',
    'categories': ['æ¯æ—¥æ˜ç»†', 3, 0, len(sales_data)+2, 0],
    'values': ['æ¯æ—¥æ˜ç»†', 3, 1, len(sales_data)+2, 1],
    'line': {'color': '#4472C4', 'width': 2},
})
chart1.set_title({'name': 'æ¯æ—¥é”€å”®é¢è¶‹åŠ¿'})
chart1.set_x_axis({'name': 'æ—¥æœŸ'})
chart1.set_y_axis({'name': 'é”€å”®é¢ï¼ˆå…ƒï¼‰'})
chart1.set_size({'width': 720, 'height': 400})
chart_sheet.insert_chart('B2', chart1)

# åˆ›å»ºåˆ©æ¶¦æŸ±çŠ¶å›¾
chart2 = workbook.add_chart({'type': 'column'})
chart2.add_series({
    'name': 'æ¯æ—¥åˆ©æ¶¦',
    'categories': ['æ¯æ—¥æ˜ç»†', 3, 0, len(sales_data)+2, 0],
    'values': ['æ¯æ—¥æ˜ç»†', 3, 3, len(sales_data)+2, 3],
})
chart2.set_title({'name': 'æ¯æ—¥åˆ©æ¶¦æŸ±çŠ¶å›¾'})
chart2.set_x_axis({'name': 'æ—¥æœŸ'})
chart2.set_y_axis({'name': 'åˆ©æ¶¦ï¼ˆå…ƒï¼‰'})
chart2.set_size({'width': 720, 'height': 400})
chart_sheet.insert_chart('B25', chart2)

workbook.close()
print("âœ… å®Œæ•´çš„æœˆåº¦é”€å”®åˆ†ææŠ¥è¡¨ç”ŸæˆæˆåŠŸï¼")
print("ğŸ“Š æŠ¥è¡¨åŒ…å«ï¼š")
print("   - æ¯æ—¥æ˜ç»†è¡¨ï¼ˆå¸¦æ¡ä»¶æ ¼å¼ï¼‰")
print("   - æ±‡æ€»åˆ†æè¡¨ï¼ˆè‡ªåŠ¨è®¡ç®—æŒ‡æ ‡ï¼‰")
print("   - å›¾è¡¨åˆ†æè¡¨ï¼ˆè¶‹åŠ¿å›¾å’ŒæŸ±çŠ¶å›¾ï¼‰")
````

````python
# æ•ˆç‡æå‡ç¤ºä¾‹ï¼šæ‰¹é‡ç”Ÿæˆå®¢æˆ·æŠ¥è¡¨

import xlsxwriter

# æ¨¡æ‹Ÿå®¢æˆ·æ•°æ®ï¼ˆå®é™…åº”ç”¨ä¸­ä»æ•°æ®åº“è·å–ï¼‰
customers = [
    {'name': 'å®¢æˆ·A', 'region': 'åä¸œ', 'sales': 500000, 'growth': 0.15},
    {'name': 'å®¢æˆ·B', 'region': 'åå—', 'sales': 450000, 'growth': 0.12},
    {'name': 'å®¢æˆ·C', 'region': 'ååŒ—', 'sales': 600000, 'growth': 0.18},
    {'name': 'å®¢æˆ·D', 'region': 'åä¸œ', 'sales': 380000, 'growth': 0.10},
    {'name': 'å®¢æˆ·E', 'region': 'è¥¿å—', 'sales': 420000, 'growth': 0.14},
]

# å®šä¹‰åˆ›å»ºæŠ¥è¡¨çš„å‡½æ•°ï¼ˆä»£ç å¤ç”¨ï¼‰
def create_customer_report(customer, workbook):
    """ä¸ºå•ä¸ªå®¢æˆ·åˆ›å»ºæŠ¥è¡¨"""
    worksheet = workbook.add_worksheet(customer['name'])
    
    # æ ¼å¼å®šä¹‰
    title_format = workbook.add_format({
        'bold': True,
        'font_size': 14,
        'align': 'center',
        'bg_color': '#4472C4',
        'font_color': '#FFFFFF',
    })
    
    header_format = workbook.add_format({
        'bold': True,
        'bg_color': '#B4C6E7',
        'border': 1,
    })
    
    data_format = workbook.add_format({
        'border': 1,
        'num_format': '#,##0',
    })
    
    percent_format = workbook.add_format({
        'border': 1,
        'num_format': '0.00%',
    })
    
    # å†™å…¥å®¢æˆ·ä¿¡æ¯
    worksheet.merge_range('A1:B1', f"{customer['name']} - é”€å”®æŠ¥è¡¨", title_format)
    worksheet.write(2, 0, 'åŒºåŸŸ', header_format)
    worksheet.write(2, 1, customer['region'], data_format)
    worksheet.write(3, 0, 'é”€å”®é¢', header_format)
    worksheet.write(3, 1, customer['sales'], data_format)
    worksheet.write(4, 0, 'å¢é•¿ç‡', header_format)
    worksheet.write(4, 1, customer['growth'], percent_format)
    
    worksheet.set_column('A:B', 15)

# æ‰¹é‡ç”Ÿæˆæ‰€æœ‰å®¢æˆ·æŠ¥è¡¨
workbook = xlsxwriter.Workbook('æ•ˆç‡ç¤ºä¾‹_æ‰¹é‡å®¢æˆ·æŠ¥è¡¨.xlsx')

for customer in customers:
    create_customer_report(customer, workbook)

workbook.close()

print(f"âœ… æˆåŠŸä¸º {len(customers)} ä¸ªå®¢æˆ·ç”ŸæˆæŠ¥è¡¨ï¼")
print("ğŸ’¡ å¦‚æœæ‰‹åŠ¨æ“ä½œï¼Œéœ€è¦ï¼š")
print(f"   - åˆ›å»º {len(customers)} ä¸ªå·¥ä½œè¡¨")
print(f"   - è¾“å…¥æ•°æ®å¹¶è®¾ç½®æ ¼å¼")
print(f"   - é¢„è®¡è€—æ—¶ï¼š{len(customers) * 3} åˆ†é’Ÿ")
print("ğŸ’¡ ä½¿ç”¨ä»£ç ï¼š")
print("   - ç¼–å†™å‡½æ•°ä¸€æ¬¡ï¼Œæ‰¹é‡è°ƒç”¨")
print("   - è¿è¡Œæ—¶é—´ï¼šå‡ ç§’é’Ÿ")
print(f"   - æ•ˆç‡æå‡ï¼š{len(customers) * 3 * 60 // 5} å€ä»¥ä¸Šï¼")
````

````python
# ç»ƒä¹ 1å‚è€ƒç­”æ¡ˆï¼šå‘˜å·¥å·¥èµ„è¡¨

import xlsxwriter

workbook = xlsxwriter.Workbook('ç»ƒä¹ 1_å‘˜å·¥å·¥èµ„è¡¨.xlsx')
worksheet = workbook.add_worksheet('å·¥èµ„è¡¨')

# å®šä¹‰æ ¼å¼
title_format = workbook.add_format({
    'bold': True,
    'font_size': 16,
    'align': 'center',
    'bg_color': '#4472C4',
    'font_color': '#FFFFFF',
})

header_format = workbook.add_format({
    'bold': True,
    'bg_color': '#B4C6E7',
    'border': 1,
    'align': 'center',
})

currency_format = workbook.add_format({
    'num_format': 'Â¥#,##0',
    'border': 1,
})

text_format = workbook.add_format({
    'border': 1,
})

# å‘˜å·¥æ•°æ®
employees = [
    ['å¼ ä¸‰', 'æŠ€æœ¯éƒ¨', 12000, 5000],
    ['æå››', 'é”€å”®éƒ¨', 10000, 8000],
    ['ç‹äº”', 'æŠ€æœ¯éƒ¨', 15000, 6000],
    ['èµµå…­', 'äººäº‹éƒ¨', 9000, 3000],
    ['é’±ä¸ƒ', 'æŠ€æœ¯éƒ¨', 18000, 7000],
]

# å†™å…¥æ ‡é¢˜
worksheet.merge_range('A1:E1', 'å‘˜å·¥å·¥èµ„è¡¨', title_format)
worksheet.set_row(0, 30)

# å†™å…¥è¡¨å¤´
headers = ['å§“å', 'éƒ¨é—¨', 'åŸºæœ¬å·¥èµ„', 'ç»©æ•ˆå¥–é‡‘', 'å®å‘å·¥èµ„']
worksheet.write_row(2, 0, headers, header_format)

# å†™å…¥æ•°æ®
for row_num, (name, dept, base, bonus) in enumerate(employees, start=3):
    worksheet.write(row_num, 0, name, text_format)
    worksheet.write(row_num, 1, dept, text_format)
    worksheet.write(row_num, 2, base, currency_format)
    worksheet.write(row_num, 3, bonus, currency_format)
    # è®¡ç®—å®å‘å·¥èµ„
    worksheet.write_formula(row_num, 4, f'=C{row_num+1}+D{row_num+1}', currency_format)

# æ¡ä»¶æ ¼å¼ï¼šé«˜äº®é«˜å·¥èµ„å‘˜å·¥ï¼ˆå®å‘å·¥èµ„>15000ï¼‰
worksheet.conditional_format('E4:E8', {
    'type': 'cell',
    'criteria': '>',
    'value': 15000,
    'format': workbook.add_format({'bg_color': '#FFEB9C', 'border': 1, 'num_format': 'Â¥#,##0'})
})

# æ±‡æ€»è¡Œ
total_row = len(employees) + 3
total_format = workbook.add_format({
    'bold': True,
    'bg_color': '#FFC000',
    'border': 1,
})

worksheet.write(total_row, 0, 'å¹³å‡', total_format)
worksheet.write_formula(total_row, 2, '=AVERAGE(C4:C8)', total_format)
worksheet.write_formula(total_row, 3, '=AVERAGE(D4:D8)', total_format)
worksheet.write_formula(total_row, 4, '=AVERAGE(E4:E8)', total_format)

worksheet.write(total_row + 1, 0, 'æ€»è®¡', total_format)
worksheet.write_formula(total_row + 1, 2, '=SUM(C4:C8)', total_format)
worksheet.write_formula(total_row + 1, 3, '=SUM(D4:D8)', total_format)
worksheet.write_formula(total_row + 1, 4, '=SUM(E4:E8)', total_format)

# è®¾ç½®åˆ—å®½
worksheet.set_column('A:E', 15)

workbook.close()
print("âœ… ç»ƒä¹ 1å®Œæˆï¼")
````
