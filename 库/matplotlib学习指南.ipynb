## ç¯å¢ƒé…ç½®ä¸ä¸­æ–‡æ˜¾ç¤º

åœ¨ä½¿ç”¨ matplotlib ç»˜åˆ¶åŒ…å«ä¸­æ–‡çš„å›¾è¡¨æ—¶ï¼Œéœ€è¦é…ç½®ä¸­æ–‡å­—ä½“ï¼Œå¦åˆ™ä¸­æ–‡ä¼šæ˜¾ç¤ºä¸ºæ–¹æ¡†ã€‚


```python
# å¯¼å…¥å¿…è¦çš„åº“
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

# é…ç½®ä¸­æ–‡å­—ä½“æ˜¾ç¤º
# Windows ç³»ç»Ÿå¸¸ç”¨å­—ä½“ï¼šSimHeiï¼ˆé»‘ä½“ï¼‰ã€Microsoft YaHeiï¼ˆå¾®è½¯é›…é»‘ï¼‰
# Mac ç³»ç»Ÿå¸¸ç”¨å­—ä½“ï¼šArial Unicode MSã€PingFang SC
# Linux ç³»ç»Ÿå¸¸ç”¨å­—ä½“ï¼šWenQuanYi Micro Hei

# æ–¹æ³•1ï¼šè®¾ç½®å…¨å±€å­—ä½“ï¼ˆæ¨èï¼‰
plt.rcParams['font.sans-serif'] = ['SimHei']  # ç”¨æ¥æ­£å¸¸æ˜¾ç¤ºä¸­æ–‡æ ‡ç­¾
plt.rcParams['axes.unicode_minus'] = False  # ç”¨æ¥æ­£å¸¸æ˜¾ç¤ºè´Ÿå·

# æ–¹æ³•2ï¼šä½¿ç”¨ matplotlib.font_manager åŠ¨æ€æŸ¥æ‰¾å­—ä½“
from matplotlib.font_manager import FontProperties
# font = FontProperties(fname='C:/Windows/Fonts/simhei.ttf')  # Windows è·¯å¾„ç¤ºä¾‹

print("Matplotlib ç‰ˆæœ¬:", matplotlib.__version__)
print("ä¸­æ–‡æ˜¾ç¤ºé…ç½®å®Œæˆï¼")
```

```python
# æœ€ç®€å•çš„ç»˜å›¾ç¤ºä¾‹
# åˆ›å»ºæ•°æ®
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# ç»˜åˆ¶å›¾å½¢
plt.plot(x, y)

# æ·»åŠ æ ‡é¢˜å’Œæ ‡ç­¾
plt.title('åŸºç¡€æŠ˜çº¿å›¾ç¤ºä¾‹')
plt.xlabel('X è½´æ ‡ç­¾')
plt.ylabel('Y è½´æ ‡ç­¾')

# æ˜¾ç¤ºå›¾å½¢
plt.show()
```

## æŠ˜çº¿å›¾ (Line Plot)

æŠ˜çº¿å›¾ç”¨äºæ˜¾ç¤ºæ•°æ®éšæ—¶é—´æˆ–å…¶ä»–è¿ç»­å˜é‡çš„å˜åŒ–è¶‹åŠ¿ã€‚

### å¿…é€‰å‚æ•°

- **x**: x è½´æ•°æ®ï¼ˆæ•°ç»„æˆ–åˆ—è¡¨ï¼‰
- **y**: y è½´æ•°æ®ï¼ˆæ•°ç»„æˆ–åˆ—è¡¨ï¼‰

### å¸¸ç”¨å¯é€‰å‚æ•°

- **color** æˆ– **c**: çº¿æ¡é¢œè‰²ï¼ˆå­—ç¬¦ä¸²å¦‚ 'red', 'blue' æˆ–åå…­è¿›åˆ¶å¦‚ '#FF5733'ï¼‰
- **linestyle** æˆ– **ls**: çº¿æ¡æ ·å¼ï¼ˆ'-' å®çº¿, '--' è™šçº¿, '-.' ç‚¹åˆ’çº¿, ':' ç‚¹çº¿ï¼‰
- **linewidth** æˆ– **lw**: çº¿æ¡å®½åº¦ï¼ˆæ•°å€¼ï¼Œé»˜è®¤ 1.5ï¼‰
- **marker**: æ•°æ®ç‚¹æ ‡è®°æ ·å¼ï¼ˆ'o' åœ†ç‚¹, 's' æ–¹å—, '^' ä¸‰è§’, '*' æ˜Ÿå½¢ç­‰ï¼‰
- **markersize** æˆ– **ms**: æ ‡è®°å¤§å°ï¼ˆæ•°å€¼ï¼‰
- **label**: å›¾ä¾‹æ ‡ç­¾ï¼ˆå­—ç¬¦ä¸²ï¼‰
- **alpha**: é€æ˜åº¦ï¼ˆ0-1 ä¹‹é—´çš„æ•°å€¼ï¼‰


```python
# æŠ˜çº¿å›¾ç¤ºä¾‹
# ç”Ÿæˆç¤ºä¾‹æ•°æ®
x = np.linspace(0, 10, 100)  # 0åˆ°10ä¹‹é—´ç”Ÿæˆ100ä¸ªç‚¹
y1 = np.sin(x)  # æ­£å¼¦å‡½æ•°
y2 = np.cos(x)  # ä½™å¼¦å‡½æ•°

# åˆ›å»ºå›¾å½¢ï¼Œè®¾ç½®å›¾å½¢å¤§å°
plt.figure(figsize=(10, 6))

# ç»˜åˆ¶å¤šæ¡æŠ˜çº¿
plt.plot(x, y1, 
         color='blue',           # çº¿æ¡é¢œè‰²
         linestyle='-',          # å®çº¿
         linewidth=2,            # çº¿æ¡å®½åº¦
         marker='o',             # æ ‡è®°æ ·å¼ï¼ˆæ¯éš”ä¸€å®šè·ç¦»æ˜¾ç¤ºï¼‰
         markersize=4,           # æ ‡è®°å¤§å°
         label='æ­£å¼¦å‡½æ•° sin(x)', # å›¾ä¾‹æ ‡ç­¾
         alpha=0.8)              # é€æ˜åº¦

plt.plot(x, y2, 
         color='red', 
         linestyle='--',         # è™šçº¿
         linewidth=2, 
         marker='s',             # æ–¹å—æ ‡è®°
         markersize=4, 
         label='ä½™å¼¦å‡½æ•° cos(x)', 
         alpha=0.8)

# æ·»åŠ æ ‡é¢˜å’Œæ ‡ç­¾
plt.title('ä¸‰è§’å‡½æ•°æŠ˜çº¿å›¾', fontsize=16, fontweight='bold')
plt.xlabel('X è½´ï¼ˆè§’åº¦ï¼‰', fontsize=12)
plt.ylabel('Y è½´ï¼ˆå‡½æ•°å€¼ï¼‰', fontsize=12)

# æ·»åŠ ç½‘æ ¼
plt.grid(True, linestyle='--', alpha=0.5)

# æ·»åŠ å›¾ä¾‹
plt.legend(loc='best', fontsize=10)

# æ˜¾ç¤ºå›¾å½¢
plt.tight_layout()  # è‡ªåŠ¨è°ƒæ•´å¸ƒå±€
plt.show()
```

## æŸ±çŠ¶å›¾ (Bar Chart)

æŸ±çŠ¶å›¾ç”¨äºæ¯”è¾ƒä¸åŒç±»åˆ«çš„æ•°æ®ã€‚

### å¿…é€‰å‚æ•°

- **x**: x è½´ä½ç½®ï¼ˆç±»åˆ«æ ‡ç­¾çš„ä½ç½®ï¼‰
- **height**: æŸ±å­çš„é«˜åº¦ï¼ˆå¯¹åº”æ¯ä¸ªç±»åˆ«çš„æ•°å€¼ï¼‰

### å¸¸ç”¨å¯é€‰å‚æ•°

- **width**: æŸ±å­å®½åº¦ï¼ˆé»˜è®¤ 0.8ï¼‰
- **color** æˆ– **c**: æŸ±å­é¢œè‰²ï¼ˆå¯ä»¥æ˜¯å•ä¸ªé¢œè‰²æˆ–é¢œè‰²åˆ—è¡¨ï¼‰
- **edgecolor**: æŸ±å­è¾¹æ¡†é¢œè‰²
- **linewidth**: è¾¹æ¡†å®½åº¦
- **label**: å›¾ä¾‹æ ‡ç­¾
- **alpha**: é€æ˜åº¦
- **align**: å¯¹é½æ–¹å¼ï¼ˆ'center' å±…ä¸­, 'edge' è¾¹ç¼˜å¯¹é½ï¼‰


```python
# æŸ±çŠ¶å›¾ç¤ºä¾‹
# å‡†å¤‡æ•°æ®
categories = ['äº§å“A', 'äº§å“B', 'äº§å“C', 'äº§å“D', 'äº§å“E']
sales = [23, 45, 56, 78, 32]

# åˆ›å»ºå›¾å½¢
plt.figure(figsize=(10, 6))

# ç»˜åˆ¶æŸ±çŠ¶å›¾
bars = plt.bar(categories,                    # x è½´ç±»åˆ«
                sales,                         # æŸ±å­é«˜åº¦
                width=0.6,                     # æŸ±å­å®½åº¦
                color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#98D8C8'],  # ä¸åŒé¢œè‰²
                edgecolor='black',             # è¾¹æ¡†é¢œè‰²
                linewidth=1.5,                 # è¾¹æ¡†å®½åº¦
                alpha=0.8,                     # é€æ˜åº¦
                label='é”€å”®é¢')                # å›¾ä¾‹æ ‡ç­¾

# åœ¨æŸ±å­ä¸Šæ–¹æ·»åŠ æ•°å€¼æ ‡ç­¾
for i, (bar, value) in enumerate(zip(bars, sales)):
    plt.text(bar.get_x() + bar.get_width()/2,  # x ä½ç½®ï¼ˆæŸ±å­ä¸­å¿ƒï¼‰
             bar.get_height() + 1,              # y ä½ç½®ï¼ˆæŸ±å­é¡¶éƒ¨ä¸Šæ–¹ï¼‰
             str(value),                        # æ˜¾ç¤ºçš„æ–‡æœ¬
             ha='center',                       # æ°´å¹³å¯¹é½æ–¹å¼
             va='bottom',                       # å‚ç›´å¯¹é½æ–¹å¼
             fontsize=11)

# æ·»åŠ æ ‡é¢˜å’Œæ ‡ç­¾
plt.title('å„äº§å“é”€å”®é¢å¯¹æ¯”', fontsize=16, fontweight='bold')
plt.xlabel('äº§å“ç±»åˆ«', fontsize=12)
plt.ylabel('é”€å”®é¢ï¼ˆä¸‡å…ƒï¼‰', fontsize=12)

# æ·»åŠ ç½‘æ ¼ï¼ˆä»… y è½´ï¼‰
plt.grid(True, axis='y', linestyle='--', alpha=0.3)

# æ˜¾ç¤ºå›¾å½¢
plt.tight_layout()
plt.show()
```

```python
# åˆ†ç»„æŸ±çŠ¶å›¾ç¤ºä¾‹ï¼ˆå¯¹æ¯”å¤šä¸ªæ•°æ®ç³»åˆ—ï¼‰
categories = ['ç¬¬ä¸€å­£åº¦', 'ç¬¬äºŒå­£åº¦', 'ç¬¬ä¸‰å­£åº¦', 'ç¬¬å››å­£åº¦']
sales_2023 = [120, 135, 148, 160]
sales_2024 = [130, 145, 165, 180]

# è®¾ç½®æŸ±å­ä½ç½®
x = np.arange(len(categories))
width = 0.35  # æŸ±å­å®½åº¦

# åˆ›å»ºå›¾å½¢
plt.figure(figsize=(10, 6))

# ç»˜åˆ¶ä¸¤ç»„æŸ±å­
bars1 = plt.bar(x - width/2,           # ç¬¬ä¸€ç»„æŸ±å­ä½ç½®ï¼ˆå·¦ç§»ï¼‰
                sales_2023,            # ç¬¬ä¸€ç»„æ•°æ®
                width,                 # æŸ±å­å®½åº¦
                label='2023å¹´é”€å”®é¢',  # å›¾ä¾‹æ ‡ç­¾
                color='#3498db',       # é¢œè‰²
                alpha=0.8)

bars2 = plt.bar(x + width/2,           # ç¬¬äºŒç»„æŸ±å­ä½ç½®ï¼ˆå³ç§»ï¼‰
                sales_2024,            # ç¬¬äºŒç»„æ•°æ®
                width,                 # æŸ±å­å®½åº¦
                label='2024å¹´é”€å”®é¢',  # å›¾ä¾‹æ ‡ç­¾
                color='#e74c3c',       # é¢œè‰²
                alpha=0.8)

# è®¾ç½® x è½´æ ‡ç­¾
plt.xlabel('å­£åº¦', fontsize=12)
plt.ylabel('é”€å”®é¢ï¼ˆä¸‡å…ƒï¼‰', fontsize=12)
plt.title('2023å¹´ä¸2024å¹´å­£åº¦é”€å”®é¢å¯¹æ¯”', fontsize=16, fontweight='bold')
plt.xticks(x, categories)  # è®¾ç½® x è½´åˆ»åº¦æ ‡ç­¾

# æ·»åŠ å›¾ä¾‹
plt.legend(fontsize=10)

# æ·»åŠ ç½‘æ ¼
plt.grid(True, axis='y', linestyle='--', alpha=0.3)

# æ˜¾ç¤ºå›¾å½¢
plt.tight_layout()
plt.show()
```

## æ•£ç‚¹å›¾ (Scatter Plot)

æ•£ç‚¹å›¾ç”¨äºæ˜¾ç¤ºä¸¤ä¸ªå˜é‡ä¹‹é—´çš„å…³ç³»ï¼Œå¸¸ç”¨äºç›¸å…³æ€§åˆ†æã€‚

### å¿…é€‰å‚æ•°

- **x**: x è½´æ•°æ®ï¼ˆæ•°ç»„æˆ–åˆ—è¡¨ï¼‰
- **y**: y è½´æ•°æ®ï¼ˆæ•°ç»„æˆ–åˆ—è¡¨ï¼‰

### å¸¸ç”¨å¯é€‰å‚æ•°

- **s**: æ ‡è®°å¤§å°ï¼ˆå¯ä»¥æ˜¯å•ä¸ªå€¼æˆ–æ•°ç»„ï¼‰
- **c**: æ ‡è®°é¢œè‰²ï¼ˆå¯ä»¥æ˜¯å•ä¸ªé¢œè‰²ã€é¢œè‰²åˆ—è¡¨æˆ–æ•°å€¼æ•°ç»„ç”¨äºé¢œè‰²æ˜ å°„ï¼‰
- **marker**: æ ‡è®°æ ·å¼ï¼ˆ'o' åœ†ç‚¹, 's' æ–¹å—, '^' ä¸‰è§’ç­‰ï¼‰
- **alpha**: é€æ˜åº¦
- **edgecolors**: æ ‡è®°è¾¹æ¡†é¢œè‰²
- **cmap**: é¢œè‰²æ˜ å°„ï¼ˆå½“ c æ˜¯æ•°å€¼æ•°ç»„æ—¶ä½¿ç”¨ï¼Œå¦‚ 'viridis', 'coolwarm'ï¼‰


```python
# æ•£ç‚¹å›¾ç¤ºä¾‹
# ç”Ÿæˆç¤ºä¾‹æ•°æ®ï¼ˆæ¨¡æ‹Ÿèº«é«˜å’Œä½“é‡çš„å…³ç³»ï¼‰
np.random.seed(42)  # è®¾ç½®éšæœºç§å­ï¼Œä¿è¯ç»“æœå¯å¤ç°
height = np.random.normal(170, 10, 100)  # å¹³å‡èº«é«˜170cmï¼Œæ ‡å‡†å·®10
weight = height * 0.5 + np.random.normal(0, 5, 100)  # ä½“é‡ä¸èº«é«˜ç›¸å…³

# åˆ›å»ºå›¾å½¢
plt.figure(figsize=(10, 6))

# ç»˜åˆ¶æ•£ç‚¹å›¾
plt.scatter(height,                    # x è½´æ•°æ®
            weight,                    # y è½´æ•°æ®
            s=100,                     # æ ‡è®°å¤§å°
            c='steelblue',             # æ ‡è®°é¢œè‰²
            marker='o',                # æ ‡è®°æ ·å¼
            alpha=0.6,                 # é€æ˜åº¦
            edgecolors='black',        # è¾¹æ¡†é¢œè‰²
            linewidths=1)              # è¾¹æ¡†å®½åº¦

# æ·»åŠ æ ‡é¢˜å’Œæ ‡ç­¾
plt.title('èº«é«˜ä¸ä½“é‡å…³ç³»æ•£ç‚¹å›¾', fontsize=16, fontweight='bold')
plt.xlabel('èº«é«˜ï¼ˆcmï¼‰', fontsize=12)
plt.ylabel('ä½“é‡ï¼ˆkgï¼‰', fontsize=12)

# æ·»åŠ ç½‘æ ¼
plt.grid(True, linestyle='--', alpha=0.3)

# æ˜¾ç¤ºå›¾å½¢
plt.tight_layout()
plt.show()
```

```python
# å½©è‰²æ•£ç‚¹å›¾ç¤ºä¾‹ï¼ˆä½¿ç”¨é¢œè‰²æ˜ å°„ï¼‰
# ç”Ÿæˆæ•°æ®
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)  # éšæœºé¢œè‰²å€¼ï¼ˆ0-1ä¹‹é—´ï¼‰
sizes = 1000 * np.random.rand(100)  # éšæœºå¤§å°

# åˆ›å»ºå›¾å½¢
plt.figure(figsize=(10, 6))

# ç»˜åˆ¶å½©è‰²æ•£ç‚¹å›¾
scatter = plt.scatter(x, y,                    # x, y æ•°æ®
                      s=sizes,                  # å¤§å°æ•°ç»„
                      c=colors,                 # é¢œè‰²æ•°ç»„
                      cmap='viridis',           # é¢œè‰²æ˜ å°„
                      alpha=0.6,                # é€æ˜åº¦
                      edgecolors='black',       # è¾¹æ¡†é¢œè‰²
                      linewidths=1)

# æ·»åŠ é¢œè‰²æ¡
plt.colorbar(scatter, label='é¢œè‰²å€¼')

# æ·»åŠ æ ‡é¢˜å’Œæ ‡ç­¾
plt.title('å½©è‰²æ•£ç‚¹å›¾ç¤ºä¾‹ï¼ˆå¤§å°å’Œé¢œè‰²éƒ½å˜åŒ–ï¼‰', fontsize=16, fontweight='bold')
plt.xlabel('X å€¼', fontsize=12)
plt.ylabel('Y å€¼', fontsize=12)

# æ·»åŠ ç½‘æ ¼
plt.grid(True, linestyle='--', alpha=0.3)

# æ˜¾ç¤ºå›¾å½¢
plt.tight_layout()
plt.show()
```

## é¥¼å›¾ (Pie Chart)

é¥¼å›¾ç”¨äºæ˜¾ç¤ºå„éƒ¨åˆ†å æ•´ä½“çš„æ¯”ä¾‹ã€‚

### å¿…é€‰å‚æ•°

- **x**: æ•°æ®æ•°ç»„ï¼ˆå„éƒ¨åˆ†çš„å¤§å°ï¼‰

### å¸¸ç”¨å¯é€‰å‚æ•°

- **labels**: å„éƒ¨åˆ†æ ‡ç­¾ï¼ˆåˆ—è¡¨ï¼‰
- **colors**: å„éƒ¨åˆ†é¢œè‰²ï¼ˆé¢œè‰²åˆ—è¡¨ï¼‰
- **autopct**: æ˜¾ç¤ºç™¾åˆ†æ¯”æ ¼å¼ï¼ˆå¦‚ '%.1f%%' æ˜¾ç¤ºä¸€ä½å°æ•°ï¼‰
- **startangle**: èµ·å§‹è§’åº¦ï¼ˆ0-360åº¦ï¼Œé»˜è®¤0ï¼‰
- **explode**: çªå‡ºæ˜¾ç¤ºæŸäº›éƒ¨åˆ†ï¼ˆåˆ—è¡¨ï¼Œ0è¡¨ç¤ºä¸çªå‡ºï¼Œå¤§äº0è¡¨ç¤ºçªå‡ºè·ç¦»ï¼‰
- **shadow**: æ˜¯å¦æ˜¾ç¤ºé˜´å½±ï¼ˆTrue/Falseï¼‰
- **textprops**: æ–‡æœ¬å±æ€§ï¼ˆå­—å…¸ï¼Œå¦‚ {'fontsize': 12}ï¼‰


```python
# é¥¼å›¾ç¤ºä¾‹
# å‡†å¤‡æ•°æ®
categories = ['äº§å“A', 'äº§å“B', 'äº§å“C', 'äº§å“D', 'å…¶ä»–']
sizes = [30, 25, 20, 15, 10]
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#95A5A6']
explode = (0.1, 0, 0, 0, 0)  # çªå‡ºæ˜¾ç¤ºç¬¬ä¸€éƒ¨åˆ†ï¼ˆäº§å“Aï¼‰

# åˆ›å»ºå›¾å½¢
plt.figure(figsize=(10, 8))

# ç»˜åˆ¶é¥¼å›¾
plt.pie(sizes,                              # æ•°æ®
        explode=explode,                    # çªå‡ºæ˜¾ç¤º
        labels=categories,                  # æ ‡ç­¾
        colors=colors,                      # é¢œè‰²
        autopct='%1.1f%%',                 # æ˜¾ç¤ºç™¾åˆ†æ¯”ï¼ˆä¿ç•™1ä½å°æ•°ï¼‰
        startangle=90,                      # èµ·å§‹è§’åº¦ï¼ˆä»é¡¶éƒ¨å¼€å§‹ï¼‰
        shadow=True,                        # æ˜¾ç¤ºé˜´å½±
        textprops={'fontsize': 12})        # æ–‡æœ¬å±æ€§

# æ·»åŠ æ ‡é¢˜
plt.title('äº§å“é”€å”®å æ¯”åˆ†å¸ƒ', fontsize=16, fontweight='bold')

# ç¡®ä¿é¥¼å›¾æ˜¯åœ†å½¢
plt.axis('equal')

# æ˜¾ç¤ºå›¾å½¢
plt.tight_layout()
plt.show()
```

## ç›´æ–¹å›¾ (Histogram)

ç›´æ–¹å›¾ç”¨äºæ˜¾ç¤ºæ•°æ®çš„åˆ†å¸ƒæƒ…å†µã€‚

### å¿…é€‰å‚æ•°

- **x**: æ•°æ®æ•°ç»„

### å¸¸ç”¨å¯é€‰å‚æ•°

- **bins**: æŸ±å­æ•°é‡æˆ–è¾¹ç•Œï¼ˆæ•´æ•°æˆ–æ•°ç»„ï¼Œé»˜è®¤10ï¼‰

## ç®±çº¿å›¾ (Box Plot)

ç®±çº¿å›¾ç”¨äºæ˜¾ç¤ºæ•°æ®çš„åˆ†å¸ƒã€ä¸­ä½æ•°ã€å››åˆ†ä½æ•°å’Œå¼‚å¸¸å€¼ã€‚

### å¿…é€‰å‚æ•°

- **x**: æ•°æ®ï¼ˆå¯ä»¥æ˜¯æ•°ç»„ã€åˆ—è¡¨æˆ–åˆ—è¡¨çš„åˆ—è¡¨ï¼‰

### å¸¸ç”¨å¯é€‰å‚æ•°

- **labels**: å„ç»„æ ‡ç­¾ï¼ˆåˆ—è¡¨ï¼‰
- **notch**: æ˜¯å¦æ˜¾ç¤ºç¼ºå£ï¼ˆTrue/Falseï¼Œç”¨äºæ¯”è¾ƒä¸­ä½æ•°ï¼‰
- **patch_artist**: æ˜¯å¦å¡«å……ç®±å­ï¼ˆTrue/Falseï¼‰
- **showmeans**: æ˜¯å¦æ˜¾ç¤ºå‡å€¼ï¼ˆTrue/Falseï¼‰
- **showfliers**: æ˜¯å¦æ˜¾ç¤ºå¼‚å¸¸å€¼ï¼ˆTrue/Falseï¼‰
- **vert**: æ˜¯å¦å‚ç›´æ˜¾ç¤ºï¼ˆTrue å‚ç›´, False æ°´å¹³ï¼‰


```python
# ç®±çº¿å›¾ç¤ºä¾‹
# å‡†å¤‡æ•°æ®ï¼ˆä¸‰ä¸ªç»„çš„æ•°æ®ï¼‰
np.random.seed(42)
group1 = np.random.normal(50, 10, 100)
group2 = np.random.normal(60, 12, 100)
group3 = np.random.normal(55, 8, 100)

data = [group1, group2, group3]
labels = ['ç»„åˆ«A', 'ç»„åˆ«B', 'ç»„åˆ«C']

# åˆ›å»ºå›¾å½¢
plt.figure(figsize=(10, 6))

# ç»˜åˆ¶ç®±çº¿å›¾
box = plt.boxplot(data,                      # æ•°æ®
                  labels=labels,              # æ ‡ç­¾
                  patch_artist=True,          # å¡«å……ç®±å­
                  notch=True,                 # æ˜¾ç¤ºç¼ºå£
                  showmeans=True,             # æ˜¾ç¤ºå‡å€¼
                  showfliers=True)            # æ˜¾ç¤ºå¼‚å¸¸å€¼

# è®¾ç½®ç®±å­é¢œè‰²
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

# æ·»åŠ æ ‡é¢˜å’Œæ ‡ç­¾
plt.title('ä¸‰ç»„æ•°æ®åˆ†å¸ƒç®±çº¿å›¾', fontsize=16, fontweight='bold')
plt.xlabel('ç»„åˆ«', fontsize=12)
plt.ylabel('æ•°å€¼', fontsize=12)

# æ·»åŠ ç½‘æ ¼
plt.grid(True, axis='y', linestyle='--', alpha=0.3)

# æ˜¾ç¤ºå›¾å½¢
plt.tight_layout()
plt.show()
```

## å­å›¾å¸ƒå±€ (Subplots)

ä½¿ç”¨å­å›¾å¯ä»¥åœ¨ä¸€ä¸ªå›¾å½¢ä¸­æ˜¾ç¤ºå¤šä¸ªå›¾è¡¨ã€‚

```python
# å­å›¾ç¤ºä¾‹ - æ–¹æ³•1ï¼šä½¿ç”¨ plt.subplot()
# åˆ›å»ºæ•°æ®
x = np.linspace(0, 10, 100)

# åˆ›å»º 2x2 çš„å­å›¾å¸ƒå±€
plt.figure(figsize=(12, 10))

# ç¬¬ä¸€ä¸ªå­å›¾ï¼šæŠ˜çº¿å›¾
plt.subplot(2, 2, 1)  # 2è¡Œ2åˆ—ï¼Œç¬¬1ä¸ªä½ç½®
plt.plot(x, np.sin(x), 'b-', linewidth=2)
plt.title('æ­£å¼¦å‡½æ•°', fontsize=12)
plt.grid(True, alpha=0.3)

# ç¬¬äºŒä¸ªå­å›¾ï¼šæŸ±çŠ¶å›¾
plt.subplot(2, 2, 2)  # 2è¡Œ2åˆ—ï¼Œç¬¬2ä¸ªä½ç½®
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.bar(categories, values, color='steelblue', alpha=0.7)
plt.title('æŸ±çŠ¶å›¾', fontsize=12)
plt.grid(True, axis='y', alpha=0.3)

# ç¬¬ä¸‰ä¸ªå­å›¾ï¼šæ•£ç‚¹å›¾
plt.subplot(2, 2, 3)  # 2è¡Œ2åˆ—ï¼Œç¬¬3ä¸ªä½ç½®
x_scatter = np.random.randn(100)
y_scatter = np.random.randn(100)
plt.scatter(x_scatter, y_scatter, alpha=0.6, c='green')
plt.title('æ•£ç‚¹å›¾', fontsize=12)
plt.grid(True, alpha=0.3)

# ç¬¬å››ä¸ªå­å›¾ï¼šé¥¼å›¾
plt.subplot(2, 2, 4)  # 2è¡Œ2åˆ—ï¼Œç¬¬4ä¸ªä½ç½®
sizes = [30, 25, 20, 25]
plt.pie(sizes, labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%', startangle=90)
plt.title('é¥¼å›¾', fontsize=12)

# è°ƒæ•´å¸ƒå±€
plt.tight_layout()
plt.show()
```

```python
# ç»ƒä¹ 3ï¼šè‡ªå®šä¹‰æ ·å¼å›¾è¡¨
# å‡†å¤‡æ•°æ®
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
revenue = [450, 520, 480, 600]
profit = [120, 150, 130, 180]
profit_rate = [26.7, 28.8, 27.1, 30.0]  # åˆ©æ¶¦ç‡ï¼ˆ%ï¼‰

# åˆ›å»ºå›¾å½¢å’Œåæ ‡è½´
fig, ax1 = plt.subplots(figsize=(12, 7))

# è®¾ç½®é¢œè‰²æ–¹æ¡ˆ
color1 = '#2E86AB'  # è“è‰²
color2 = '#A23B72'  # ç´«è‰²

# ç»˜åˆ¶æŸ±çŠ¶å›¾ï¼ˆæ”¶å…¥ï¼‰
bars = ax1.bar(quarters, revenue, 
               width=0.6, 
               color=color1, 
               alpha=0.8, 
               edgecolor='white',
               linewidth=2,
               label='è¥ä¸šæ”¶å…¥ï¼ˆä¸‡å…ƒï¼‰')

# åœ¨æŸ±å­ä¸Šæ–¹æ·»åŠ æ•°å€¼æ ‡ç­¾
for bar, value in zip(bars, revenue):
    ax1.text(bar.get_x() + bar.get_width()/2, 
             bar.get_height() + 10, 
             f'{value}',
             ha='center', 
             va='bottom', 
             fontsize=11,
             fontweight='bold')

# åˆ›å»ºç¬¬äºŒä¸ª y è½´ï¼ˆç”¨äºåˆ©æ¶¦ç‡ï¼‰
ax2 = ax1.twinx()

# ç»˜åˆ¶æŠ˜çº¿å›¾ï¼ˆåˆ©æ¶¦ç‡ï¼‰
line = ax2.plot(quarters, profit_rate, 
                color=color2, 
                marker='o', 
                linewidth=3, 
                markersize=10,
                label='åˆ©æ¶¦ç‡ï¼ˆ%ï¼‰',
                linestyle='-',
                markerfacecolor='white',
                markeredgewidth=2,
                markeredgecolor=color2)

# åœ¨æŠ˜çº¿ç‚¹ä¸Šæ·»åŠ æ ‡ç­¾
for i, (q, rate) in enumerate(zip(quarters, profit_rate)):
    ax2.text(i, rate + 0.5, 
             f'{rate:.1f}%', 
             ha='center', 
             va='bottom', 
             fontsize=10,
             fontweight='bold',
             color=color2)

# è®¾ç½®æ ‡é¢˜
ax1.set_title('2024å¹´å­£åº¦è¥ä¸šæ”¶å…¥ä¸åˆ©æ¶¦ç‡åˆ†æ', 
              fontsize=16, 
              fontweight='bold', 
              pad=20)

# è®¾ç½®æ ‡ç­¾
ax1.set_xlabel('å­£åº¦', fontsize=12, fontweight='bold')
ax1.set_ylabel('è¥ä¸šæ”¶å…¥ï¼ˆä¸‡å…ƒï¼‰', fontsize=12, fontweight='bold', color=color1)
ax2.set_ylabel('åˆ©æ¶¦ç‡ï¼ˆ%ï¼‰', fontsize=12, fontweight='bold', color=color2)

# è®¾ç½® y è½´é¢œè‰²
ax1.tick_params(axis='y', labelcolor=color1)
ax2.tick_params(axis='y', labelcolor=color2)

# æ·»åŠ å›¾ä¾‹
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left', fontsize=10)

# æ·»åŠ ç½‘æ ¼
ax1.grid(True, axis='y', linestyle='--', alpha=0.3, color=color1)

# æ·»åŠ æ³¨é‡Šï¼ˆæ ‡æ³¨æœ€ä½³å­£åº¦ï¼‰
ax1.annotate('æœ€ä½³å­£åº¦', 
             xy=(3, 600), 
             xytext=(3.3, 550),
             arrowprops=dict(arrowstyle='->', color='red', lw=2),
             fontsize=11,
             fontweight='bold',
             color='red',
             bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.5))

# è°ƒæ•´å¸ƒå±€
plt.tight_layout()
plt.show()
```
