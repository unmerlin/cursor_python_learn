## Matplotlib 简介

Matplotlib 是 Python 中最流行的数据可视化库之一，提供了类似 MATLAB 的绘图接口。它可以创建各种静态、动态和交互式图表。


## 环境配置与中文显示

在使用 matplotlib 绘制包含中文的图表时，需要配置中文字体，否则中文会显示为方框。


```python
# 导入必要的库
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

# 配置中文字体显示
# Windows 系统常用字体：SimHei（黑体）、Microsoft YaHei（微软雅黑）
# Mac 系统常用字体：Arial Unicode MS、PingFang SC
# Linux 系统常用字体：WenQuanYi Micro Hei

# 方法1：设置全局字体（推荐）
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

# 方法2：使用 matplotlib.font_manager 动态查找字体
from matplotlib.font_manager import FontProperties
# font = FontProperties(fname='C:/Windows/Fonts/simhei.ttf')  # Windows 路径示例

print("Matplotlib 版本:", matplotlib.__version__)
print("中文显示配置完成！")
```

## 基础绘图

### 基本绘图流程

1. 准备数据
2. 创建图形和坐标轴
3. 绘制图形
4. 添加标签、标题等
5. 显示或保存图形


```python
# 最简单的绘图示例
# 创建数据
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# 绘制图形
plt.plot(x, y)

# 添加标题和标签
plt.title('基础折线图示例')
plt.xlabel('X 轴标签')
plt.ylabel('Y 轴标签')

# 显示图形
plt.show()
```

## 折线图 (Line Plot)

折线图用于显示数据随时间或其他连续变量的变化趋势。

### 必选参数

- **x**: x 轴数据（数组或列表）
- **y**: y 轴数据（数组或列表）

### 常用可选参数

- **color** 或 **c**: 线条颜色（字符串如 'red', 'blue' 或十六进制如 '#FF5733'）
- **linestyle** 或 **ls**: 线条样式（'-' 实线, '--' 虚线, '-.' 点划线, ':' 点线）
- **linewidth** 或 **lw**: 线条宽度（数值，默认 1.5）
- **marker**: 数据点标记样式（'o' 圆点, 's' 方块, '^' 三角, '*' 星形等）
- **markersize** 或 **ms**: 标记大小（数值）
- **label**: 图例标签（字符串）
- **alpha**: 透明度（0-1 之间的数值）


```python
# 折线图示例
# 生成示例数据
x = np.linspace(0, 10, 100)  # 0到10之间生成100个点
y1 = np.sin(x)  # 正弦函数
y2 = np.cos(x)  # 余弦函数

# 创建图形，设置图形大小
plt.figure(figsize=(10, 6))

# 绘制多条折线
plt.plot(x, y1, 
         color='blue',           # 线条颜色
         linestyle='-',          # 实线
         linewidth=2,            # 线条宽度
         marker='o',             # 标记样式（每隔一定距离显示）
         markersize=4,           # 标记大小
         label='正弦函数 sin(x)', # 图例标签
         alpha=0.8)              # 透明度

plt.plot(x, y2, 
         color='red', 
         linestyle='--',         # 虚线
         linewidth=2, 
         marker='s',             # 方块标记
         markersize=4, 
         label='余弦函数 cos(x)', 
         alpha=0.8)

# 添加标题和标签
plt.title('三角函数折线图', fontsize=16, fontweight='bold')
plt.xlabel('X 轴（角度）', fontsize=12)
plt.ylabel('Y 轴（函数值）', fontsize=12)

# 添加网格
plt.grid(True, linestyle='--', alpha=0.5)

# 添加图例
plt.legend(loc='best', fontsize=10)

# 显示图形
plt.tight_layout()  # 自动调整布局
plt.show()
```

## 柱状图 (Bar Chart)

柱状图用于比较不同类别的数据。

### 必选参数

- **x**: x 轴位置（类别标签的位置）
- **height**: 柱子的高度（对应每个类别的数值）

### 常用可选参数

- **width**: 柱子宽度（默认 0.8）
- **color** 或 **c**: 柱子颜色（可以是单个颜色或颜色列表）
- **edgecolor**: 柱子边框颜色
- **linewidth**: 边框宽度
- **label**: 图例标签
- **alpha**: 透明度
- **align**: 对齐方式（'center' 居中, 'edge' 边缘对齐）


```python
# 柱状图示例
# 准备数据
categories = ['产品A', '产品B', '产品C', '产品D', '产品E']
sales = [23, 45, 56, 78, 32]

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制柱状图
bars = plt.bar(categories,                    # x 轴类别
                sales,                         # 柱子高度
                width=0.6,                     # 柱子宽度
                color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#98D8C8'],  # 不同颜色
                edgecolor='black',             # 边框颜色
                linewidth=1.5,                 # 边框宽度
                alpha=0.8,                     # 透明度
                label='销售额')                # 图例标签

# 在柱子上方添加数值标签
for i, (bar, value) in enumerate(zip(bars, sales)):
    plt.text(bar.get_x() + bar.get_width()/2,  # x 位置（柱子中心）
             bar.get_height() + 1,              # y 位置（柱子顶部上方）
             str(value),                        # 显示的文本
             ha='center',                       # 水平对齐方式
             va='bottom',                       # 垂直对齐方式
             fontsize=11)

# 添加标题和标签
plt.title('各产品销售额对比', fontsize=16, fontweight='bold')
plt.xlabel('产品类别', fontsize=12)
plt.ylabel('销售额（万元）', fontsize=12)

# 添加网格（仅 y 轴）
plt.grid(True, axis='y', linestyle='--', alpha=0.3)

# 显示图形
plt.tight_layout()
plt.show()
```

```python
# 分组柱状图示例（对比多个数据系列）
categories = ['第一季度', '第二季度', '第三季度', '第四季度']
sales_2023 = [120, 135, 148, 160]
sales_2024 = [130, 145, 165, 180]

# 设置柱子位置
x = np.arange(len(categories))
width = 0.35  # 柱子宽度

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制两组柱子
bars1 = plt.bar(x - width/2,           # 第一组柱子位置（左移）
                sales_2023,            # 第一组数据
                width,                 # 柱子宽度
                label='2023年销售额',  # 图例标签
                color='#3498db',       # 颜色
                alpha=0.8)

bars2 = plt.bar(x + width/2,           # 第二组柱子位置（右移）
                sales_2024,            # 第二组数据
                width,                 # 柱子宽度
                label='2024年销售额',  # 图例标签
                color='#e74c3c',       # 颜色
                alpha=0.8)

# 设置 x 轴标签
plt.xlabel('季度', fontsize=12)
plt.ylabel('销售额（万元）', fontsize=12)
plt.title('2023年与2024年季度销售额对比', fontsize=16, fontweight='bold')
plt.xticks(x, categories)  # 设置 x 轴刻度标签

# 添加图例
plt.legend(fontsize=10)

# 添加网格
plt.grid(True, axis='y', linestyle='--', alpha=0.3)

# 显示图形
plt.tight_layout()
plt.show()
```

## 散点图 (Scatter Plot)

散点图用于显示两个变量之间的关系，常用于相关性分析。

### 必选参数

- **x**: x 轴数据（数组或列表）
- **y**: y 轴数据（数组或列表）

### 常用可选参数

- **s**: 标记大小（可以是单个值或数组）
- **c**: 标记颜色（可以是单个颜色、颜色列表或数值数组用于颜色映射）
- **marker**: 标记样式（'o' 圆点, 's' 方块, '^' 三角等）
- **alpha**: 透明度
- **edgecolors**: 标记边框颜色
- **cmap**: 颜色映射（当 c 是数值数组时使用，如 'viridis', 'coolwarm'）


```python
# 散点图示例
# 生成示例数据（模拟身高和体重的关系）
np.random.seed(42)  # 设置随机种子，保证结果可复现
height = np.random.normal(170, 10, 100)  # 平均身高170cm，标准差10
weight = height * 0.5 + np.random.normal(0, 5, 100)  # 体重与身高相关

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制散点图
plt.scatter(height,                    # x 轴数据
            weight,                    # y 轴数据
            s=100,                     # 标记大小
            c='steelblue',             # 标记颜色
            marker='o',                # 标记样式
            alpha=0.6,                 # 透明度
            edgecolors='black',        # 边框颜色
            linewidths=1)              # 边框宽度

# 添加标题和标签
plt.title('身高与体重关系散点图', fontsize=16, fontweight='bold')
plt.xlabel('身高（cm）', fontsize=12)
plt.ylabel('体重（kg）', fontsize=12)

# 添加网格
plt.grid(True, linestyle='--', alpha=0.3)

# 显示图形
plt.tight_layout()
plt.show()
```

```python
# 彩色散点图示例（使用颜色映射）
# 生成数据
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)  # 随机颜色值（0-1之间）
sizes = 1000 * np.random.rand(100)  # 随机大小

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制彩色散点图
scatter = plt.scatter(x, y,                    # x, y 数据
                      s=sizes,                  # 大小数组
                      c=colors,                 # 颜色数组
                      cmap='viridis',           # 颜色映射
                      alpha=0.6,                # 透明度
                      edgecolors='black',       # 边框颜色
                      linewidths=1)

# 添加颜色条
plt.colorbar(scatter, label='颜色值')

# 添加标题和标签
plt.title('彩色散点图示例（大小和颜色都变化）', fontsize=16, fontweight='bold')
plt.xlabel('X 值', fontsize=12)
plt.ylabel('Y 值', fontsize=12)

# 添加网格
plt.grid(True, linestyle='--', alpha=0.3)

# 显示图形
plt.tight_layout()
plt.show()
```

## 饼图 (Pie Chart)

饼图用于显示各部分占整体的比例。

### 必选参数

- **x**: 数据数组（各部分的大小）

### 常用可选参数

- **labels**: 各部分标签（列表）
- **colors**: 各部分颜色（颜色列表）
- **autopct**: 显示百分比格式（如 '%.1f%%' 显示一位小数）
- **startangle**: 起始角度（0-360度，默认0）
- **explode**: 突出显示某些部分（列表，0表示不突出，大于0表示突出距离）
- **shadow**: 是否显示阴影（True/False）
- **textprops**: 文本属性（字典，如 {'fontsize': 12}）


```python
# 饼图示例
# 准备数据
categories = ['产品A', '产品B', '产品C', '产品D', '其他']
sizes = [30, 25, 20, 15, 10]
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#95A5A6']
explode = (0.1, 0, 0, 0, 0)  # 突出显示第一部分（产品A）

# 创建图形
plt.figure(figsize=(10, 8))

# 绘制饼图
plt.pie(sizes,                              # 数据
        explode=explode,                    # 突出显示
        labels=categories,                  # 标签
        colors=colors,                      # 颜色
        autopct='%1.1f%%',                 # 显示百分比（保留1位小数）
        startangle=90,                      # 起始角度（从顶部开始）
        shadow=True,                        # 显示阴影
        textprops={'fontsize': 12})        # 文本属性

# 添加标题
plt.title('产品销售占比分布', fontsize=16, fontweight='bold')

# 确保饼图是圆形
plt.axis('equal')

# 显示图形
plt.tight_layout()
plt.show()
```

## 直方图 (Histogram)

直方图用于显示数据的分布情况。

### 必选参数

- **x**: 数据数组

### 常用可选参数

- **bins**: 柱子数量或边界（整数或数组，默认10）
- **range**: 数据范围（元组，如 (0, 100)）
- **color**: 柱子颜色
- **edgecolor**: 边框颜色
- **alpha**: 透明度
- **density**: 是否显示密度（True 显示概率密度，False 显示频数）
- **orientation**: 方向（'vertical' 垂直, 'horizontal' 水平）


```python
# 直方图示例
# 生成示例数据（模拟考试成绩）
np.random.seed(42)
scores = np.random.normal(75, 15, 1000)  # 平均分75，标准差15

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制直方图
plt.hist(scores,                          # 数据
         bins=30,                         # 柱子数量
         color='steelblue',               # 颜色
         edgecolor='black',               # 边框颜色
         alpha=0.7,                       # 透明度
         density=False)                   # 显示频数（不是密度）

# 添加标题和标签
plt.title('学生考试成绩分布直方图', fontsize=16, fontweight='bold')
plt.xlabel('分数', fontsize=12)
plt.ylabel('频数', fontsize=12)

# 添加网格
plt.grid(True, axis='y', linestyle='--', alpha=0.3)

# 显示图形
plt.tight_layout()
plt.show()
```

```python
# 叠加直方图示例（对比两组数据）
np.random.seed(42)
group1 = np.random.normal(70, 12, 500)
group2 = np.random.normal(80, 12, 500)

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制两个直方图
plt.hist(group1, 
         bins=30, 
         color='blue', 
         alpha=0.5, 
         label='组别A', 
         edgecolor='black')

plt.hist(group2, 
         bins=30, 
         color='red', 
         alpha=0.5, 
         label='组别B', 
         edgecolor='black')

# 添加标题和标签
plt.title('两组数据分布对比', fontsize=16, fontweight='bold')
plt.xlabel('数值', fontsize=12)
plt.ylabel('频数', fontsize=12)

# 添加图例
plt.legend(fontsize=10)

# 添加网格
plt.grid(True, axis='y', linestyle='--', alpha=0.3)

# 显示图形
plt.tight_layout()
plt.show()
```

## 箱线图 (Box Plot)

箱线图用于显示数据的分布、中位数、四分位数和异常值。

### 必选参数

- **x**: 数据（可以是数组、列表或列表的列表）

### 常用可选参数

- **labels**: 各组标签（列表）
- **notch**: 是否显示缺口（True/False，用于比较中位数）
- **patch_artist**: 是否填充箱子（True/False）
- **showmeans**: 是否显示均值（True/False）
- **showfliers**: 是否显示异常值（True/False）
- **vert**: 是否垂直显示（True 垂直, False 水平）


```python
# 箱线图示例
# 准备数据（三个组的数据）
np.random.seed(42)
group1 = np.random.normal(50, 10, 100)
group2 = np.random.normal(60, 12, 100)
group3 = np.random.normal(55, 8, 100)

data = [group1, group2, group3]
labels = ['组别A', '组别B', '组别C']

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制箱线图
box = plt.boxplot(data,                      # 数据
                  labels=labels,              # 标签
                  patch_artist=True,          # 填充箱子
                  notch=True,                 # 显示缺口
                  showmeans=True,             # 显示均值
                  showfliers=True)            # 显示异常值

# 设置箱子颜色
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

# 添加标题和标签
plt.title('三组数据分布箱线图', fontsize=16, fontweight='bold')
plt.xlabel('组别', fontsize=12)
plt.ylabel('数值', fontsize=12)

# 添加网格
plt.grid(True, axis='y', linestyle='--', alpha=0.3)

# 显示图形
plt.tight_layout()
plt.show()
```

## 子图布局 (Subplots)

使用子图可以在一个图形中显示多个图表。

### 常用方法

- **plt.subplot(nrows, ncols, index)**: 创建子图
- **plt.subplots(nrows, ncols)**: 一次性创建多个子图，返回 figure 和 axes 对象
- **fig.add_subplot()**: 使用 figure 对象添加子图


```python
# 子图示例 - 方法1：使用 plt.subplot()
# 创建数据
x = np.linspace(0, 10, 100)

# 创建 2x2 的子图布局
plt.figure(figsize=(12, 10))

# 第一个子图：折线图
plt.subplot(2, 2, 1)  # 2行2列，第1个位置
plt.plot(x, np.sin(x), 'b-', linewidth=2)
plt.title('正弦函数', fontsize=12)
plt.grid(True, alpha=0.3)

# 第二个子图：柱状图
plt.subplot(2, 2, 2)  # 2行2列，第2个位置
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.bar(categories, values, color='steelblue', alpha=0.7)
plt.title('柱状图', fontsize=12)
plt.grid(True, axis='y', alpha=0.3)

# 第三个子图：散点图
plt.subplot(2, 2, 3)  # 2行2列，第3个位置
x_scatter = np.random.randn(100)
y_scatter = np.random.randn(100)
plt.scatter(x_scatter, y_scatter, alpha=0.6, c='green')
plt.title('散点图', fontsize=12)
plt.grid(True, alpha=0.3)

# 第四个子图：饼图
plt.subplot(2, 2, 4)  # 2行2列，第4个位置
sizes = [30, 25, 20, 25]
plt.pie(sizes, labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%', startangle=90)
plt.title('饼图', fontsize=12)

# 调整布局
plt.tight_layout()
plt.show()
```

```python
# 子图示例 - 方法2：使用 plt.subplots()（推荐方法）
# 创建 2x2 的子图，返回 figure 和 axes 对象
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 准备数据
x = np.linspace(0, 10, 100)

# 第一个子图：折线图
axes[0, 0].plot(x, np.sin(x), 'b-', linewidth=2)
axes[0, 0].set_title('正弦函数', fontsize=12)
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y')

# 第二个子图：柱状图
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
axes[0, 1].bar(categories, values, color='steelblue', alpha=0.7)
axes[0, 1].set_title('柱状图', fontsize=12)
axes[0, 1].grid(True, axis='y', alpha=0.3)

# 第三个子图：散点图
x_scatter = np.random.randn(100)
y_scatter = np.random.randn(100)
axes[1, 0].scatter(x_scatter, y_scatter, alpha=0.6, c='green')
axes[1, 0].set_title('散点图', fontsize=12)
axes[1, 0].grid(True, alpha=0.3)

# 第四个子图：饼图
sizes = [30, 25, 20, 25]
axes[1, 1].pie(sizes, labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title('饼图', fontsize=12)

# 添加总标题
fig.suptitle('多种图表类型展示', fontsize=16, fontweight='bold', y=0.995)

# 调整布局
plt.tight_layout()
plt.show()
```

## 实践练习

### 练习1：销售数据分析

创建一个包含以下内容的综合图表：
1. 使用柱状图显示各月份销售额
2. 使用折线图叠加显示目标销售额
3. 添加适当的标签、标题和图例


```python
# 练习1：销售数据分析
# 准备数据
months = ['1月', '2月', '3月', '4月', '5月', '6月', 
          '7月', '8月', '9月', '10月', '11月', '12月']
actual_sales = [120, 135, 148, 160, 175, 190, 205, 220, 210, 195, 180, 200]
target_sales = [130, 140, 150, 160, 170, 180, 190, 200, 210, 200, 190, 200]

# 创建图形
fig, ax = plt.subplots(figsize=(14, 7))

# 绘制柱状图（实际销售额）
bars = ax.bar(months, actual_sales, 
              width=0.6, 
              color='steelblue', 
              alpha=0.7, 
              label='实际销售额',
              edgecolor='black',
              linewidth=1)

# 绘制折线图（目标销售额）
ax.plot(months, target_sales, 
        color='red', 
        marker='o', 
        linewidth=2.5, 
        markersize=8, 
        label='目标销售额',
        linestyle='--')

# 在柱子上方添加数值标签
for i, (bar, value) in enumerate(zip(bars, actual_sales)):
    ax.text(bar.get_x() + bar.get_width()/2, 
            bar.get_height() + 5, 
            str(value), 
            ha='center', 
            va='bottom', 
            fontsize=9)

# 设置标题和标签
ax.set_title('2024年月度销售额对比分析', fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('月份', fontsize=12)
ax.set_ylabel('销售额（万元）', fontsize=12)

# 旋转 x 轴标签，避免重叠
plt.xticks(rotation=45, ha='right')

# 添加图例
ax.legend(loc='upper left', fontsize=11)

# 添加网格
ax.grid(True, axis='y', linestyle='--', alpha=0.3)

# 调整布局
plt.tight_layout()
plt.show()
```

### 练习2：学生成绩分析

创建一个包含多个子图的综合分析：
1. 子图1：各科目平均分柱状图
2. 子图2：成绩分布直方图
3. 子图3：数学与英语成绩散点图
4. 子图4：各班级成绩箱线图


```python
# 练习2：学生成绩分析
# 生成模拟数据
np.random.seed(42)

# 各科目平均分
subjects = ['语文', '数学', '英语', '物理', '化学']
avg_scores = [78, 82, 75, 80, 77]

# 成绩分布数据
all_scores = np.random.normal(78, 12, 500)

# 数学和英语成绩（模拟相关性）
math_scores = np.random.normal(82, 10, 100)
english_scores = math_scores * 0.7 + np.random.normal(20, 8, 100)

# 三个班级的成绩
class1 = np.random.normal(75, 10, 50)
class2 = np.random.normal(80, 10, 50)
class3 = np.random.normal(78, 10, 50)

# 创建 2x2 子图
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 子图1：各科目平均分柱状图
axes[0, 0].bar(subjects, avg_scores, color='steelblue', alpha=0.7, edgecolor='black')
axes[0, 0].set_title('各科目平均分', fontsize=12, fontweight='bold')
axes[0, 0].set_ylabel('分数', fontsize=10)
axes[0, 0].grid(True, axis='y', alpha=0.3)
# 添加数值标签
for i, (subject, score) in enumerate(zip(subjects, avg_scores)):
    axes[0, 0].text(i, score + 1, f'{score:.1f}', ha='center', va='bottom', fontsize=9)

# 子图2：成绩分布直方图
axes[0, 1].hist(all_scores, bins=30, color='green', alpha=0.7, edgecolor='black')
axes[0, 1].set_title('成绩分布直方图', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('分数', fontsize=10)
axes[0, 1].set_ylabel('频数', fontsize=10)
axes[0, 1].grid(True, axis='y', alpha=0.3)

# 子图3：数学与英语成绩散点图
axes[1, 0].scatter(math_scores, english_scores, alpha=0.6, c='purple', s=50, edgecolors='black', linewidths=0.5)
axes[1, 0].set_title('数学与英语成绩关系', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('数学成绩', fontsize=10)
axes[1, 0].set_ylabel('英语成绩', fontsize=10)
axes[1, 0].grid(True, alpha=0.3)

# 子图4：各班级成绩箱线图
box_data = [class1, class2, class3]
box_labels = ['班级A', '班级B', '班级C']
bp = axes[1, 1].boxplot(box_data, labels=box_labels, patch_artist=True, notch=True, showmeans=True)
# 设置箱子颜色
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
axes[1, 1].set_title('各班级成绩分布', fontsize=12, fontweight='bold')
axes[1, 1].set_ylabel('分数', fontsize=10)
axes[1, 1].grid(True, axis='y', alpha=0.3)

# 添加总标题
fig.suptitle('学生成绩综合分析', fontsize=16, fontweight='bold', y=0.995)

# 调整布局
plt.tight_layout()
plt.show()
```

### 练习3：自定义样式图表

创建一个美观的图表，包含：
1. 自定义颜色方案
2. 添加数据标签
3. 使用不同的图表元素（网格、图例、注释等）


```python
# 练习3：自定义样式图表
# 准备数据
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
revenue = [450, 520, 480, 600]
profit = [120, 150, 130, 180]
profit_rate = [26.7, 28.8, 27.1, 30.0]  # 利润率（%）

# 创建图形和坐标轴
fig, ax1 = plt.subplots(figsize=(12, 7))

# 设置颜色方案
color1 = '#2E86AB'  # 蓝色
color2 = '#A23B72'  # 紫色

# 绘制柱状图（收入）
bars = ax1.bar(quarters, revenue, 
               width=0.6, 
               color=color1, 
               alpha=0.8, 
               edgecolor='white',
               linewidth=2,
               label='营业收入（万元）')

# 在柱子上方添加数值标签
for bar, value in zip(bars, revenue):
    ax1.text(bar.get_x() + bar.get_width()/2, 
             bar.get_height() + 10, 
             f'{value}',
             ha='center', 
             va='bottom', 
             fontsize=11,
             fontweight='bold')

# 创建第二个 y 轴（用于利润率）
ax2 = ax1.twinx()

# 绘制折线图（利润率）
line = ax2.plot(quarters, profit_rate, 
                color=color2, 
                marker='o', 
                linewidth=3, 
                markersize=10,
                label='利润率（%）',
                linestyle='-',
                markerfacecolor='white',
                markeredgewidth=2,
                markeredgecolor=color2)

# 在折线点上添加标签
for i, (q, rate) in enumerate(zip(quarters, profit_rate)):
    ax2.text(i, rate + 0.5, 
             f'{rate:.1f}%', 
             ha='center', 
             va='bottom', 
             fontsize=10,
             fontweight='bold',
             color=color2)

# 设置标题
ax1.set_title('2024年季度营业收入与利润率分析', 
              fontsize=16, 
              fontweight='bold', 
              pad=20)

# 设置标签
ax1.set_xlabel('季度', fontsize=12, fontweight='bold')
ax1.set_ylabel('营业收入（万元）', fontsize=12, fontweight='bold', color=color1)
ax2.set_ylabel('利润率（%）', fontsize=12, fontweight='bold', color=color2)

# 设置 y 轴颜色
ax1.tick_params(axis='y', labelcolor=color1)
ax2.tick_params(axis='y', labelcolor=color2)

# 添加图例
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left', fontsize=10)

# 添加网格
ax1.grid(True, axis='y', linestyle='--', alpha=0.3, color=color1)

# 添加注释（标注最佳季度）
ax1.annotate('最佳季度', 
             xy=(3, 600), 
             xytext=(3.3, 550),
             arrowprops=dict(arrowstyle='->', color='red', lw=2),
             fontsize=11,
             fontweight='bold',
             color='red',
             bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.5))

# 调整布局
plt.tight_layout()
plt.show()
```
