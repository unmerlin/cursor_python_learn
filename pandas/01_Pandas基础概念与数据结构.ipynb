```python
data = [10, 20, 30, 40, 50]
index = ['a', 'b', 'c', 'd', 'e']
series2 = pd.Series(data, index=index)

print("å¸¦è‡ªå®šä¹‰ç´¢å¼•çš„Seriesï¼š")
print(series2)
print("\n" + "="*50)

print("\né€šè¿‡ç´¢å¼•è®¿é—®ï¼š")
print(f"series2['a'] = {series2['a']}")
print(f"series2['c'] = {series2['c']}")

print(f"\né€šè¿‡ä½ç½®è®¿é—®ï¼š")
print(f"series2.iloc[0] = {series2.iloc[0]}")
print(f"series2.iloc[2] = {series2.iloc[2]}")
```

### 3. åˆ›å»ºSeries - ä»å­—å…¸åˆ›å»º

```python
data_dict = {
    'åŒ—äº¬': 2154,
    'ä¸Šæµ·': 2487,
    'å¹¿å·': 1868,
    'æ·±åœ³': 1756
}
series3 = pd.Series(data_dict)

print("ä»å­—å…¸åˆ›å»ºçš„Seriesï¼š")
print(series3)
print("\n" + "="*50)

print("\né€šè¿‡åŸå¸‚åè®¿é—®äººå£ï¼š")
print(f"åŒ—äº¬äººå£: {series3['åŒ—äº¬']} ä¸‡äºº")
print(f"ä¸Šæµ·äººå£: {series3['ä¸Šæµ·']} ä¸‡äºº")
```

### 4. Seriesçš„åŸºæœ¬æ“ä½œ

```python
scores = pd.Series([85, 92, 78, 96, 88], index=['å¼ ä¸‰', 'æå››', 'ç‹äº”', 'èµµå…­', 'é’±ä¸ƒ'])

print("åŸå§‹æ•°æ®ï¼š")
print(scores)
print("\n" + "="*50)

print("\nåŸºæœ¬ç»Ÿè®¡ä¿¡æ¯ï¼š")
print(f"å¹³å‡å€¼: {scores.mean():.2f}")
print(f"æœ€å¤§å€¼: {scores.max()}")
print(f"æœ€å°å€¼: {scores.min()}")
print(f"æ€»å’Œ: {scores.sum()}")
print(f"æ ‡å‡†å·®: {scores.std():.2f}")

print("\nå‘é‡åŒ–æ“ä½œï¼ˆæ‰€æœ‰åˆ†æ•°åŠ 10ï¼‰ï¼š")
scores_plus_10 = scores + 10
print(scores_plus_10)

print("\nç­›é€‰å¤§äº90åˆ†çš„æˆç»©ï¼š")
high_scores = scores[scores > 90]
print(high_scores)
```
