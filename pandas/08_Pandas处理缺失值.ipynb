## â­ ä½¿ç”¨dropna()åˆ é™¤ç¼ºå¤±å€¼

## â­ ä½¿ç”¨fillna()å¡«å……ç¼ºå¤±å€¼

### 1. ä½¿ç”¨å›ºå®šå€¼å¡«å……

```python
# ä½¿ç”¨å›ºå®šå€¼å¡«å……æ‰€æœ‰ç¼ºå¤±å€¼
df_fill = df.copy()
df_fill = df_fill.fillna(0)
print("ç”¨0å¡«å……æ‰€æœ‰ç¼ºå¤±å€¼ï¼š")
print(df_fill)

# æ³¨æ„ï¼šå­—ç¬¦ä¸²åˆ—ä¹Ÿä¼šè¢«å¡«å……ä¸º0ï¼ˆå¯èƒ½ä¸åˆé€‚ï¼‰
```

### 2. ä½¿ç”¨å­—å…¸æŒ‡å®šä¸åŒåˆ—çš„å¡«å……å€¼

```python
# ä¸ºä¸åŒåˆ—æŒ‡å®šä¸åŒçš„å¡«å……å€¼
df_fill = df.copy()
df_fill = df_fill.fillna({
    'å¹´é¾„': df_fill['å¹´é¾„'].mean(),  # ç”¨å¹³å‡å€¼å¡«å……
    'è–ªèµ„': df_fill['è–ªèµ„'].median(),  # ç”¨ä¸­ä½æ•°å¡«å……
    'åŸå¸‚': 'æœªçŸ¥',  # ç”¨å­—ç¬¦ä¸²å¡«å……
    'éƒ¨é—¨': 'æœªåˆ†é…'  # ç”¨å­—ç¬¦ä¸²å¡«å……
})
print("ä½¿ç”¨å­—å…¸æŒ‡å®šå¡«å……å€¼ï¼š")
print(df_fill)
```

### 3. å‰å‘å¡«å……ï¼ˆforward fillï¼‰

```python
# method='ffill'æˆ–'pad': ç”¨å‰ä¸€ä¸ªå€¼å¡«å……
df_fill = df.copy()
df_fill['å¹´é¾„'] = df_fill['å¹´é¾„'].fillna(method='ffill')
print("å‰å‘å¡«å……'å¹´é¾„'åˆ—ï¼š")
print(df_fill[['å§“å', 'å¹´é¾„']])

# å®é™…åº”ç”¨ï¼šæ—¶é—´åºåˆ—æ•°æ®ï¼Œç”¨å‰ä¸€ä¸ªæ—¶é—´ç‚¹çš„å€¼å¡«å……
```

### 4. åå‘å¡«å……ï¼ˆbackward fillï¼‰

```python
# method='bfill'æˆ–'backfill': ç”¨åä¸€ä¸ªå€¼å¡«å……
df_fill = df.copy()
df_fill['å¹´é¾„'] = df_fill['å¹´é¾„'].fillna(method='bfill')
print("åå‘å¡«å……'å¹´é¾„'åˆ—ï¼š")
print(df_fill[['å§“å', 'å¹´é¾„']])

# å®é™…åº”ç”¨ï¼šç”¨ä¸‹ä¸€ä¸ªå€¼å¡«å……ç¼ºå¤±å€¼
```

### 5. ä½¿ç”¨ç»Ÿè®¡å€¼å¡«å……

```python
# ä½¿ç”¨å¹³å‡å€¼å¡«å……
df_fill = df.copy()
df_fill['å¹´é¾„'] = df_fill['å¹´é¾„'].fillna(df_fill['å¹´é¾„'].mean())
print("ç”¨å¹³å‡å€¼å¡«å……'å¹´é¾„'åˆ—ï¼š")
print(df_fill[['å§“å', 'å¹´é¾„']])

# å…¶ä»–ç»Ÿè®¡å€¼ï¼š
# - mean(): å¹³å‡å€¼
# - median(): ä¸­ä½æ•°
# - mode(): ä¼—æ•°ï¼ˆå­—ç¬¦ä¸²åˆ—ï¼‰
```

## â­ æ ¹æ®åœºæ™¯é€‰æ‹©åˆé€‚çš„å¤„ç†æ–¹æ³•

### 1. ç¼ºå¤±å€¼è¾ƒå°‘çš„å¤„ç†

```python
# ç¼ºå¤±å€¼æ¯”ä¾‹å°äº5%ï¼Œå¯ä»¥åˆ é™¤
missing_ratio = df.isnull().sum() / len(df)
print("ç¼ºå¤±å€¼æ¯”ä¾‹ï¼š")
print(missing_ratio.round(3))

# åˆ é™¤ç¼ºå¤±å€¼æ¯”ä¾‹å°äº0.2çš„è¡Œ
df_clean = df.dropna(thresh=int(len(df.columns) * 0.8))
print("\nåˆ é™¤ç¼ºå¤±å€¼è¿‡å¤šçš„è¡Œï¼š")
print(df_clean)
```

### 2. æ•°å€¼å‹ç¼ºå¤±å€¼å¤„ç†

```python
# æ•°å€¼å‹åˆ—ï¼šä½¿ç”¨ç»Ÿè®¡å€¼å¡«å……
df_numeric = df.copy()
numeric_cols = ['å¹´é¾„', 'è–ªèµ„']

for col in numeric_cols:
    if df_numeric[col].isnull().any():
        df_numeric[col] = df_numeric[col].fillna(df_numeric[col].median())

print("æ•°å€¼å‹åˆ—ç”¨ä¸­ä½æ•°å¡«å……ï¼š")
print(df_numeric[numeric_cols])
```

### 3. åˆ†ç±»å‹ç¼ºå¤±å€¼å¤„ç†

```python
# åˆ†ç±»å‹åˆ—ï¼šä½¿ç”¨ä¼—æ•°æˆ–'æœªçŸ¥'å¡«å……
df_categorical = df.copy()
categorical_cols = ['åŸå¸‚', 'éƒ¨é—¨']

for col in categorical_cols:
    if df_categorical[col].isnull().any():
        # ä½¿ç”¨ä¼—æ•°å¡«å……ï¼Œå¦‚æœæ²¡æœ‰ä¼—æ•°åˆ™ç”¨'æœªçŸ¥'
        mode_value = df_categorical[col].mode()
        fill_value = mode_value[0] if len(mode_value) > 0 else 'æœªçŸ¥'
        df_categorical[col] = df_categorical[col].fillna(fill_value)

print("åˆ†ç±»å‹åˆ—ç”¨ä¼—æ•°å¡«å……ï¼š")
print(df_categorical[categorical_cols])
```

### 4. ç»¼åˆå¤„ç†å‡½æ•°

```python
def handle_missing_values(df, numeric_strategy='median', categorical_strategy='mode'):
    """ç»¼åˆå¤„ç†ç¼ºå¤±å€¼çš„å‡½æ•°"""
    df_clean = df.copy()
    
    # å¤„ç†æ•°å€¼å‹åˆ—
    numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        if df_clean[col].isnull().any():
            if numeric_strategy == 'mean':
                df_clean[col] = df_clean[col].fillna(df_clean[col].mean())
            elif numeric_strategy == 'median':
                df_clean[col] = df_clean[col].fillna(df_clean[col].median())
    
    # å¤„ç†åˆ†ç±»å‹åˆ—
    categorical_cols = df_clean.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        if df_clean[col].isnull().any():
            if categorical_strategy == 'mode':
                mode_value = df_clean[col].mode()
                fill_value = mode_value[0] if len(mode_value) > 0 else 'æœªçŸ¥'
                df_clean[col] = df_clean[col].fillna(fill_value)
            else:
                df_clean[col] = df_clean[col].fillna('æœªçŸ¥')
    
    return df_clean

# ä½¿ç”¨ç¤ºä¾‹
df_processed = handle_missing_values(df)
print("ç»¼åˆå¤„ç†åçš„æ•°æ®ï¼š")
print(df_processed)
print("\nç¼ºå¤±å€¼æ£€æŸ¥ï¼š")
print(df_processed.isnull().sum())
```
