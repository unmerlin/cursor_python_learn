## â­ å¤„ç†è½¬æ¢é”™è¯¯å’Œå¼‚å¸¸

### 1. å¤„ç†æ— æ³•è½¬æ¢çš„å€¼

```python
# ä½¿ç”¨try-exceptå¤„ç†è½¬æ¢é”™è¯¯
def safe_convert(series, target_type, errors='coerce'):
    """å®‰å…¨è½¬æ¢æ•°æ®ç±»å‹"""
    try:
        if target_type == 'numeric':
            return pd.to_numeric(series, errors=errors)
        elif target_type == 'datetime':
            return pd.to_datetime(series, errors=errors)
        else:
            return series.astype(target_type)
    except Exception as e:
        print(f"è½¬æ¢å¤±è´¥: {e}")
        return series

# ä½¿ç”¨ç¤ºä¾‹
df_safe = df.copy()
df_safe['ID_safe'] = safe_convert(df_safe['ID'], 'numeric')
print("å®‰å…¨è½¬æ¢ï¼š")
print(df_safe[['ID', 'ID_safe']])
```

### 2. æ£€æŸ¥è½¬æ¢ç»“æœ

```python
# è½¬æ¢åæ£€æŸ¥æ•°æ®ç±»å‹
df_convert = df.copy()
df_convert['ID'] = pd.to_numeric(df_convert['ID'])

print(f"è½¬æ¢åçš„æ•°æ®ç±»å‹: {df_convert['ID'].dtype}")
print(f"æ˜¯å¦æœ‰ç¼ºå¤±å€¼: {df_convert['ID'].isnull().any()}")
print(f"æ•°æ®èŒƒå›´: {df_convert['ID'].min()} - {df_convert['ID'].max()}")

# å®é™…åº”ç”¨ï¼šè½¬æ¢åéªŒè¯ç»“æœ
```

### 3. æ‰¹é‡ç±»å‹è½¬æ¢

```python
def convert_dataframe_types(df, type_mapping):
    """æ‰¹é‡è½¬æ¢DataFrameçš„æ•°æ®ç±»å‹"""
    df_convert = df.copy()
    
    for col, target_type in type_mapping.items():
        if col in df_convert.columns:
            try:
                if target_type == 'numeric':
                    df_convert[col] = pd.to_numeric(df_convert[col], errors='coerce')
                elif target_type == 'datetime':
                    df_convert[col] = pd.to_datetime(df_convert[col], errors='coerce')
                else:
                    df_convert[col] = df_convert[col].astype(target_type)
            except Exception as e:
                print(f"åˆ— '{col}' è½¬æ¢å¤±è´¥: {e}")
    
    return df_convert

# ä½¿ç”¨ç¤ºä¾‹
type_mapping = {
    'ID': 'numeric',
    'æ—¥æœŸ': 'datetime',
    'éƒ¨é—¨': 'category'
}
df_batch = convert_dataframe_types(df, type_mapping)
print("æ‰¹é‡è½¬æ¢åçš„æ•°æ®ç±»å‹ï¼š")
print(df_batch.dtypes)
```
