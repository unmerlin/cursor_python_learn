## ğŸ“¦ å¯¼å…¥åº“å¹¶åˆ›å»ºç¤ºä¾‹æ•°æ®

```python
import pandas as pd
import numpy as np

# åˆ›å»ºç¤ºä¾‹æ•°æ®
data = {
    'å§“å': ['å¼ ä¸‰', 'æå››', 'ç‹äº”', 'èµµå…­', 'é’±ä¸ƒ', 'å­™å…«', 'å‘¨ä¹', 'å´å'],
    'å¹´é¾„': [25, 30, 28, 35, 32, 27, 29, 31],
    'è–ªèµ„': [8000, 12000, 9000, 15000, 11000, 9500, 10000, 8500],
    'åŸå¸‚': ['åŒ—äº¬', 'ä¸Šæµ·', 'å¹¿å·', 'æ·±åœ³', 'åŒ—äº¬', 'ä¸Šæµ·', 'å¹¿å·', 'æ·±åœ³'],
    'éƒ¨é—¨': ['é”€å”®éƒ¨', 'æŠ€æœ¯éƒ¨', 'é”€å”®éƒ¨', 'æŠ€æœ¯éƒ¨', 'å¸‚åœºéƒ¨', 'æŠ€æœ¯éƒ¨', 'é”€å”®éƒ¨', 'å¸‚åœºéƒ¨'],
    'é‚®ç®±': ['zhang@example.com', 'li@test.com', 'wang@example.com', 
             'zhao@test.com', 'qian@example.com', 'sun@test.com', 
             'zhou@example.com', 'wu@test.com']
}
df = pd.DataFrame(data)

print("ç¤ºä¾‹æ•°æ®ï¼š")
print(df)
```

# ç¬¬ä¸€éƒ¨åˆ†ï¼šåŸºæœ¬æ¡ä»¶ç­›é€‰

## â­ ä½¿ç”¨æ¯”è¾ƒè¿ç®—ç¬¦ç­›é€‰

### 1. å¤§äºã€å°äºç­›é€‰

```python
# ç­›é€‰å¹´é¾„å¤§äº30çš„è¡Œ
result = df[df['å¹´é¾„'] > 30]
print("å¹´é¾„å¤§äº30ï¼š")
print(result)
```

### 2. å¤§äºç­‰äºã€å°äºç­‰äºç­›é€‰

```python
# ç­›é€‰è–ªèµ„å¤§äºç­‰äº10000çš„è¡Œ
result = df[df['è–ªèµ„'] >= 10000]
print("è–ªèµ„>=10000ï¼š")
print(result)
```

### 3. ç­‰äºç­›é€‰

```python
# ç­›é€‰åŸå¸‚ä¸º'åŒ—äº¬'çš„è¡Œ
result = df[df['åŸå¸‚'] == 'åŒ—äº¬']
print("åŸå¸‚æ˜¯åŒ—äº¬ï¼š")
print(result)
```

### 4. ä¸ç­‰äºç­›é€‰

```python
# ç­›é€‰åŸå¸‚ä¸æ˜¯'åŒ—äº¬'çš„è¡Œ
result = df[df['åŸå¸‚'] != 'åŒ—äº¬']
print("åŸå¸‚ä¸æ˜¯åŒ—äº¬ï¼š")
print(result)
```

# ç¬¬äºŒéƒ¨åˆ†ï¼šå¤šæ¡ä»¶ç­›é€‰

## â­ ç»„åˆå¤šä¸ªæ¡ä»¶

### 1. ä½¿ç”¨&ï¼ˆä¸ï¼‰ç»„åˆæ¡ä»¶

```python
# ç­›é€‰å¹´é¾„>28 ä¸” è–ªèµ„>10000ï¼ˆå¿…é¡»ç”¨æ‹¬å·ï¼‰
result = df[(df['å¹´é¾„'] > 28) & (df['è–ªèµ„'] > 10000)]
print("å¹´é¾„>28 ä¸” è–ªèµ„>10000ï¼š")
print(result)

# æ³¨æ„ï¼šæ¯ä¸ªæ¡ä»¶å¿…é¡»ç”¨æ‹¬å·æ‹¬èµ·æ¥
```

### 2. ä½¿ç”¨|ï¼ˆæˆ–ï¼‰ç»„åˆæ¡ä»¶

```python
# ç­›é€‰åŸå¸‚æ˜¯'åŒ—äº¬'æˆ–'ä¸Šæµ·'
result = df[(df['åŸå¸‚'] == 'åŒ—äº¬') | (df['åŸå¸‚'] == 'ä¸Šæµ·')]
print("åŸå¸‚æ˜¯åŒ—äº¬æˆ–ä¸Šæµ·ï¼š")
print(result)
```

### 3. ä½¿ç”¨~ï¼ˆéï¼‰æ¡ä»¶

```python
# ç­›é€‰å¹´é¾„ä¸å¤§äº30çš„è¡Œ
result = df[~(df['å¹´é¾„'] > 30)]
print("å¹´é¾„ä¸å¤§äº30ï¼š")
print(result)

# ç­‰ä»·äº df[df['å¹´é¾„'] <= 30]
```

### 4. å¤æ‚æ¡ä»¶ç»„åˆ

```python
# ç­›é€‰ï¼ˆå¹´é¾„>28ä¸”è–ªèµ„>10000ï¼‰æˆ–ï¼ˆåŸå¸‚æ˜¯åŒ—äº¬ï¼‰
result = df[((df['å¹´é¾„'] > 28) & (df['è–ªèµ„'] > 10000)) | (df['åŸå¸‚'] == 'åŒ—äº¬')]
print("å¤æ‚æ¡ä»¶ç»„åˆï¼š")
print(result)
```

# ç¬¬ä¸‰éƒ¨åˆ†ï¼šå­—ç¬¦ä¸²ç­›é€‰

## â­ æ–‡æœ¬æ•°æ®çš„ç­›é€‰æ–¹æ³•

### 1. contains() - åŒ…å«ç‰¹å®šå­—ç¬¦ä¸²

```python
# ç­›é€‰éƒ¨é—¨åç§°åŒ…å«'é”€å”®'çš„è¡Œ
result = df[df['éƒ¨é—¨'].str.contains('é”€å”®')]
print("éƒ¨é—¨åŒ…å«'é”€å”®'ï¼š")
print(result)

# æ³¨æ„ï¼šå¿…é¡»ä½¿ç”¨.strè®¿é—®å™¨
```

### 2. startswith() - ä»¥ç‰¹å®šå­—ç¬¦ä¸²å¼€å¤´

```python
# ç­›é€‰é‚®ç®±ä»¥'zhang'å¼€å¤´çš„è¡Œ
result = df[df['é‚®ç®±'].str.startswith('zhang')]
print("é‚®ç®±ä»¥'zhang'å¼€å¤´ï¼š")
print(result)
```

### 3. endswith() - ä»¥ç‰¹å®šå­—ç¬¦ä¸²ç»“å°¾

```python
# ç­›é€‰é‚®ç®±ä»¥'example.com'ç»“å°¾çš„è¡Œ
result = df[df['é‚®ç®±'].str.endswith('example.com')]
print("é‚®ç®±ä»¥'example.com'ç»“å°¾ï¼š")
print(result)
```

### 4. å­—ç¬¦ä¸²é•¿åº¦ç­›é€‰

```python
# ç­›é€‰å§“åé•¿åº¦å¤§äº2çš„è¡Œ
result = df[df['å§“å'].str.len() > 2]
print("å§“åé•¿åº¦>2ï¼š")
print(result)
```

### 5. æ­£åˆ™è¡¨è¾¾å¼ç­›é€‰

```python
# ä½¿ç”¨æ­£åˆ™è¡¨è¾¾å¼ç­›é€‰ï¼ˆéœ€è¦å¯¼å…¥reï¼‰
import re

# ç­›é€‰é‚®ç®±åŒ…å«æ•°å­—çš„è¡Œ
result = df[df['é‚®ç®±'].str.contains(r'\d', regex=True)]
print("é‚®ç®±åŒ…å«æ•°å­—ï¼š")
print(result)

# regex=True è¡¨ç¤ºä½¿ç”¨æ­£åˆ™è¡¨è¾¾å¼
```

## â­ æ•°å€¼æ•°æ®çš„èŒƒå›´ç­›é€‰

### 1. between() - æ•°å€¼èŒƒå›´ç­›é€‰

```python
# ç­›é€‰å¹´é¾„åœ¨28åˆ°32ä¹‹é—´çš„è¡Œï¼ˆåŒ…å«è¾¹ç•Œï¼‰
result = df[df['å¹´é¾„'].between(28, 32)]
print("å¹´é¾„åœ¨28-32ä¹‹é—´ï¼š")
print(result)
```

### 2. between() - ä¸åŒ…å«è¾¹ç•Œ

```python
# ç­›é€‰å¹´é¾„åœ¨28åˆ°32ä¹‹é—´ï¼ˆä¸åŒ…å«è¾¹ç•Œï¼‰
result = df[df['å¹´é¾„'].between(28, 32, inclusive='neither')]
print("å¹´é¾„åœ¨28-32ä¹‹é—´ï¼ˆä¸åŒ…å«è¾¹ç•Œï¼‰ï¼š")
print(result)

# inclusiveå‚æ•°ï¼š
# - 'both': åŒ…å«ä¸¤ç«¯ï¼ˆé»˜è®¤ï¼‰
# - 'left': åªåŒ…å«å·¦ç«¯
# - 'right': åªåŒ…å«å³ç«¯
# - 'neither': éƒ½ä¸åŒ…å«
```

### 3. å¤šåˆ—æ•°å€¼èŒƒå›´ç­›é€‰

```python
# ç­›é€‰å¹´é¾„å’Œè–ªèµ„éƒ½åœ¨æŒ‡å®šèŒƒå›´å†…çš„è¡Œ
result = df[(df['å¹´é¾„'].between(28, 32)) & (df['è–ªèµ„'].between(9000, 12000))]
print("å¹´é¾„28-32 ä¸” è–ªèµ„9000-12000ï¼š")
print(result)
```

# ç¬¬äº”éƒ¨åˆ†ï¼šä½¿ç”¨isin()ç­›é€‰

## â­ å¤šå€¼ç­›é€‰çš„ä¾¿æ·æ–¹æ³•

### 2. ~isin() - å€¼ä¸åœ¨åˆ—è¡¨ä¸­

### 3. å¤šåˆ—isin()ç­›é€‰

```python
# ç­›é€‰åŸå¸‚å’Œéƒ¨é—¨éƒ½åœ¨æŒ‡å®šåˆ—è¡¨ä¸­çš„è¡Œ
result = df[df['åŸå¸‚'].isin(['åŒ—äº¬', 'ä¸Šæµ·']) & df['éƒ¨é—¨'].isin(['é”€å”®éƒ¨', 'æŠ€æœ¯éƒ¨'])]
print("åŸå¸‚åœ¨æŒ‡å®šåˆ—è¡¨ ä¸” éƒ¨é—¨åœ¨æŒ‡å®šåˆ—è¡¨ï¼š")
print(result)
```
