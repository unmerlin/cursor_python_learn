```python
# æŒ‡å®šå·¥ä½œè¡¨åç§°
df.to_excel('data_sheet.xlsx', index=False, sheet_name='å‘˜å·¥ä¿¡æ¯')
print("å·²å¯¼å‡ºï¼Œå·¥ä½œè¡¨åä¸º'å‘˜å·¥ä¿¡æ¯'")

# sheet_nameå‚æ•°ï¼šæŒ‡å®šå·¥ä½œè¡¨åç§°ï¼ˆé»˜è®¤'Sheet1'ï¼‰
```

```python
# å¯¼å‡ºå¤šä¸ªå·¥ä½œè¡¨åˆ°åŒä¸€ä¸ªExcelæ–‡ä»¶
with pd.ExcelWriter('data_multi.xlsx') as writer:
    df.to_excel(writer, sheet_name='å‘˜å·¥ä¿¡æ¯', index=False)
    df.groupby('éƒ¨é—¨')['å·¥èµ„'].mean().to_excel(writer, sheet_name='éƒ¨é—¨ç»Ÿè®¡', index=True)

print("å·²å¯¼å‡ºå¤šä¸ªå·¥ä½œè¡¨")

# ä½¿ç”¨ExcelWriterå¯ä»¥å¯¼å‡ºå¤šä¸ªå·¥ä½œè¡¨
```

### 5. è®¾ç½®åˆ—å®½ï¼ˆä½¿ç”¨xlsxwriterï¼‰

```python
# ä½¿ç”¨xlsxwriterè®¾ç½®æ ¼å¼
with pd.ExcelWriter('data_formatted.xlsx', engine='xlsxwriter') as writer:
    df.to_excel(writer, sheet_name='Sheet1', index=False)
    
    # è·å–å·¥ä½œè¡¨å¯¹è±¡
    worksheet = writer.sheets['Sheet1']
    
    # è®¾ç½®åˆ—å®½
    worksheet.set_column('A:D', 15)

print("å·²å¯¼å‡ºï¼Œå·²è®¾ç½®åˆ—å®½")

# xlsxwriteræ”¯æŒæ›´å¤šæ ¼å¼è®¾ç½®
```
