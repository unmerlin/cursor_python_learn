## ğŸ“¦ å¯¼å…¥åº“å¹¶åˆ›å»ºç¤ºä¾‹æ•°æ®

```python
import pandas as pd
import numpy as np

# åˆ›å»ºç¤ºä¾‹æ•°æ®
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'å§“å': ['å¼ ä¸‰', 'æå››', 'ç‹äº”', 'èµµå…­'],
    'éƒ¨é—¨': ['é”€å”®', 'æŠ€æœ¯', 'é”€å”®', 'æŠ€æœ¯']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3, 5],
    'å·¥èµ„': [8000, 12000, 10000, 15000],
    'å¥–é‡‘': [2000, 3000, 2500, 4000]
})

print("æ•°æ®è¡¨1ï¼š")
print(df1)
print("\næ•°æ®è¡¨2ï¼š")
print(df2)
```

# ç¬¬ä¸€éƒ¨åˆ†ï¼šmerge() - åŸºäºé”®å€¼åˆå¹¶

## â­ merge()æ˜¯æœ€å¸¸ç”¨çš„åˆå¹¶æ–¹æ³•ï¼Œç±»ä¼¼SQLçš„JOINæ“ä½œ

### 1. å†…è¿æ¥ï¼ˆinner joinï¼‰

```python
# å†…è¿æ¥ï¼šåªä¿ç•™ä¸¤ä¸ªè¡¨ä¸­éƒ½æœ‰çš„é”®å€¼
result = pd.merge(df1, df2, on='ID', how='inner')
print("å†…è¿æ¥ç»“æœï¼š")
print(result)

# å‚æ•°è¯´æ˜ï¼š
# - on: åˆå¹¶çš„é”®ï¼ˆåˆ—åï¼‰
# - how: è¿æ¥ç±»å‹ï¼ˆ'inner'é»˜è®¤å€¼ï¼‰
```

### 2. å·¦è¿æ¥ï¼ˆleft joinï¼‰

```python
# å·¦è¿æ¥ï¼šä¿ç•™å·¦è¡¨çš„æ‰€æœ‰è®°å½•
result = pd.merge(df1, df2, on='ID', how='left')
print("å·¦è¿æ¥ç»“æœï¼š")
print(result)

# å·¦è¡¨ä¸­æ²¡æœ‰åŒ¹é…çš„è®°å½•ï¼Œå³è¡¨åˆ—æ˜¾ç¤ºNaN
```

### 3. å³è¿æ¥ï¼ˆright joinï¼‰

```python
# å³è¿æ¥ï¼šä¿ç•™å³è¡¨çš„æ‰€æœ‰è®°å½•
result = pd.merge(df1, df2, on='ID', how='right')
print("å³è¿æ¥ç»“æœï¼š")
print(result)

# å³è¡¨ä¸­æ²¡æœ‰åŒ¹é…çš„è®°å½•ï¼Œå·¦è¡¨åˆ—æ˜¾ç¤ºNaN
```

### 4. å¤–è¿æ¥ï¼ˆouter joinï¼‰

```python
# å¤–è¿æ¥ï¼šä¿ç•™ä¸¤ä¸ªè¡¨çš„æ‰€æœ‰è®°å½•
result = pd.merge(df1, df2, on='ID', how='outer')
print("å¤–è¿æ¥ç»“æœï¼š")
print(result)

# æ²¡æœ‰åŒ¹é…çš„è®°å½•ï¼Œå¯¹åº”åˆ—æ˜¾ç¤ºNaN
```

### 5. ä¸åŒåˆ—ååˆå¹¶

```python
# å½“ä¸¤ä¸ªè¡¨çš„é”®åˆ—åä¸åŒæ—¶ï¼Œä½¿ç”¨left_onå’Œright_on
df3 = pd.DataFrame({
    'å‘˜å·¥ID': [1, 2, 3],
    'å¹´é¾„': [25, 30, 28]
})

result = pd.merge(df1, df3, left_on='ID', right_on='å‘˜å·¥ID', how='left')
print("ä¸åŒåˆ—ååˆå¹¶ï¼š")
print(result)

# left_on: å·¦è¡¨çš„é”®åˆ—
# right_on: å³è¡¨çš„é”®åˆ—
```

### 6. å¤šåˆ—åˆå¹¶

```python
# åŸºäºå¤šåˆ—è¿›è¡Œåˆå¹¶
df4 = pd.DataFrame({
    'ID': [1, 2, 3],
    'éƒ¨é—¨': ['é”€å”®', 'æŠ€æœ¯', 'é”€å”®'],
    'ç»©æ•ˆ': ['A', 'B', 'A']
})

result = pd.merge(df1, df4, on=['ID', 'éƒ¨é—¨'], how='inner')
print("å¤šåˆ—åˆå¹¶ï¼š")
print(result)

# onå‚æ•°å¯ä»¥æ˜¯åˆ—è¡¨ï¼ŒæŒ‡å®šå¤šä¸ªé”®åˆ—
```

### 7. å¤„ç†é‡å¤åˆ—åï¼ˆsuffixesï¼‰

```python
# å½“ä¸¤ä¸ªè¡¨æœ‰ç›¸åŒçš„åˆ—åæ—¶ï¼Œä½¿ç”¨suffixeså‚æ•°æ·»åŠ åç¼€
df5 = pd.DataFrame({
    'ID': [1, 2, 3],
    'éƒ¨é—¨': ['é”€å”®', 'æŠ€æœ¯', 'äººäº‹']
})

result = pd.merge(df1, df5, on='ID', suffixes=('_å·¦è¡¨', '_å³è¡¨'))
print("å¤„ç†é‡å¤åˆ—åï¼š")
print(result)

# suffixes: ä¸ºé‡å¤åˆ—åæ·»åŠ åç¼€ï¼ˆé»˜è®¤('_x', '_y')ï¼‰
```

### 8. åŸºäºç´¢å¼•åˆå¹¶

```python
# ä½¿ç”¨left_indexå’Œright_indexåŸºäºç´¢å¼•åˆå¹¶
df1_indexed = df1.set_index('ID')
df2_indexed = df2.set_index('ID')

result = pd.merge(df1_indexed, df2_indexed, left_index=True, right_index=True, how='inner')
print("åŸºäºç´¢å¼•åˆå¹¶ï¼š")
print(result)

# left_index=True: ä½¿ç”¨å·¦è¡¨çš„ç´¢å¼•ä½œä¸ºé”®
# right_index=True: ä½¿ç”¨å³è¡¨çš„ç´¢å¼•ä½œä¸ºé”®
```

# ç¬¬äºŒéƒ¨åˆ†ï¼šconcat() - è¿æ¥å¤šä¸ªDataFrame

## â­ concat()ç”¨äºæ²¿è½´è¿æ¥å¤šä¸ªDataFrame

### 1. å‚ç›´è¿æ¥ï¼ˆaxis=0ï¼‰

```python
# å‚ç›´è¿æ¥ï¼šå°†å¤šä¸ªDataFrameæŒ‰è¡Œå †å 
df_a = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df_b = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

result = pd.concat([df_a, df_b], axis=0)
print("å‚ç›´è¿æ¥ï¼š")
print(result)

# axis=0: æ²¿è¡Œæ–¹å‘è¿æ¥ï¼ˆé»˜è®¤å€¼ï¼‰
```

### 2. æ°´å¹³è¿æ¥ï¼ˆaxis=1ï¼‰

```python
# æ°´å¹³è¿æ¥ï¼šå°†å¤šä¸ªDataFrameæŒ‰åˆ—æ‹¼æ¥
df_c = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

result = pd.concat([df_a, df_c], axis=1)
print("æ°´å¹³è¿æ¥ï¼š")
print(result)

# axis=1: æ²¿åˆ—æ–¹å‘è¿æ¥
```

### 3. å¿½ç•¥ç´¢å¼•ï¼ˆignore_indexï¼‰

```python
# ignore_index=True: é‡æ–°ç”Ÿæˆç´¢å¼•
result = pd.concat([df_a, df_b], axis=0, ignore_index=True)
print("å¿½ç•¥åŸç´¢å¼•ï¼š")
print(result)

# åŸç´¢å¼•è¢«å¿½ç•¥ï¼Œç”Ÿæˆæ–°çš„è¿ç»­ç´¢å¼•
```

### 4. æ·»åŠ é”®æ ‡è¯†ï¼ˆkeysï¼‰

```python
# keyså‚æ•°ï¼šä¸ºæ¯ä¸ªDataFrameæ·»åŠ æ ‡è¯†
result = pd.concat([df_a, df_b], axis=0, keys=['è¡¨1', 'è¡¨2'])
print("æ·»åŠ é”®æ ‡è¯†ï¼š")
print(result)

# åˆ›å»ºå¤šçº§ç´¢å¼•ï¼Œå¯ä»¥åŒºåˆ†æ•°æ®æ¥æº
```

### 5. å¤„ç†ä¸åŒåˆ—ï¼ˆjoinï¼‰

```python
# å½“DataFrameåˆ—ä¸åŒæ—¶ï¼Œä½¿ç”¨joinå‚æ•°
df_d = pd.DataFrame({'A': [1, 2], 'C': [3, 4]})

# join='outer': ä¿ç•™æ‰€æœ‰åˆ—ï¼ˆé»˜è®¤ï¼‰
result_outer = pd.concat([df_a, df_d], axis=0, join='outer')
print("å¤–è¿æ¥ï¼ˆä¿ç•™æ‰€æœ‰åˆ—ï¼‰ï¼š")
print(result_outer)

# join='inner': åªä¿ç•™å…±åŒåˆ—
result_inner = pd.concat([df_a, df_d], axis=0, join='inner')
print("\nå†…è¿æ¥ï¼ˆåªä¿ç•™å…±åŒåˆ—ï¼‰ï¼š")
print(result_inner)
```

# ç¬¬ä¸‰éƒ¨åˆ†ï¼šjoin() - åŸºäºç´¢å¼•åˆå¹¶

```python
# join()é»˜è®¤åŸºäºç´¢å¼•åˆå¹¶
df_left = pd.DataFrame({'A': [1, 2], 'B': [3, 4]}, index=['x', 'y'])
df_right = pd.DataFrame({'C': [5, 6], 'D': [7, 8]}, index=['x', 'y'])

result = df_left.join(df_right)
print("åŸºæœ¬joinæ“ä½œï¼š")
print(result)

# join()é»˜è®¤æ˜¯å·¦è¿æ¥
```
