# ç¬¬ä¸‰éƒ¨åˆ†ï¼šæå–æ—¶é—´ä¿¡æ¯

## â­ ä»æ—¶é—´ä¸­æå–å„ç§ä¿¡æ¯

### 1. æå–å¹´ã€æœˆã€æ—¥

```python
# åˆ›å»ºæ—¶é—´åºåˆ—æ•°æ®
dates = pd.date_range('2024-01-01', periods=5, freq='D')
df = pd.DataFrame({'æ—¥æœŸ': dates, 'å€¼': range(5)})

# æå–å¹´ã€æœˆã€æ—¥
df['å¹´'] = df['æ—¥æœŸ'].dt.year
df['æœˆ'] = df['æ—¥æœŸ'].dt.month
df['æ—¥'] = df['æ—¥æœŸ'].dt.day

print("æå–å¹´ã€æœˆã€æ—¥ï¼š")
print(df)
```

### 2. æå–æ˜ŸæœŸã€å­£åº¦

```python
# æå–æ˜ŸæœŸå’Œå­£åº¦
df['æ˜ŸæœŸ'] = df['æ—¥æœŸ'].dt.dayofweek  # 0=å‘¨ä¸€, 6=å‘¨æ—¥
df['æ˜ŸæœŸå'] = df['æ—¥æœŸ'].dt.day_name()  # æ˜ŸæœŸåç§°
df['å­£åº¦'] = df['æ—¥æœŸ'].dt.quarter

print("æå–æ˜ŸæœŸå’Œå­£åº¦ï¼š")
print(df[['æ—¥æœŸ', 'æ˜ŸæœŸ', 'æ˜ŸæœŸå', 'å­£åº¦']])
```

### 3. æå–æ—¶ã€åˆ†ã€ç§’

```python
# åˆ›å»ºå¸¦æ—¶é—´çš„æ—¥æœŸ
datetime_index = pd.date_range('2024-01-01 10:30:45', periods=3, freq='H')
df_time = pd.DataFrame({'æ—¶é—´': datetime_index})

# æå–æ—¶ã€åˆ†ã€ç§’
df_time['æ—¶'] = df_time['æ—¶é—´'].dt.hour
df_time['åˆ†'] = df_time['æ—¶é—´'].dt.minute
df_time['ç§’'] = df_time['æ—¶é—´'].dt.second

print("æå–æ—¶ã€åˆ†ã€ç§’ï¼š")
print(df_time)
```

### 4. åˆ¤æ–­æ˜¯å¦ä¸ºå·¥ä½œæ—¥

```python
# åˆ¤æ–­æ˜¯å¦ä¸ºå·¥ä½œæ—¥
dates = pd.date_range('2024-01-01', periods=7, freq='D')
df_work = pd.DataFrame({'æ—¥æœŸ': dates})
df_work['æ˜¯å¦å·¥ä½œæ—¥'] = df_work['æ—¥æœŸ'].dt.is_weekday
df_work['æ˜¯å¦å‘¨æœ«'] = df_work['æ—¥æœŸ'].dt.dayofweek.isin([5, 6])

print("åˆ¤æ–­å·¥ä½œæ—¥ï¼š")
print(df_work)
```

# ç¬¬å››éƒ¨åˆ†ï¼šæ—¶é—´è¿ç®—

## â­ æ—¶é—´çš„åŠ å‡å’Œå·®å€¼è®¡ç®—

### 1. æ—¶é—´åŠ å‡

```python
# æ—¶é—´åŠ å‡
date = pd.to_datetime('2024-01-15')

# åŠ å¤©æ•°
date_plus_7 = date + pd.Timedelta(days=7)
print(f"åŸæ—¥æœŸï¼š{date}")
print(f"åŠ 7å¤©ï¼š{date_plus_7}")

# åŠ æœˆä»½ï¼ˆä½¿ç”¨DateOffsetï¼‰
date_plus_month = date + pd.DateOffset(months=1)
print(f"åŠ 1ä¸ªæœˆï¼š{date_plus_month}")
```

```python
# è®¡ç®—æ—¶é—´å·®
date1 = pd.to_datetime('2024-01-15')
date2 = pd.to_datetime('2024-01-20')

diff = date2 - date1
print(f"æ—¶é—´å·®ï¼š{diff}")
print(f"å¤©æ•°ï¼š{diff.days}")

# æ—¶é—´å·®ç±»å‹æ˜¯Timedelta
```

```python
# è®¡ç®—æ—¶é—´åºåˆ—ä¸­ç›¸é‚»æ—¶é—´çš„å·®å€¼
dates = pd.date_range('2024-01-01', periods=5, freq='D')
df = pd.DataFrame({'æ—¥æœŸ': dates, 'å€¼': range(5)})

# è®¡ç®—ç›¸é‚»æ—¶é—´å·®
df['æ—¶é—´å·®'] = df['æ—¥æœŸ'].diff()
print("æ—¶é—´å·®ï¼š")
print(df)
```

# ç¬¬äº”éƒ¨åˆ†ï¼šæ—¶é—´é‡é‡‡æ ·ï¼ˆresampleï¼‰

## â­ é‡é‡‡æ ·æ˜¯æ—¶é—´åºåˆ—åˆ†æçš„æ ¸å¿ƒåŠŸèƒ½

### 1. åˆ›å»ºç¤ºä¾‹æ•°æ®

```python
# åˆ›å»ºæ—¥åº¦æ•°æ®
dates = pd.date_range('2024-01-01', periods=30, freq='D')
df = pd.DataFrame({
    'é”€å”®é¢': np.random.randint(1000, 5000, 30)
}, index=dates)

print("åŸå§‹æ•°æ®ï¼ˆæ—¥åº¦ï¼‰ï¼š")
print(df.head())
```

### 2. é™é‡‡æ · - æ—¥åº¦è½¬æœˆåº¦

```python
# é™é‡‡æ ·ï¼šå°†æ—¥åº¦æ•°æ®èšåˆä¸ºæœˆåº¦æ•°æ®
monthly = df.resample('M').sum()  # 'M'=æœˆæœ«
print("æœˆåº¦æ±‡æ€»ï¼ˆæ±‚å’Œï¼‰ï¼š")
print(monthly)

# å¸¸ç”¨é¢‘ç‡ï¼š'D'(æ—¥), 'W'(å‘¨), 'M'(æœˆ), 'Q'(å­£), 'Y'(å¹´)
```

### 3. é™é‡‡æ · - ä½¿ç”¨ä¸åŒèšåˆå‡½æ•°

```python
# ä½¿ç”¨ä¸åŒçš„èšåˆå‡½æ•°
monthly_mean = df.resample('M').mean()  # å¹³å‡å€¼
monthly_max = df.resample('M').max()    # æœ€å¤§å€¼
monthly_min = df.resample('M').min()    # æœ€å°å€¼

print("æœˆåº¦å¹³å‡å€¼ï¼š")
print(monthly_mean)
print("\næœˆåº¦æœ€å¤§å€¼ï¼š")
print(monthly_max)
```

### 4. å‡é‡‡æ · - æœˆåº¦è½¬æ—¥åº¦

```python
# å‡é‡‡æ ·ï¼šå°†æœˆåº¦æ•°æ®æ‰©å±•ä¸ºæ—¥åº¦æ•°æ®
monthly_data = df.resample('M').sum()

# å‰å‘å¡«å……
daily_ffill = monthly_data.resample('D').ffill()
print("å‡é‡‡æ ·ï¼ˆå‰å‘å¡«å……ï¼‰ï¼š")
print(daily_ffill.head(10))

# æ’å€¼å¡«å……
daily_interpolate = monthly_data.resample('D').interpolate()
print("\nå‡é‡‡æ ·ï¼ˆæ’å€¼ï¼‰ï¼š")
print(daily_interpolate.head(10))
```

# ç¬¬å…­éƒ¨åˆ†ï¼šæ—¶é—´çª—å£æ“ä½œ

## â­ æ»šåŠ¨çª—å£å’Œæ‰©å±•çª—å£

### 1. æ»šåŠ¨çª—å£ï¼ˆrollingï¼‰

```python
# åˆ›å»ºç¤ºä¾‹æ•°æ®
dates = pd.date_range('2024-01-01', periods=10, freq='D')
df = pd.DataFrame({'é”€å”®é¢': np.random.randint(1000, 5000, 10)}, index=dates)

# 3æ—¥æ»šåŠ¨å¹³å‡
df['3æ—¥å¹³å‡'] = df['é”€å”®é¢'].rolling(window=3).mean()
print("æ»šåŠ¨çª—å£ï¼ˆ3æ—¥å¹³å‡ï¼‰ï¼š")
print(df)
```

### 2. æ»šåŠ¨çª—å£ - ä¸åŒèšåˆå‡½æ•°

```python
# æ»šåŠ¨çª—å£çš„å…¶ä»–å‡½æ•°
df['3æ—¥æ€»å’Œ'] = df['é”€å”®é¢'].rolling(window=3).sum()
df['3æ—¥æœ€å¤§å€¼'] = df['é”€å”®é¢'].rolling(window=3).max()
df['3æ—¥æœ€å°å€¼'] = df['é”€å”®é¢'].rolling(window=3).min()

print("æ»šåŠ¨çª—å£ç»Ÿè®¡ï¼š")
print(df[['é”€å”®é¢', '3æ—¥æ€»å’Œ', '3æ—¥æœ€å¤§å€¼', '3æ—¥æœ€å°å€¼']])
```

### 3. æ‰©å±•çª—å£ï¼ˆexpandingï¼‰

```python
# æ‰©å±•çª—å£ï¼šä»å¼€å§‹åˆ°å½“å‰çš„æ‰€æœ‰æ•°æ®
df['ç´¯è®¡å¹³å‡'] = df['é”€å”®é¢'].expanding().mean()
df['ç´¯è®¡æ€»å’Œ'] = df['é”€å”®é¢'].expanding().sum()

print("æ‰©å±•çª—å£ï¼š")
print(df[['é”€å”®é¢', 'ç´¯è®¡å¹³å‡', 'ç´¯è®¡æ€»å’Œ']])
```

### 4. æ—¶é—´çª—å£ï¼ˆåŸºäºæ—¶é—´é•¿åº¦ï¼‰

```python
# åŸºäºæ—¶é—´é•¿åº¦åˆ›å»ºçª—å£ï¼ˆè€Œä¸æ˜¯å›ºå®šæ•°é‡ï¼‰
dates = pd.date_range('2024-01-01', periods=10, freq='D')
df = pd.DataFrame({'é”€å”®é¢': np.random.randint(1000, 5000, 10)}, index=dates)

# 7å¤©æ»šåŠ¨çª—å£
df['7å¤©å¹³å‡'] = df['é”€å”®é¢'].rolling(window='7D').mean()
print("åŸºäºæ—¶é—´çš„æ»šåŠ¨çª—å£ï¼š")
print(df)
```
