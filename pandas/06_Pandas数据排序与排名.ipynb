## ğŸ“¦ å¯¼å…¥åº“å¹¶åˆ›å»ºç¤ºä¾‹æ•°æ®

```python
import pandas as pd
import numpy as np

# åˆ›å»ºç¤ºä¾‹æ•°æ®
data = {
    'å§“å': ['å¼ ä¸‰', 'æå››', 'ç‹äº”', 'èµµå…­', 'é’±ä¸ƒ', 'å­™å…«'],
    'å¹´é¾„': [25, 30, 28, 35, 32, 27],
    'è–ªèµ„': [8000, 12000, 9000, 15000, 11000, 9500],
    'åŸå¸‚': ['åŒ—äº¬', 'ä¸Šæµ·', 'å¹¿å·', 'æ·±åœ³', 'åŒ—äº¬', 'ä¸Šæµ·'],
    'éƒ¨é—¨': ['é”€å”®', 'æŠ€æœ¯', 'é”€å”®', 'æŠ€æœ¯', 'å¸‚åœº', 'æŠ€æœ¯']
}
df = pd.DataFrame(data)

print("ç¤ºä¾‹æ•°æ®ï¼š")
print(df)
```

# ç¬¬ä¸€éƒ¨åˆ†ï¼šsort_values() - æŒ‰å€¼æ’åº

## â­ æ ¹æ®åˆ—çš„å€¼è¿›è¡Œæ’åº

### 1. å•åˆ—å‡åºæ’åº

```python
# æŒ‰å¹´é¾„å‡åºæ’åºï¼ˆä»å°åˆ°å¤§ï¼‰
result = df.sort_values('å¹´é¾„')
print("æŒ‰å¹´é¾„å‡åºæ’åºï¼š")
print(result)

# æ³¨æ„ï¼šé»˜è®¤æ˜¯å‡åºï¼ˆascending=Trueï¼‰
```

### 2. å•åˆ—é™åºæ’åº

```python
# æŒ‰è–ªèµ„é™åºæ’åºï¼ˆä»å¤§åˆ°å°ï¼‰
result = df.sort_values('è–ªèµ„', ascending=False)
print("æŒ‰è–ªèµ„é™åºæ’åºï¼š")
print(result)
```

### 3. å¤šåˆ—æ’åº

```python
# å…ˆæŒ‰åŸå¸‚æ’åºï¼Œå†æŒ‰è–ªèµ„æ’åº
result = df.sort_values(['åŸå¸‚', 'è–ªèµ„'])
print("å…ˆæŒ‰åŸå¸‚ï¼Œå†æŒ‰è–ªèµ„æ’åºï¼š")
print(result)

# å¤šåˆ—æ’åºæ—¶ï¼Œå…ˆæŒ‰ç¬¬ä¸€åˆ—ï¼Œç›¸åŒå€¼å†æŒ‰ç¬¬äºŒåˆ—
```

### 4. å¤šåˆ—ä¸åŒæ’åºæ–¹å‘

```python
# åŸå¸‚å‡åºï¼Œè–ªèµ„é™åº
result = df.sort_values(['åŸå¸‚', 'è–ªèµ„'], ascending=[True, False])
print("åŸå¸‚å‡åºï¼Œè–ªèµ„é™åºï¼š")
print(result)

# ascendingå‚æ•°å¯ä»¥æ˜¯åˆ—è¡¨ï¼Œå¯¹åº”æ¯åˆ—çš„æ’åºæ–¹å‘
```

### 5. æ’åºåé‡ç½®ç´¢å¼•

```python
# æ’åºåé‡ç½®ç´¢å¼•ï¼ˆç´¢å¼•ä»0å¼€å§‹é‡æ–°ç¼–å·ï¼‰
result = df.sort_values('è–ªèµ„', ascending=False).reset_index(drop=True)
print("æŒ‰è–ªèµ„é™åºï¼Œé‡ç½®ç´¢å¼•ï¼š")
print(result)

# drop=True: åˆ é™¤åŸæ¥çš„ç´¢å¼•åˆ—
# drop=False: ä¿ç•™åŸæ¥çš„ç´¢å¼•ä½œä¸ºæ–°åˆ—
```

### 6. åŸåœ°æ’åº

```python
# åŸåœ°æ’åºï¼ˆç›´æ¥ä¿®æ”¹åŸDataFrameï¼‰
df_copy = df.copy()
df_copy.sort_values('å¹´é¾„', inplace=True)
print("åŸåœ°æ’åºåçš„æ•°æ®ï¼š")
print(df_copy)

# inplace=True: ç›´æ¥ä¿®æ”¹åŸDataFrameï¼Œä¸è¿”å›æ–°å¯¹è±¡
```

---
# ç¬¬äºŒéƒ¨åˆ†ï¼šsort_index() - æŒ‰ç´¢å¼•æ’åº

## â­ æ ¹æ®ç´¢å¼•è¿›è¡Œæ’åº

### 1. æŒ‰è¡Œç´¢å¼•æ’åº

```python
# åˆ›å»ºå¸¦è‡ªå®šä¹‰ç´¢å¼•çš„æ•°æ®
df_indexed = df.copy()
df_indexed.index = ['C', 'A', 'B', 'F', 'D', 'E']

print("åŸå§‹æ•°æ®ï¼ˆç´¢å¼•æœªæ’åºï¼‰ï¼š")
print(df_indexed)

# æŒ‰ç´¢å¼•æ’åº
result = df_indexed.sort_index()
print("\næŒ‰ç´¢å¼•æ’åºåï¼š")
print(result)
```

### 2. æŒ‰åˆ—ç´¢å¼•æ’åº

```python
# æŒ‰åˆ—åæ’åºï¼ˆaxis=1è¡¨ç¤ºåˆ—ï¼‰
df_cols = df[['è–ªèµ„', 'å§“å', 'å¹´é¾„', 'åŸå¸‚']]
print("åŸå§‹åˆ—é¡ºåºï¼š")
print(df_cols.columns.tolist())

# æŒ‰åˆ—åæ’åº
result = df_cols.sort_index(axis=1)
print("\næŒ‰åˆ—åæ’åºåï¼š")
print(result.columns.tolist())
```

### 3. ç´¢å¼•é™åºæ’åº

```python
# ç´¢å¼•é™åºæ’åº
result = df_indexed.sort_index(ascending=False)
print("ç´¢å¼•é™åºæ’åºï¼š")
print(result)
```

---
# ç¬¬ä¸‰éƒ¨åˆ†ï¼šrank() - æ’å

## â­ è®¡ç®—æ•°æ®çš„æ’å

### 1. åŸºæœ¬æ’å

```python
# è®¡ç®—è–ªèµ„æ’åï¼ˆé»˜è®¤æ–¹æ³•ï¼šaverageï¼‰
df['è–ªèµ„æ’å'] = df['è–ªèµ„'].rank()
print("æ·»åŠ è–ªèµ„æ’ååˆ—ï¼š")
print(df[['å§“å', 'è–ªèµ„', 'è–ªèµ„æ’å']])

# rank()è¿”å›æ’åï¼Œ1è¡¨ç¤ºæœ€å°å€¼ï¼Œæ•°å€¼è¶Šå¤§æ’åè¶Šé«˜
```

### 2. æ’åæ–¹æ³• - min

```python
# method='min': ç›¸åŒå€¼å–æœ€å°æ’å
df['æ’å_min'] = df['è–ªèµ„'].rank(method='min')
print("ä½¿ç”¨minæ–¹æ³•æ’åï¼š")
print(df[['å§“å', 'è–ªèµ„', 'æ’å_min']])

# ç›¸åŒå€¼éƒ½å–æœ€å°æ’åï¼ˆå¦‚ä¸¤ä¸ªç¬¬2åï¼Œä¸‹ä¸€ä¸ªæ˜¯ç¬¬4åï¼‰
```

### 3. æ’åæ–¹æ³• - max

```python
# method='max': ç›¸åŒå€¼å–æœ€å¤§æ’å
df['æ’å_max'] = df['è–ªèµ„'].rank(method='max')
print("ä½¿ç”¨maxæ–¹æ³•æ’åï¼š")
print(df[['å§“å', 'è–ªèµ„', 'æ’å_max']])

# ç›¸åŒå€¼éƒ½å–æœ€å¤§æ’åï¼ˆå¦‚ä¸¤ä¸ªç¬¬3åï¼Œä¸‹ä¸€ä¸ªæ˜¯ç¬¬5åï¼‰
```

### 4. æ’åæ–¹æ³• - first

```python
# method='first': æŒ‰å‡ºç°é¡ºåºæ’å
df['æ’å_first'] = df['è–ªèµ„'].rank(method='first')
print("ä½¿ç”¨firstæ–¹æ³•æ’åï¼š")
print(df[['å§“å', 'è–ªèµ„', 'æ’å_first']])

# ç›¸åŒå€¼æŒ‰åœ¨æ•°æ®ä¸­å‡ºç°çš„é¡ºåºæ’å
```

### 5. é™åºæ’å

```python
# ascending=False: é™åºæ’åï¼ˆæ•°å€¼è¶Šå¤§æ’åè¶Šå°ï¼‰
df['è–ªèµ„æ’å_é™åº'] = df['è–ªèµ„'].rank(ascending=False)
print("é™åºæ’åï¼ˆè–ªèµ„è¶Šé«˜æ’åè¶Šå°ï¼‰ï¼š")
print(df[['å§“å', 'è–ªèµ„', 'è–ªèµ„æ’å_é™åº']])

# å®é™…åº”ç”¨ï¼šæ’è¡Œæ¦œï¼ˆç¬¬1åæ˜¯æœ€å¥½çš„ï¼‰
```

# ç¬¬å››éƒ¨åˆ†ï¼šå¤„ç†ç¼ºå¤±å€¼çš„æ’åº

## â­ æ’åºæ—¶ç¼ºå¤±å€¼çš„å¤„ç†

### 1. ç¼ºå¤±å€¼æ’åœ¨æœ€å

```python
# åˆ›å»ºåŒ…å«ç¼ºå¤±å€¼çš„æ•°æ®
df_na = df.copy()
df_na.loc[2, 'è–ªèµ„'] = np.nan

print("åŒ…å«ç¼ºå¤±å€¼çš„æ•°æ®ï¼š")
print(df_na[['å§“å', 'è–ªèµ„']])

# ç¼ºå¤±å€¼é»˜è®¤æ’åœ¨æœ€å
result = df_na.sort_values('è–ªèµ„')
print("\næ’åºåï¼ˆç¼ºå¤±å€¼åœ¨æœ€åï¼‰ï¼š")
print(result[['å§“å', 'è–ªèµ„']])
```

### 2. ç¼ºå¤±å€¼æ’åœ¨æœ€å‰

```python
# na_position='first': ç¼ºå¤±å€¼æ’åœ¨æœ€å‰
result = df_na.sort_values('è–ªèµ„', na_position='first')
print("æ’åºåï¼ˆç¼ºå¤±å€¼åœ¨æœ€å‰ï¼‰ï¼š")
print(result[['å§“å', 'è–ªèµ„']])

# na_positionå‚æ•°ï¼š
# - 'last': ç¼ºå¤±å€¼æ’åœ¨æœ€åï¼ˆé»˜è®¤ï¼‰
# - 'first': ç¼ºå¤±å€¼æ’åœ¨æœ€å‰
```

```python
# åˆ›å»ºè–ªèµ„æ’è¡Œæ¦œ
df_rank = df.copy()
df_rank = df_rank.sort_values('è–ªèµ„', ascending=False)
df_rank['æ’å'] = range(1, len(df_rank) + 1)

print("è–ªèµ„æ’è¡Œæ¦œï¼š")
print(df_rank[['æ’å', 'å§“å', 'è–ªèµ„']])

# å®é™…åº”ç”¨ï¼šå‘˜å·¥ç»©æ•ˆæ’åã€é”€å”®æ’è¡Œæ¦œç­‰
```

```python
# åœ¨æ¯ä¸ªéƒ¨é—¨å†…æŒ‰è–ªèµ„æ’åº
df_grouped = df.sort_values(['éƒ¨é—¨', 'è–ªèµ„'], ascending=[True, False])
print("æŒ‰éƒ¨é—¨åˆ†ç»„ï¼Œéƒ¨é—¨å†…æŒ‰è–ªèµ„é™åºï¼š")
print(df_grouped[['éƒ¨é—¨', 'å§“å', 'è–ªèµ„']])

# å®é™…åº”ç”¨ï¼šå„éƒ¨é—¨å†…éƒ¨æ’å
```
