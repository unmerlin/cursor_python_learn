## ğŸ“¦ å¯¼å…¥åº“å¹¶åˆ›å»ºç¤ºä¾‹æ•°æ®

```python
import pandas as pd
import numpy as np

# åˆ›å»ºç¤ºä¾‹æ•°æ®
data = {
    'å§“å': ['å¼ ä¸‰', 'æå››', 'ç‹äº”', 'èµµå…­', 'é’±ä¸ƒ'],
    'å¹´é¾„': [25, 30, 28, 35, 32],
    'è–ªèµ„': [8000, 12000, 9000, 15000, 11000],
    'åŸå¸‚': ['åŒ—äº¬', 'ä¸Šæµ·', 'å¹¿å·', 'æ·±åœ³', 'åŒ—äº¬']
}
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D', 'E'])

print("ç¤ºä¾‹æ•°æ®ï¼š")
print(df)
```

# ç¬¬ä¸€éƒ¨åˆ†ï¼šloc - æ ‡ç­¾ç´¢å¼•

## â­ ä½¿ç”¨æ ‡ç­¾ï¼ˆç´¢å¼•åå’Œåˆ—åï¼‰é€‰æ‹©æ•°æ®

`loc` åŸºäºæ ‡ç­¾è¿›è¡Œç´¢å¼•ï¼Œä½¿ç”¨è¡Œç´¢å¼•åå’Œåˆ—åæ¥é€‰æ‹©æ•°æ®ã€‚

### 1. é€‰æ‹©å•è¡Œ

```python
# é€‰æ‹©ç´¢å¼•ä¸º'A'çš„è¡Œ
row_a = df.loc['A']
print("é€‰æ‹©ç´¢å¼•'A'çš„è¡Œï¼š")
print(row_a)
print(f"\nç±»å‹: {type(row_a)}")  # Seriesç±»å‹
```

### 2. é€‰æ‹©å¤šè¡Œ

```python
# é€‰æ‹©å¤šä¸ªè¡Œï¼ˆä½¿ç”¨åˆ—è¡¨ï¼‰
rows = df.loc[['A', 'C', 'E']]
print("é€‰æ‹©ç´¢å¼•'A', 'C', 'E'çš„è¡Œï¼š")
print(rows)
```

### 3. é€‰æ‹©è¡Œå’Œåˆ—ï¼ˆè¡Œåœ¨å‰ï¼Œåˆ—åœ¨åï¼‰

```python
# é€‰æ‹©ç‰¹å®šè¡Œå’Œåˆ—
result = df.loc['A', 'å§“å']  # è¡Œ'A'ï¼Œåˆ—'å§“å'
print(f"é€‰æ‹©è¡Œ'A'åˆ—'å§“å': {result}")

# é€‰æ‹©è¡Œå’Œå¤šä¸ªåˆ—
result2 = df.loc['A', ['å§“å', 'å¹´é¾„']]
print("\né€‰æ‹©è¡Œ'A'ï¼Œåˆ—'å§“å'å’Œ'å¹´é¾„'ï¼š")
print(result2)
```

### 4. é€‰æ‹©å¤šè¡Œå¤šåˆ—

```python
# é€‰æ‹©å¤šè¡Œå¤šåˆ—
result = df.loc[['A', 'B'], ['å§“å', 'è–ªèµ„']]
print("é€‰æ‹©è¡Œ'A','B'ï¼Œåˆ—'å§“å','è–ªèµ„'ï¼š")
print(result)
```

### 5. ä½¿ç”¨åˆ‡ç‰‡é€‰æ‹©è¡Œ

```python
# ä½¿ç”¨åˆ‡ç‰‡ï¼ˆåŒ…å«ç»“æŸä½ç½®ï¼‰
result = df.loc['A':'C']  # ä»'A'åˆ°'C'ï¼ˆåŒ…å«'C'ï¼‰
print("é€‰æ‹©ä»'A'åˆ°'C'çš„è¡Œï¼š")
print(result)

# æ³¨æ„ï¼šlocçš„åˆ‡ç‰‡æ˜¯åŒ…å«ç»“æŸä½ç½®çš„ï¼Œä¸Pythonåˆ—è¡¨ä¸åŒ
```

### 6. ä½¿ç”¨æ¡ä»¶é€‰æ‹©è¡Œ

```python
# ä½¿ç”¨æ¡ä»¶é€‰æ‹©è¡Œ
result = df.loc[df['å¹´é¾„'] > 30]
print("é€‰æ‹©å¹´é¾„å¤§äº30çš„è¡Œï¼š")
print(result)

# æ¡ä»¶é€‰æ‹©åï¼Œå†é€‰æ‹©ç‰¹å®šåˆ—
result2 = df.loc[df['å¹´é¾„'] > 30, ['å§“å', 'è–ªèµ„']]
print("\nå¹´é¾„>30çš„è¡Œï¼Œåªæ˜¾ç¤ºå§“åå’Œè–ªèµ„ï¼š")
print(result2)
```

# ç¬¬äºŒéƒ¨åˆ†ï¼šiloc - ä½ç½®ç´¢å¼•

## â­ ä½¿ç”¨ä½ç½®ï¼ˆæ•´æ•°ä½ç½®ï¼‰é€‰æ‹©æ•°æ®

`iloc` åŸºäºæ•´æ•°ä½ç½®è¿›è¡Œç´¢å¼•ï¼Œç±»ä¼¼äºPythonåˆ—è¡¨çš„ç´¢å¼•æ–¹å¼ã€‚

### 1. é€‰æ‹©å•è¡Œ

```python
# é€‰æ‹©ç¬¬ä¸€è¡Œï¼ˆç´¢å¼•ä»0å¼€å§‹ï¼‰
row_0 = df.iloc[0]
print("é€‰æ‹©ç¬¬ä¸€è¡Œï¼š")
print(row_0)
```

### 2. é€‰æ‹©å¤šè¡Œ

```python
# é€‰æ‹©å¤šè¡Œï¼ˆä½¿ç”¨åˆ—è¡¨ï¼‰
rows = df.iloc[[0, 2, 4]]
print("é€‰æ‹©ç¬¬1ã€3ã€5è¡Œï¼š")
print(rows)
```

### 3. ä½¿ç”¨åˆ‡ç‰‡é€‰æ‹©è¡Œ

```python
# ä½¿ç”¨åˆ‡ç‰‡ï¼ˆä¸åŒ…å«ç»“æŸä½ç½®ï¼Œä¸Pythonåˆ—è¡¨ç›¸åŒï¼‰
result = df.iloc[0:3]  # ç¬¬1åˆ°ç¬¬3è¡Œï¼ˆä¸åŒ…å«ç¬¬4è¡Œï¼‰
print("é€‰æ‹©ç¬¬1åˆ°ç¬¬3è¡Œï¼š")
print(result)
```

### 4. é€‰æ‹©è¡Œå’Œåˆ—

```python
# é€‰æ‹©ç‰¹å®šè¡Œå’Œåˆ—
result = df.iloc[0, 0]  # ç¬¬1è¡Œï¼Œç¬¬1åˆ—
print(f"ç¬¬1è¡Œç¬¬1åˆ—: {result}")

# é€‰æ‹©è¡Œå’Œå¤šä¸ªåˆ—
result2 = df.iloc[0, [0, 1]]  # ç¬¬1è¡Œï¼Œç¬¬1å’Œç¬¬2åˆ—
print("\nç¬¬1è¡Œï¼Œç¬¬1å’Œç¬¬2åˆ—ï¼š")
print(result2)
```

### 5. é€‰æ‹©å¤šè¡Œå¤šåˆ—

```python
# é€‰æ‹©å¤šè¡Œå¤šåˆ—
result = df.iloc[0:2, 0:2]  # å‰2è¡Œï¼Œå‰2åˆ—
print("å‰2è¡Œï¼Œå‰2åˆ—ï¼š")
print(result)
```

### 6. é€‰æ‹©æ‰€æœ‰è¡Œæˆ–æ‰€æœ‰åˆ—

```python
# é€‰æ‹©æ‰€æœ‰è¡Œï¼Œç‰¹å®šåˆ—
result = df.iloc[:, [0, 2]]  # æ‰€æœ‰è¡Œï¼Œç¬¬1å’Œç¬¬3åˆ—
print("æ‰€æœ‰è¡Œï¼Œç¬¬1å’Œç¬¬3åˆ—ï¼š")
print(result)

# é€‰æ‹©ç‰¹å®šè¡Œï¼Œæ‰€æœ‰åˆ—
result2 = df.iloc[0:2, :]  # å‰2è¡Œï¼Œæ‰€æœ‰åˆ—
print("\nå‰2è¡Œï¼Œæ‰€æœ‰åˆ—ï¼š")
print(result2)
```

---
# ç¬¬ä¸‰éƒ¨åˆ†ï¼šå¸ƒå°”ç´¢å¼•

## â­ ä½¿ç”¨æ¡ä»¶è¡¨è¾¾å¼ç­›é€‰æ•°æ®

### 1. å•æ¡ä»¶ç­›é€‰

```python
# ç­›é€‰å¹´é¾„å¤§äº30çš„è¡Œ
result = df[df['å¹´é¾„'] > 30]
print("å¹´é¾„å¤§äº30çš„è¡Œï¼š")
print(result)

# ç­‰ä»·å†™æ³•
result2 = df.loc[df['å¹´é¾„'] > 30]
print("\nä½¿ç”¨locçš„ç­‰ä»·å†™æ³•ï¼š")
print(result2)
```

### 2. å¤šæ¡ä»¶ç­›é€‰ - ä½¿ç”¨&ï¼ˆä¸ï¼‰

```python
# å¤šæ¡ä»¶ç­›é€‰ï¼ˆå¿…é¡»ç”¨æ‹¬å·ï¼‰
result = df[(df['å¹´é¾„'] > 28) & (df['è–ªèµ„'] > 10000)]
print("å¹´é¾„>28 ä¸” è–ªèµ„>10000ï¼š")
print(result)

# æ³¨æ„ï¼šå¿…é¡»ç”¨æ‹¬å·ï¼Œå¦åˆ™ä¼šæŠ¥é”™
```

### 3. å¤šæ¡ä»¶ç­›é€‰ - ä½¿ç”¨|ï¼ˆæˆ–ï¼‰

```python
# ä½¿ç”¨æˆ–æ¡ä»¶
result = df[(df['åŸå¸‚'] == 'åŒ—äº¬') | (df['åŸå¸‚'] == 'ä¸Šæµ·')]
print("åŸå¸‚æ˜¯åŒ—äº¬æˆ–ä¸Šæµ·ï¼š")
print(result)
```

### 4. å¤šæ¡ä»¶ç­›é€‰ - ä½¿ç”¨~ï¼ˆéï¼‰

```python
# ä½¿ç”¨éæ¡ä»¶
result = df[~(df['åŸå¸‚'] == 'åŒ—äº¬')]  # ä¸æ˜¯åŒ—äº¬
print("åŸå¸‚ä¸æ˜¯åŒ—äº¬ï¼š")
print(result)
```

### 5. ä½¿ç”¨isin()ç­›é€‰

```python
# ç­›é€‰å€¼åœ¨åˆ—è¡¨ä¸­çš„è¡Œ
result = df[df['åŸå¸‚'].isin(['åŒ—äº¬', 'ä¸Šæµ·'])]
print("åŸå¸‚æ˜¯åŒ—äº¬æˆ–ä¸Šæµ·ï¼ˆä½¿ç”¨isinï¼‰ï¼š")
print(result)

# isin()æ¯”å¤šä¸ªæˆ–æ¡ä»¶æ›´ç®€æ´
```

# ç¬¬å››éƒ¨åˆ†ï¼šé€‰æ‹©åˆ—

## â­ å„ç§åˆ—é€‰æ‹©æ–¹æ³•

### 1. ä½¿ç”¨åˆ—åé€‰æ‹©å•åˆ—

```python
# é€‰æ‹©å•åˆ—ï¼ˆè¿”å›Seriesï¼‰
col = df['å§“å']
print("é€‰æ‹©'å§“å'åˆ—ï¼š")
print(col)
print(f"\nç±»å‹: {type(col)}")
```

### 2. ä½¿ç”¨åˆ—åé€‰æ‹©å¤šåˆ—

```python
# é€‰æ‹©å¤šåˆ—ï¼ˆè¿”å›DataFrameï¼‰
cols = df[['å§“å', 'å¹´é¾„']]
print("é€‰æ‹©'å§“å'å’Œ'å¹´é¾„'åˆ—ï¼š")
print(cols)
print(f"\nç±»å‹: {type(cols)}")
```

### 3. ä½¿ç”¨ç‚¹å·è®¿é—®åˆ—

```python
# ä½¿ç”¨ç‚¹å·è®¿é—®ï¼ˆåˆ—åå¿…é¡»æ˜¯æœ‰æ•ˆçš„Pythonæ ‡è¯†ç¬¦ï¼‰
col = df.å§“å
print("ä½¿ç”¨ç‚¹å·è®¿é—®'å§“å'åˆ—ï¼š")
print(col.head())

# æ³¨æ„ï¼šå¦‚æœåˆ—ååŒ…å«ç©ºæ ¼æˆ–ç‰¹æ®Šå­—ç¬¦ï¼Œä¸èƒ½ä½¿ç”¨ç‚¹å·
```

# ç¬¬äº”éƒ¨åˆ†ï¼šé«˜çº§ç´¢å¼•æŠ€å·§

## â­ ç»„åˆä½¿ç”¨å¤šç§ç´¢å¼•æ–¹å¼

### 1. æ¡ä»¶ç­›é€‰åé€‰æ‹©åˆ—

```python
# å…ˆç­›é€‰è¡Œï¼Œå†é€‰æ‹©åˆ—
result = df[df['å¹´é¾„'] > 30][['å§“å', 'è–ªèµ„']]
print("å¹´é¾„>30çš„è¡Œï¼Œåªæ˜¾ç¤ºå§“åå’Œè–ªèµ„ï¼š")
print(result)

# æ›´æ¨èä½¿ç”¨locï¼ˆæ€§èƒ½æ›´å¥½ï¼‰
result2 = df.loc[df['å¹´é¾„'] > 30, ['å§“å', 'è–ªèµ„']]
print("\nä½¿ç”¨locçš„å†™æ³•ï¼ˆæ¨èï¼‰ï¼š")
print(result2)
```

### 2. ä½¿ç”¨query()æ–¹æ³•

```python
# ä½¿ç”¨query()æ–¹æ³•ï¼ˆæ›´ç®€æ´çš„æ¡ä»¶ç­›é€‰ï¼‰
result = df.query('å¹´é¾„ > 30')
print("ä½¿ç”¨queryç­›é€‰å¹´é¾„>30ï¼š")
print(result)

# å¤šæ¡ä»¶
result2 = df.query('å¹´é¾„ > 28 and è–ªèµ„ > 10000')
print("\nä½¿ç”¨queryå¤šæ¡ä»¶ç­›é€‰ï¼š")
print(result2)
```

### 3. ä½¿ç”¨atå’Œiatè®¿é—®å•ä¸ªå€¼

```python
# at: ä½¿ç”¨æ ‡ç­¾è®¿é—®å•ä¸ªå€¼ï¼ˆæ¯”locæ›´å¿«ï¼‰
value = df.at['A', 'å§“å']
print(f"ä½¿ç”¨atè®¿é—®: {value}")

# iat: ä½¿ç”¨ä½ç½®è®¿é—®å•ä¸ªå€¼ï¼ˆæ¯”ilocæ›´å¿«ï¼‰
value2 = df.iat[0, 0]
print(f"ä½¿ç”¨iatè®¿é—®: {value2}")

# å®é™…åº”ç”¨ï¼šè®¿é—®å•ä¸ªå€¼æ—¶ï¼Œatå’Œiatæ¯”locå’Œilocæ›´å¿«
```

### 4. ä½¿ç”¨where()æ–¹æ³•

```python
# where()ï¼šæ»¡è¶³æ¡ä»¶çš„ä¿ç•™ï¼Œä¸æ»¡è¶³çš„å˜ä¸ºNaN
result = df.where(df['å¹´é¾„'] > 30)
print("ä½¿ç”¨whereç­›é€‰ï¼ˆä¸æ»¡è¶³æ¡ä»¶çš„å˜ä¸ºNaNï¼‰ï¼š")
print(result)

# å®é™…åº”ç”¨ï¼šä¿ç•™æ•°æ®å½¢çŠ¶ï¼Œåªæ˜¾ç¤ºæ»¡è¶³æ¡ä»¶çš„æ•°æ®
```

### 5. ä½¿ç”¨mask()æ–¹æ³•

```python
# mask()ï¼šä¸whereç›¸åï¼Œä¸æ»¡è¶³æ¡ä»¶çš„ä¿ç•™ï¼Œæ»¡è¶³çš„å˜ä¸ºNaN
result = df.mask(df['å¹´é¾„'] > 30)
print("ä½¿ç”¨maskç­›é€‰ï¼ˆæ»¡è¶³æ¡ä»¶çš„å˜ä¸ºNaNï¼‰ï¼š")
print(result)
```
