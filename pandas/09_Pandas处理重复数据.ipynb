## ğŸ“¦ å¯¼å…¥åº“å¹¶åˆ›å»ºåŒ…å«é‡å¤å€¼çš„æ•°æ®

```python
import pandas as pd
import numpy as np

# åˆ›å»ºåŒ…å«é‡å¤å€¼çš„æ•°æ®
data = {
    'å§“å': ['å¼ ä¸‰', 'æå››', 'ç‹äº”', 'å¼ ä¸‰', 'æå››', 'èµµå…­'],
    'å¹´é¾„': [25, 30, 28, 25, 30, 35],
    'è–ªèµ„': [8000, 12000, 9000, 8000, 12000, 15000],
    'åŸå¸‚': ['åŒ—äº¬', 'ä¸Šæµ·', 'å¹¿å·', 'åŒ—äº¬', 'ä¸Šæµ·', 'æ·±åœ³'],
    'éƒ¨é—¨': ['é”€å”®', 'æŠ€æœ¯', 'é”€å”®', 'é”€å”®', 'æŠ€æœ¯', 'æŠ€æœ¯']
}
df = pd.DataFrame(data)

print("åŒ…å«é‡å¤å€¼çš„æ•°æ®ï¼š")
print(df)
```

---
# ç¬¬ä¸€éƒ¨åˆ†ï¼šæ£€æµ‹é‡å¤æ•°æ®

## â­ ä½¿ç”¨duplicated()æ£€æµ‹é‡å¤è¡Œ

### 1. æ£€æµ‹å®Œå…¨é‡å¤çš„è¡Œ

```python
# æ£€æµ‹é‡å¤è¡Œï¼ˆè¿”å›å¸ƒå°”Seriesï¼‰
duplicated = df.duplicated()
print("é‡å¤è¡Œæ£€æµ‹ï¼ˆTrueè¡¨ç¤ºé‡å¤ï¼‰ï¼š")
print(duplicated)

# ç»Ÿè®¡é‡å¤è¡Œæ•°é‡
print(f"\né‡å¤è¡Œæ•°é‡: {duplicated.sum()}")
```

### 2. æŸ¥çœ‹é‡å¤çš„è¡Œ

```python
# æ˜¾ç¤ºæ‰€æœ‰é‡å¤çš„è¡Œ
duplicated_rows = df[df.duplicated()]
print("æ‰€æœ‰é‡å¤çš„è¡Œï¼š")
print(duplicated_rows)

# æ˜¾ç¤ºæ‰€æœ‰é‡å¤çš„è¡Œï¼ˆåŒ…æ‹¬ç¬¬ä¸€æ¬¡å‡ºç°çš„ï¼‰
all_duplicated = df[df.duplicated(keep=False)]
print("\næ‰€æœ‰é‡å¤çš„è¡Œï¼ˆåŒ…æ‹¬ç¬¬ä¸€æ¬¡å‡ºç°çš„ï¼‰ï¼š")
print(all_duplicated)
```

### 3. keepå‚æ•° - æ§åˆ¶æ ‡è®°ç­–ç•¥

```python
# keep='first': ç¬¬ä¸€æ¬¡å‡ºç°çš„ä¸æ ‡è®°ä¸ºé‡å¤ï¼ˆé»˜è®¤ï¼‰
dup_first = df.duplicated(keep='first')
print("keep='first'ï¼ˆé»˜è®¤ï¼‰ï¼š")
print(dup_first)

# keep='last': æœ€åä¸€æ¬¡å‡ºç°çš„ä¸æ ‡è®°ä¸ºé‡å¤
dup_last = df.duplicated(keep='last')
print("\nkeep='last'ï¼š")
print(dup_last)

# keep=False: æ‰€æœ‰é‡å¤çš„éƒ½æ ‡è®°ä¸ºTrue
dup_all = df.duplicated(keep=False)
print("\nkeep=Falseï¼ˆæ‰€æœ‰é‡å¤çš„ï¼‰ï¼š")
print(dup_all)
```

### 4. åŸºäºç‰¹å®šåˆ—æ£€æµ‹é‡å¤

```python
# åªåŸºäº'å§“å'åˆ—æ£€æµ‹é‡å¤
dup_name = df.duplicated(subset=['å§“å'])
print("åŸºäº'å§“å'åˆ—æ£€æµ‹é‡å¤ï¼š")
print(dup_name)

# åŸºäºå¤šåˆ—æ£€æµ‹é‡å¤
dup_multi = df.duplicated(subset=['å§“å', 'å¹´é¾„'])
print("\nåŸºäº'å§“å'å’Œ'å¹´é¾„'åˆ—æ£€æµ‹é‡å¤ï¼š")
print(dup_multi)
```

---
# ç¬¬äºŒéƒ¨åˆ†ï¼šåˆ é™¤é‡å¤æ•°æ®

## â­ ä½¿ç”¨drop_duplicates()åˆ é™¤é‡å¤è¡Œ

### 1. åˆ é™¤å®Œå…¨é‡å¤çš„è¡Œï¼ˆé»˜è®¤ï¼‰

```python
# åˆ é™¤é‡å¤è¡Œï¼ˆé»˜è®¤ä¿ç•™ç¬¬ä¸€ä¸ªï¼‰
df_drop = df.drop_duplicates()
print("åˆ é™¤é‡å¤è¡Œåï¼ˆä¿ç•™ç¬¬ä¸€ä¸ªï¼‰ï¼š")
print(df_drop)
print(f"\nåŸå§‹è¡Œæ•°: {len(df)}, å»é‡åè¡Œæ•°: {len(df_drop)}")
```

### 2. keepå‚æ•° - ä¿ç•™ç­–ç•¥

```python
# keep='first': ä¿ç•™ç¬¬ä¸€æ¬¡å‡ºç°çš„ï¼ˆé»˜è®¤ï¼‰
df_first = df.drop_duplicates(keep='first')
print("ä¿ç•™ç¬¬ä¸€ä¸ªï¼š")
print(df_first)

# keep='last': ä¿ç•™æœ€åä¸€æ¬¡å‡ºç°çš„
df_last = df.drop_duplicates(keep='last')
print("\nä¿ç•™æœ€åä¸€ä¸ªï¼š")
print(df_last)
```

### 3. åŸºäºç‰¹å®šåˆ—å»é‡

```python
# åªåŸºäº'å§“å'åˆ—å»é‡
df_drop = df.drop_duplicates(subset=['å§“å'])
print("åŸºäº'å§“å'åˆ—å»é‡ï¼š")
print(df_drop)

# å®é™…åº”ç”¨ï¼šæ ¹æ®IDæˆ–å”¯ä¸€æ ‡è¯†ç¬¦å»é‡
```

### 4. åŸºäºå¤šåˆ—å»é‡

```python
# åŸºäºå¤šåˆ—å»é‡ï¼ˆæ‰€æœ‰åˆ—éƒ½ç›¸åŒæ‰ç®—é‡å¤ï¼‰
df_drop = df.drop_duplicates(subset=['å§“å', 'å¹´é¾„', 'åŸå¸‚'])
print("åŸºäºå¤šåˆ—å»é‡ï¼š")
print(df_drop)

# å®é™…åº”ç”¨ï¼šæ ¹æ®å¤šä¸ªå­—æ®µç»„åˆåˆ¤æ–­é‡å¤
```

### 5. åŸåœ°åˆ é™¤

```python
# åŸåœ°åˆ é™¤é‡å¤è¡Œï¼ˆç›´æ¥ä¿®æ”¹åŸDataFrameï¼‰
df_inplace = df.copy()
df_inplace.drop_duplicates(inplace=True)
print("åŸåœ°åˆ é™¤é‡å¤è¡Œï¼š")
print(df_inplace)

# inplace=True: ç›´æ¥ä¿®æ”¹åŸDataFrame
```

---
# ç¬¬ä¸‰éƒ¨åˆ†ï¼šé‡å¤æ•°æ®ç»Ÿè®¡å’Œåˆ†æ

## â­ åˆ†æé‡å¤æ•°æ®çš„æƒ…å†µ

### 1. ç»Ÿè®¡é‡å¤è¡Œæ•°é‡

```python
# ç»Ÿè®¡é‡å¤è¡Œæ•°é‡
total_duplicates = df.duplicated().sum()
print(f"å®Œå…¨é‡å¤çš„è¡Œæ•°: {total_duplicates}")

# ç»Ÿè®¡åŸºäºç‰¹å®šåˆ—çš„é‡å¤
name_duplicates = df.duplicated(subset=['å§“å']).sum()
print(f"'å§“å'åˆ—é‡å¤çš„è¡Œæ•°: {name_duplicates}")

# è®¡ç®—é‡å¤ç‡
duplicate_ratio = total_duplicates / len(df) * 100
print(f"\né‡å¤ç‡: {duplicate_ratio:.2f}%")
```

### 2. æŸ¥çœ‹é‡å¤è¡Œçš„è¯¦ç»†ä¿¡æ¯

```python
# è·å–æ‰€æœ‰é‡å¤è¡Œçš„è¯¦ç»†ä¿¡æ¯
duplicated_info = df[df.duplicated(keep=False)].sort_values('å§“å')
print("æ‰€æœ‰é‡å¤è¡Œçš„è¯¦ç»†ä¿¡æ¯ï¼š")
print(duplicated_info)

# å®é™…åº”ç”¨ï¼šåˆ†æé‡å¤æ•°æ®çš„åŸå› 
```

### 3. ç»Ÿè®¡æ¯åˆ—çš„å”¯ä¸€å€¼æ•°é‡

```python
# ç»Ÿè®¡æ¯åˆ—çš„å”¯ä¸€å€¼æ•°é‡
unique_counts = df.nunique()
print("æ¯åˆ—çš„å”¯ä¸€å€¼æ•°é‡ï¼š")
print(unique_counts)

# æ€»è¡Œæ•°
total_rows = len(df)
print(f"\næ€»è¡Œæ•°: {total_rows}")

# è®¡ç®—é‡å¤å¯èƒ½æ€§
duplicate_potential = (total_rows - unique_counts) / total_rows * 100
print("\næ¯åˆ—çš„é‡å¤å¯èƒ½æ€§ï¼ˆ%ï¼‰ï¼š")
print(duplicate_potential.round(2))
```

# ç¬¬å››éƒ¨åˆ†ï¼šå®é™…åº”ç”¨åœºæ™¯

## â­ ä¸åŒåœºæ™¯çš„å»é‡ç­–ç•¥

### 1. æ ¹æ®IDå»é‡ï¼ˆä¿ç•™æœ€æ–°è®°å½•ï¼‰

```python
# åˆ›å»ºå¸¦æ—¶é—´æˆ³çš„æ•°æ®
data_time = {
    'ID': ['A001', 'A002', 'A001', 'A003', 'A002'],
    'å§“å': ['å¼ ä¸‰', 'æå››', 'å¼ ä¸‰', 'ç‹äº”', 'æå››'],
    'æ›´æ–°æ—¶é—´': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']
}
df_time = pd.DataFrame(data_time)
df_time['æ›´æ–°æ—¶é—´'] = pd.to_datetime(df_time['æ›´æ–°æ—¶é—´'])

# å…ˆæŒ‰æ›´æ–°æ—¶é—´æ’åºï¼Œå†æ ¹æ®IDå»é‡ï¼ˆä¿ç•™æœ€æ–°çš„ï¼‰
df_latest = df_time.sort_values('æ›´æ–°æ—¶é—´').drop_duplicates(subset=['ID'], keep='last')
print("ä¿ç•™æ¯ä¸ªIDçš„æœ€æ–°è®°å½•ï¼š")
print(df_latest)
```

### 2. æ ¹æ®å¤šåˆ—å»é‡ï¼ˆä¸šåŠ¡é€»è¾‘ï¼‰

```python
# ä¸šåŠ¡åœºæ™¯ï¼šåŒä¸€äººåŒä¸€åŸå¸‚åªä¿ç•™ä¸€æ¡è®°å½•
df_business = df.drop_duplicates(subset=['å§“å', 'åŸå¸‚'], keep='first')
print("åŒä¸€äººåŒä¸€åŸå¸‚åªä¿ç•™ä¸€æ¡ï¼š")
print(df_business)

# å®é™…åº”ç”¨ï¼šå®¢æˆ·æ•°æ®å»é‡ã€è®¢å•å»é‡ç­‰
```

### 3. ä¿ç•™é‡å¤å€¼ä¸­æŸä¸ªå­—æ®µæœ€å¤§çš„è®°å½•

```python
# ä¿ç•™è–ªèµ„æœ€é«˜çš„è®°å½•
df_max = df.sort_values('è–ªèµ„', ascending=False).drop_duplicates(subset=['å§“å'], keep='first')
print("ä¿ç•™æ¯ä¸ªå§“åä¸­è–ªèµ„æœ€é«˜çš„è®°å½•ï¼š")
print(df_max[['å§“å', 'è–ªèµ„']])

# å®é™…åº”ç”¨ï¼šä¿ç•™æœ€ä¼˜è®°å½•
```

### 4. ç»¼åˆå»é‡å‡½æ•°

```python
def smart_deduplicate(df, subset=None, keep='first', sort_by=None, ascending=False):
    """æ™ºèƒ½å»é‡å‡½æ•°"""
    df_clean = df.copy()
    
    # å¦‚æœæŒ‡å®šäº†æ’åºå­—æ®µï¼Œå…ˆæ’åº
    if sort_by:
        df_clean = df_clean.sort_values(sort_by, ascending=ascending)
    
    # å»é‡
    if subset:
        df_clean = df_clean.drop_duplicates(subset=subset, keep=keep)
    else:
        df_clean = df_clean.drop_duplicates(keep=keep)
    
    return df_clean

# ä½¿ç”¨ç¤ºä¾‹ï¼šä¿ç•™è–ªèµ„æœ€é«˜çš„è®°å½•
df_smart = smart_deduplicate(df, subset=['å§“å'], sort_by='è–ªèµ„', ascending=False)
print("æ™ºèƒ½å»é‡ï¼ˆä¿ç•™è–ªèµ„æœ€é«˜ï¼‰ï¼š")
print(df_smart[['å§“å', 'è–ªèµ„']])
```
