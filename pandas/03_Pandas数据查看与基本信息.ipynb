```python
def quick_overview(df):
    """å¿«é€Ÿæ•°æ®æ¦‚è§ˆå‡½æ•°"""
    print("="*50)
    print("æ•°æ®æ¦‚è§ˆ")
    print("="*50)
    print(f"å½¢çŠ¶: {df.shape[0]} è¡Œ Ã— {df.shape[1]} åˆ—")
    print(f"\nç¼ºå¤±å€¼ç»Ÿè®¡ï¼š")
    print(df.isnull().sum())
    print(f"\næ•°æ®ç±»å‹ï¼š")
    print(df.dtypes)
    print(f"\næ•°å€¼åˆ—ç»Ÿè®¡ï¼š")
    print(df.describe())

# ä½¿ç”¨ç¤ºä¾‹
quick_overview(df)
```

### 2. æ•°æ®è´¨é‡æŠ¥å‘Š

```python
def data_quality_report(df):
    """ç”Ÿæˆæ•°æ®è´¨é‡æŠ¥å‘Š"""
    report = pd.DataFrame({
        'åˆ—å': df.columns,
        'æ•°æ®ç±»å‹': df.dtypes.values,
        'éç©ºå€¼æ•°é‡': df.notna().sum().values,
        'ç¼ºå¤±å€¼æ•°é‡': df.isnull().sum().values,
        'ç¼ºå¤±å€¼æ¯”ä¾‹': (df.isnull().sum() / len(df) * 100).values,
        'å”¯ä¸€å€¼æ•°é‡': [df[col].nunique() for col in df.columns]
    })
    return report

# ç”ŸæˆæŠ¥å‘Š
quality_report = data_quality_report(df)
print("æ•°æ®è´¨é‡æŠ¥å‘Šï¼š")
print(quality_report)

# å®é™…åº”ç”¨ï¼šå¿«é€Ÿäº†è§£æ•°æ®è´¨é‡ï¼Œå‘ç°æ•°æ®é—®é¢˜
```
