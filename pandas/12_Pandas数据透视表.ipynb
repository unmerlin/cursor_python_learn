```python
import pandas as pd
import numpy as np

# åˆ›å»ºç¤ºä¾‹é”€å”®æ•°æ®
data = {
    'é”€å”®å‘˜': ['å¼ ä¸‰', 'æå››', 'ç‹äº”', 'å¼ ä¸‰', 'æå››', 'ç‹äº”', 'å¼ ä¸‰', 'æå››', 'ç‹äº”'],
    'äº§å“': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'æœˆä»½': [1, 1, 1, 2, 2, 2, 3, 3, 3],
    'é”€å”®é¢': [1000, 1500, 1200, 2000, 1800, 1600, 900, 1100, 1000],
    'æ•°é‡': [10, 15, 12, 20, 18, 16, 9, 11, 10]
}
df = pd.DataFrame(data)

print("ç¤ºä¾‹é”€å”®æ•°æ®ï¼š")
print(df)
```
