## 2. å®‰è£…ä¸ç¯å¢ƒé…ç½®


```python
# å®‰è£…Seabornï¼ˆå¦‚æœå°šæœªå®‰è£…ï¼‰
# !pip install seaborn matplotlib pandas numpy

# å¯¼å…¥å¿…è¦çš„åº“
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# è®¾ç½®ä¸­æ–‡å­—ä½“æ”¯æŒï¼ˆè§£å†³ä¸­æ–‡æ˜¾ç¤ºé—®é¢˜ï¼‰
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei']  # ç”¨æ¥æ­£å¸¸æ˜¾ç¤ºä¸­æ–‡æ ‡ç­¾
plt.rcParams['axes.unicode_minus'] = False  # ç”¨æ¥æ­£å¸¸æ˜¾ç¤ºè´Ÿå·

# è®¾ç½®Seabornæ ·å¼
sns.set_style("whitegrid")  # è®¾ç½®èƒŒæ™¯ç½‘æ ¼æ ·å¼
sns.set_palette("husl")  # è®¾ç½®è°ƒè‰²æ¿

# è®¾ç½®å›¾è¡¨å¤§å°
plt.rcParams['figure.figsize'] = (10, 6)

print("Seabornç‰ˆæœ¬:", sns.__version__)
print("ç¯å¢ƒé…ç½®å®Œæˆï¼")
```

## 3. Seaborn vs Matplotlib

### Seabornç›¸æ¯”Matplotlibçš„ä¸»è¦ä¼˜åŠ¿ï¼š

1. **æ›´ç¾è§‚çš„é»˜è®¤æ ·å¼**ï¼šSeabornæä¾›äº†æ›´ç°ä»£ã€æ›´ä¸“ä¸šçš„é»˜è®¤å›¾è¡¨æ ·å¼
2. **æ›´ç®€æ´çš„API**ï¼šç”¨æ›´å°‘çš„ä»£ç å®ç°æ›´å¤æ‚çš„å¯è§†åŒ–
3. **ç»Ÿè®¡åŠŸèƒ½é›†æˆ**ï¼šå†…ç½®ç»Ÿè®¡ä¼°è®¡å’Œè¯¯å·®æ¡ç­‰åŠŸèƒ½
4. **æ›´å¥½çš„é¢œè‰²æ”¯æŒ**ï¼šæä¾›å¤šç§è°ƒè‰²æ¿ï¼Œæ”¯æŒåˆ†ç±»å’Œè¿ç»­æ•°æ®
5. **å¤šå˜é‡å¯è§†åŒ–**ï¼šè½»æ¾å¤„ç†å¤šå˜é‡æ•°æ®çš„å¯è§†åŒ–
6. **è‡ªåŠ¨å›¾ä¾‹å’Œæ ‡ç­¾**ï¼šè‡ªåŠ¨å¤„ç†å›¾ä¾‹ã€æ ‡ç­¾ç­‰ç»†èŠ‚


```python
# åˆ›å»ºç¤ºä¾‹æ•°æ®ç”¨äºå¯¹æ¯”
np.random.seed(42)
data = pd.DataFrame({
    'ç±»åˆ«': ['A', 'B', 'C', 'D', 'E'] * 20,
    'æ•°å€¼': np.random.randn(100),
    'åˆ†ç»„': np.random.choice(['ç»„1', 'ç»„2'], 100)
})

# Matplotlibæ–¹å¼ï¼ˆéœ€è¦æ›´å¤šä»£ç ï¼‰
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Matplotlibç‰ˆæœ¬
axes[0].bar(data['ç±»åˆ«'].unique(), 
            [data[data['ç±»åˆ«']==cat]['æ•°å€¼'].mean() for cat in data['ç±»åˆ«'].unique()])
axes[0].set_title('Matplotlibç‰ˆæœ¬ - éœ€è¦æ‰‹åŠ¨è®¡ç®—ç»Ÿè®¡é‡', fontsize=12)
axes[0].set_xlabel('ç±»åˆ«')
axes[0].set_ylabel('å¹³å‡å€¼')

# Seabornç‰ˆæœ¬ï¼ˆè‡ªåŠ¨è®¡ç®—ç»Ÿè®¡é‡ï¼Œæ ·å¼æ›´ç¾è§‚ï¼‰
sns.barplot(data=data, x='ç±»åˆ«', y='æ•°å€¼', ax=axes[1])
axes[1].set_title('Seabornç‰ˆæœ¬ - è‡ªåŠ¨è®¡ç®—ç»Ÿè®¡é‡ï¼Œæ ·å¼æ›´ç¾è§‚', fontsize=12)

plt.tight_layout()
plt.show()

print("å¯¹æ¯”è¯´æ˜ï¼š")
print("1. Seabornè‡ªåŠ¨è®¡ç®—å¹³å‡å€¼å’Œç½®ä¿¡åŒºé—´")
print("2. Seabornæä¾›æ›´ç¾è§‚çš„é»˜è®¤æ ·å¼å’Œé¢œè‰²")
print("3. Seabornä»£ç æ›´ç®€æ´ï¼Œä¸€è¡Œä»£ç å®Œæˆç»Ÿè®¡å’Œå¯è§†åŒ–")
```

## 4. åŸºç¡€å›¾è¡¨ç±»å‹

### 4.1 æ•£ç‚¹å›¾ (scatterplot)

**Seabornä¼˜åŠ¿**ï¼šæ”¯æŒè‡ªåŠ¨åˆ†ç»„ç€è‰²ã€å›å½’çº¿ã€è¾¹é™…åˆ†å¸ƒç­‰é«˜çº§åŠŸèƒ½


```python
# åˆ›å»ºç¤ºä¾‹æ•°æ®
np.random.seed(42)
scatter_data = pd.DataFrame({
    'x': np.random.randn(200),
    'y': np.random.randn(200) * 0.5 + np.random.randn(200) * 0.3,
    'ç±»åˆ«': np.random.choice(['ç±»å‹1', 'ç±»å‹2', 'ç±»å‹3'], 200),
    'å¤§å°': np.random.randint(20, 200, 200)
})

# åŸºç¡€æ•£ç‚¹å›¾
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(data=scatter_data, x='x', y='y')
plt.title('åŸºç¡€æ•£ç‚¹å›¾')

# å¸¦åˆ†ç»„é¢œè‰²çš„æ•£ç‚¹å›¾ï¼ˆSeabornè‡ªåŠ¨å¤„ç†å›¾ä¾‹å’Œé¢œè‰²ï¼‰
plt.subplot(1, 2, 2)
sns.scatterplot(data=scatter_data, x='x', y='y', hue='ç±»åˆ«', size='å¤§å°', sizes=(50, 200))
plt.title('å¤šç»´åº¦æ•£ç‚¹å›¾ï¼ˆé¢œè‰²+å¤§å°ï¼‰')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()

print("Seabornä¼˜åŠ¿ï¼š")
print("- è‡ªåŠ¨æ ¹æ®åˆ†ç±»å˜é‡ç€è‰²")
print("- æ”¯æŒå¤§å°æ˜ å°„ï¼ˆsizeå‚æ•°ï¼‰")
print("- è‡ªåŠ¨ç”Ÿæˆå›¾ä¾‹")
print("- æ›´ç¾è§‚çš„é»˜è®¤æ ·å¼")
```

### 4.2 çº¿å›¾ (lineplot)

**Seabornä¼˜åŠ¿**ï¼šè‡ªåŠ¨è®¡ç®—ç½®ä¿¡åŒºé—´ï¼Œæ”¯æŒå¤šå˜é‡åˆ†ç»„ï¼Œæ›´ç¾è§‚çš„æ ·å¼


```python
# åˆ›å»ºæ—¶é—´åºåˆ—æ•°æ®
dates = pd.date_range('2023-01-01', periods=100, freq='D')
line_data = pd.DataFrame({
    'æ—¥æœŸ': dates,
    'é”€å”®é¢': np.cumsum(np.random.randn(100).cumsum()) + 1000,
    'äº§å“': np.random.choice(['äº§å“A', 'äº§å“B', 'äº§å“C'], 100)
})

plt.figure(figsize=(14, 5))

# åŸºç¡€çº¿å›¾
plt.subplot(1, 2, 1)
sns.lineplot(data=line_data, x='æ—¥æœŸ', y='é”€å”®é¢')
plt.title('åŸºç¡€çº¿å›¾')
plt.xticks(rotation=45)

# å¸¦åˆ†ç»„å’Œç½®ä¿¡åŒºé—´çš„çº¿å›¾ï¼ˆSeabornè‡ªåŠ¨è®¡ç®—ï¼‰
plt.subplot(1, 2, 2)
sns.lineplot(data=line_data, x='æ—¥æœŸ', y='é”€å”®é¢', hue='äº§å“', style='äº§å“', markers=True)
plt.title('åˆ†ç»„çº¿å›¾ï¼ˆè‡ªåŠ¨ç½®ä¿¡åŒºé—´ï¼‰')
plt.xticks(rotation=45)
plt.legend(title='äº§å“ç±»å‹')

plt.tight_layout()
plt.show()

print("Seabornä¼˜åŠ¿ï¼š")
print("- è‡ªåŠ¨è®¡ç®—å¹¶æ˜¾ç¤ºç½®ä¿¡åŒºé—´ï¼ˆciå‚æ•°ï¼‰")
print("- æ”¯æŒæ ·å¼æ˜ å°„ï¼ˆstyleå‚æ•°ï¼‰")
print("- è‡ªåŠ¨æ ‡è®°ç‚¹ï¼ˆmarkerså‚æ•°ï¼‰")
print("- æ›´å¹³æ»‘çš„çº¿æ¡æ ·å¼")
```

### 4.3 æŸ±çŠ¶å›¾ (barplot)

**Seabornä¼˜åŠ¿**ï¼šè‡ªåŠ¨è®¡ç®—ç»Ÿè®¡é‡ï¼ˆé»˜è®¤æ˜¯å¹³å‡å€¼ï¼‰ï¼Œæ˜¾ç¤ºè¯¯å·®æ¡ï¼Œæ”¯æŒåˆ†ç»„


```python
# åˆ›å»ºç¤ºä¾‹æ•°æ®
bar_data = pd.DataFrame({
    'æœˆä»½': ['1æœˆ', '2æœˆ', '3æœˆ', '4æœˆ', '5æœˆ', '6æœˆ'] * 10,
    'é”€å”®é¢': np.random.randint(1000, 5000, 60),
    'åœ°åŒº': np.random.choice(['åŒ—äº¬', 'ä¸Šæµ·', 'å¹¿å·'], 60)
})

plt.figure(figsize=(14, 5))

# åŸºç¡€æŸ±çŠ¶å›¾ï¼ˆè‡ªåŠ¨è®¡ç®—å¹³å‡å€¼å’Œç½®ä¿¡åŒºé—´ï¼‰
plt.subplot(1, 2, 1)
sns.barplot(data=bar_data, x='æœˆä»½', y='é”€å”®é¢')
plt.title('åŸºç¡€æŸ±çŠ¶å›¾ï¼ˆè‡ªåŠ¨è®¡ç®—å¹³å‡å€¼ï¼‰')
plt.ylabel('å¹³å‡é”€å”®é¢')

# åˆ†ç»„æŸ±çŠ¶å›¾ï¼ˆSeabornè‡ªåŠ¨å¤„ç†åˆ†ç»„å’Œé¢œè‰²ï¼‰
plt.subplot(1, 2, 2)
sns.barplot(data=bar_data, x='æœˆä»½', y='é”€å”®é¢', hue='åœ°åŒº')
plt.title('åˆ†ç»„æŸ±çŠ¶å›¾ï¼ˆè‡ªåŠ¨åˆ†ç»„å’Œé…è‰²ï¼‰')
plt.ylabel('å¹³å‡é”€å”®é¢')
plt.legend(title='åœ°åŒº')

plt.tight_layout()
plt.show()

print("Seabornä¼˜åŠ¿ï¼š")
print("- è‡ªåŠ¨è®¡ç®—ç»Ÿè®¡é‡ï¼ˆestimatorå‚æ•°ï¼Œé»˜è®¤æ˜¯meanï¼‰")
print("- è‡ªåŠ¨æ˜¾ç¤ºè¯¯å·®æ¡ï¼ˆç½®ä¿¡åŒºé—´ï¼‰")
print("- è‡ªåŠ¨å¤„ç†åˆ†ç»„å’Œé¢œè‰²æ˜ å°„")
print("- æ›´ä¸“ä¸šçš„é»˜è®¤æ ·å¼")
```

### 4.4 ç®±çº¿å›¾ (boxplot)

**Seabornä¼˜åŠ¿**ï¼šæ›´ç¾è§‚çš„æ ·å¼ï¼Œæ”¯æŒåˆ†ç»„ï¼Œè‡ªåŠ¨å¤„ç†å¼‚å¸¸å€¼æ˜¾ç¤º


```python
# åˆ›å»ºåŒ…å«å¼‚å¸¸å€¼çš„æ•°æ®
box_data = pd.DataFrame({
    'ç±»åˆ«': ['A', 'B', 'C', 'D'] * 25,
    'æ•°å€¼': np.concatenate([
        np.random.normal(20, 5, 25),
        np.random.normal(30, 8, 25),
        np.random.normal(25, 6, 25),
        np.random.normal(35, 10, 25)
    ]),
    'åˆ†ç»„': np.random.choice(['ç»„1', 'ç»„2'], 100)
})

# æ·»åŠ ä¸€äº›å¼‚å¸¸å€¼
box_data.loc[10, 'æ•°å€¼'] = 80
box_data.loc[50, 'æ•°å€¼'] = -10

plt.figure(figsize=(14, 5))

# åŸºç¡€ç®±çº¿å›¾
plt.subplot(1, 2, 1)
sns.boxplot(data=box_data, x='ç±»åˆ«', y='æ•°å€¼')
plt.title('åŸºç¡€ç®±çº¿å›¾ï¼ˆè‡ªåŠ¨æ˜¾ç¤ºå¼‚å¸¸å€¼ï¼‰')

# åˆ†ç»„ç®±çº¿å›¾
plt.subplot(1, 2, 2)
sns.boxplot(data=box_data, x='ç±»åˆ«', y='æ•°å€¼', hue='åˆ†ç»„')
plt.title('åˆ†ç»„ç®±çº¿å›¾')
plt.legend(title='åˆ†ç»„')

plt.tight_layout()
plt.show()

print("Seabornä¼˜åŠ¿ï¼š")
print("- æ›´ç¾è§‚çš„ç®±çº¿å›¾æ ·å¼")
print("- è‡ªåŠ¨è¯†åˆ«å’Œæ˜¾ç¤ºå¼‚å¸¸å€¼")
print("- æ”¯æŒåˆ†ç»„å’Œé¢œè‰²æ˜ å°„")
print("- æ›´å¥½çš„é»˜è®¤é…è‰²æ–¹æ¡ˆ")
```

### 4.5 å°æç´å›¾ (violinplot)

**Seabornä¼˜åŠ¿**ï¼šè¿™æ˜¯Seabornç‰¹æœ‰çš„å›¾è¡¨ç±»å‹ï¼Œç»“åˆäº†ç®±çº¿å›¾å’Œæ ¸å¯†åº¦ä¼°è®¡ï¼Œæ˜¾ç¤ºæ•°æ®åˆ†å¸ƒå½¢çŠ¶


```python
# ä½¿ç”¨ä¹‹å‰åˆ›å»ºçš„box_data
plt.figure(figsize=(14, 5))

# åŸºç¡€å°æç´å›¾
plt.subplot(1, 2, 1)
sns.violinplot(data=box_data, x='ç±»åˆ«', y='æ•°å€¼')
plt.title('åŸºç¡€å°æç´å›¾ï¼ˆæ˜¾ç¤ºåˆ†å¸ƒå½¢çŠ¶ï¼‰')

# åˆ†ç»„å°æç´å›¾
plt.subplot(1, 2, 2)
sns.violinplot(data=box_data, x='ç±»åˆ«', y='æ•°å€¼', hue='åˆ†ç»„', split=True)
plt.title('åˆ†ç»„å°æç´å›¾ï¼ˆsplitæ¨¡å¼ï¼‰')
plt.legend(title='åˆ†ç»„')

plt.tight_layout()
plt.show()

print("Seabornä¼˜åŠ¿ï¼š")
print("- æ˜¾ç¤ºæ•°æ®åˆ†å¸ƒçš„å®Œæ•´å½¢çŠ¶ï¼ˆè€Œä¸ä»…ä»…æ˜¯ç»Ÿè®¡é‡ï¼‰")
print("- ç»“åˆäº†ç®±çº¿å›¾å’Œå¯†åº¦å›¾çš„ä¿¡æ¯")
print("- æ”¯æŒsplitæ¨¡å¼ï¼Œä¾¿äºå¯¹æ¯”ä¸¤ä¸ªåˆ†ç»„")
print("- è¿™æ˜¯Matplotlibæ²¡æœ‰çš„é«˜çº§å›¾è¡¨ç±»å‹")
```

## 5. ç»Ÿè®¡å›¾è¡¨

### 5.1 åˆ†å¸ƒå›¾ (distplot/histplot)

**Seabornä¼˜åŠ¿**ï¼šå¯ä»¥åŒæ—¶æ˜¾ç¤ºç›´æ–¹å›¾å’Œæ ¸å¯†åº¦ä¼°è®¡ï¼Œè‡ªåŠ¨é€‰æ‹©æœ€ä¼˜bins


```python
# åˆ›å»ºç¤ºä¾‹æ•°æ®
dist_data = pd.DataFrame({
    'æ•°å€¼1': np.random.normal(0, 1, 1000),
    'æ•°å€¼2': np.random.normal(2, 1.5, 1000),
    'ç±»åˆ«': np.random.choice(['A', 'B'], 1000)
})

plt.figure(figsize=(14, 5))

# åŸºç¡€åˆ†å¸ƒå›¾ï¼ˆç›´æ–¹å›¾ + å¯†åº¦æ›²çº¿ï¼‰
plt.subplot(1, 2, 1)
sns.histplot(data=dist_data, x='æ•°å€¼1', kde=True, bins=30)
plt.title('åˆ†å¸ƒå›¾ï¼ˆç›´æ–¹å›¾ + å¯†åº¦æ›²çº¿ï¼‰')

# åˆ†ç»„åˆ†å¸ƒå›¾
plt.subplot(1, 2, 2)
sns.histplot(data=dist_data, x='æ•°å€¼1', hue='ç±»åˆ«', kde=True, alpha=0.6)
plt.title('åˆ†ç»„åˆ†å¸ƒå›¾ï¼ˆè‡ªåŠ¨å åŠ ï¼‰')
plt.legend(title='ç±»åˆ«')

plt.tight_layout()
plt.show()

print("Seabornä¼˜åŠ¿ï¼š")
print("- å¯ä»¥åŒæ—¶æ˜¾ç¤ºç›´æ–¹å›¾å’Œæ ¸å¯†åº¦ä¼°è®¡ï¼ˆkde=Trueï¼‰")
print("- è‡ªåŠ¨é€‰æ‹©æœ€ä¼˜çš„binsæ•°é‡")
print("- æ”¯æŒåˆ†ç»„å åŠ æ˜¾ç¤º")
print("- æ›´ç¾è§‚çš„é»˜è®¤æ ·å¼")
```

### 5.2 å›å½’å›¾ (regplot/lmplot)

**Seabornä¼˜åŠ¿**ï¼šè‡ªåŠ¨æ‹Ÿåˆå›å½’çº¿ï¼Œæ˜¾ç¤ºç½®ä¿¡åŒºé—´ï¼Œæ”¯æŒå¤šç§å›å½’æ¨¡å‹


```python
# åˆ›å»ºæœ‰ç›¸å…³æ€§çš„æ•°æ®
np.random.seed(42)
x = np.linspace(0, 10, 100)
y = 2 * x + 1 + np.random.randn(100) * 2
reg_data = pd.DataFrame({'x': x, 'y': y, 'ç±»åˆ«': np.random.choice(['A', 'B'], 100)})

plt.figure(figsize=(14, 5))

# åŸºç¡€å›å½’å›¾
plt.subplot(1, 2, 1)
sns.regplot(data=reg_data, x='x', y='y')
plt.title('åŸºç¡€å›å½’å›¾ï¼ˆè‡ªåŠ¨æ‹Ÿåˆç›´çº¿å’Œç½®ä¿¡åŒºé—´ï¼‰')

# åˆ†ç»„å›å½’å›¾ï¼ˆä½¿ç”¨lmplotï¼‰
plt.subplot(1, 2, 2)
# æ³¨æ„ï¼šlmplotä½¿ç”¨FacetGridï¼Œéœ€è¦å•ç‹¬åˆ›å»ºfigure
plt.close()  # å…³é—­ä¸Šä¸€ä¸ªfigure

g = sns.lmplot(data=reg_data, x='x', y='y', hue='ç±»åˆ«', height=5, aspect=1.2)
g.fig.suptitle('åˆ†ç»„å›å½’å›¾ï¼ˆæ¯ä¸ªç±»åˆ«ç‹¬ç«‹æ‹Ÿåˆï¼‰', y=1.02)
plt.show()

print("Seabornä¼˜åŠ¿ï¼š")
print("- è‡ªåŠ¨æ‹Ÿåˆå›å½’çº¿ï¼ˆçº¿æ€§æˆ–å¤šé¡¹å¼ï¼‰")
print("- è‡ªåŠ¨æ˜¾ç¤ºç½®ä¿¡åŒºé—´å¸¦")
print("- æ”¯æŒå¤šç§å›å½’æ¨¡å‹ï¼ˆorderå‚æ•°ç”¨äºå¤šé¡¹å¼å›å½’ï¼‰")
print("- å¯ä»¥è½»æ¾è¿›è¡Œåˆ†ç»„å›å½’åˆ†æ")
```

### 5.3 çƒ­åŠ›å›¾ (heatmap)

**Seabornä¼˜åŠ¿**ï¼šæ›´ç¾è§‚çš„é…è‰²æ–¹æ¡ˆï¼Œè‡ªåŠ¨æ ‡æ³¨æ•°å€¼ï¼Œæ”¯æŒèšç±»


```python
# åˆ›å»ºç›¸å…³æ€§çŸ©é˜µæ•°æ®
np.random.seed(42)
corr_data = pd.DataFrame(np.random.randn(10, 10))
corr_matrix = corr_data.corr()

plt.figure(figsize=(14, 5))

# åŸºç¡€çƒ­åŠ›å›¾
plt.subplot(1, 2, 1)
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('åŸºç¡€çƒ­åŠ›å›¾ï¼ˆè‡ªåŠ¨æ ‡æ³¨æ•°å€¼ï¼‰')

# æ›´é«˜çº§çš„çƒ­åŠ›å›¾ï¼ˆéšè—ä¸Šä¸‰è§’ï¼Œæ·»åŠ maskï¼‰
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
plt.subplot(1, 2, 2)
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='RdYlBu_r', 
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('é«˜çº§çƒ­åŠ›å›¾ï¼ˆéšè—ä¸Šä¸‰è§’ï¼‰')

plt.tight_layout()
plt.show()

print("Seabornä¼˜åŠ¿ï¼š")
print("- æ›´ç¾è§‚çš„é»˜è®¤é…è‰²æ–¹æ¡ˆï¼ˆå¤šç§cmapå¯é€‰ï¼‰")
print("- è‡ªåŠ¨æ ‡æ³¨æ•°å€¼ï¼ˆannot=Trueï¼‰")
print("- æ”¯æŒmaskå‚æ•°éšè—éƒ¨åˆ†æ•°æ®")
print("- æ›´å¥½çš„é¢œè‰²æ¡ï¼ˆcolorbarï¼‰æ§åˆ¶")
```

### 6.3 åˆ†ç±»å›¾ (catplot)

**Seabornä¼˜åŠ¿**ï¼šç»Ÿä¸€çš„æ¥å£å¤„ç†å„ç§åˆ†ç±»æ•°æ®å¯è§†åŒ–ï¼Œæ”¯æŒåˆ†é¢ï¼ˆfacetï¼‰


```python
# åˆ›å»ºåˆ†ç±»æ•°æ®
cat_data = pd.DataFrame({
    'ç±»åˆ«': ['A', 'B', 'C', 'D'] * 30,
    'æ•°å€¼': np.random.randn(120),
    'åˆ†ç»„': np.random.choice(['ç»„1', 'ç»„2'], 120),
    'æ—¶é—´': np.random.choice(['æ—©', 'æ™š'], 120)
})

# catplotæ˜¯åˆ†ç±»å›¾çš„ç»Ÿä¸€æ¥å£ï¼Œå¯ä»¥é€šè¿‡kindå‚æ•°é€‰æ‹©ä¸åŒç±»å‹
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# ç®±çº¿å›¾
sns.catplot(data=cat_data, x='ç±»åˆ«', y='æ•°å€¼', kind='box', ax=axes[0, 0])
axes[0, 0].set_title('ç®±çº¿å›¾ (kind="box")')

# å°æç´å›¾
sns.catplot(data=cat_data, x='ç±»åˆ«', y='æ•°å€¼', kind='violin', ax=axes[0, 1])
axes[0, 1].set_title('å°æç´å›¾ (kind="violin")')

# æ¡å½¢å›¾
sns.catplot(data=cat_data, x='ç±»åˆ«', y='æ•°å€¼', kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('æ¡å½¢å›¾ (kind="bar")')

# ç‚¹å›¾
sns.catplot(data=cat_data, x='ç±»åˆ«', y='æ•°å€¼', kind='point', ax=axes[1, 1])
axes[1, 1].set_title('ç‚¹å›¾ (kind="point")')

plt.tight_layout()
plt.show()

# ä½¿ç”¨åˆ†é¢åŠŸèƒ½ï¼ˆcolå‚æ•°ï¼‰
g = sns.catplot(data=cat_data, x='ç±»åˆ«', y='æ•°å€¼', kind='box', 
                col='æ—¶é—´', hue='åˆ†ç»„', height=5)
g.fig.suptitle('åˆ†é¢åˆ†ç±»å›¾ï¼ˆæŒ‰æ—¶é—´åˆ†é¢ï¼ŒæŒ‰åˆ†ç»„ç€è‰²ï¼‰', y=1.02)
plt.show()

print("Seabornä¼˜åŠ¿ï¼š")
print("- ç»Ÿä¸€çš„APIå¤„ç†æ‰€æœ‰åˆ†ç±»å›¾ç±»å‹")
print("- æ”¯æŒåˆ†é¢åŠŸèƒ½ï¼ˆcolå’Œrowå‚æ•°ï¼‰")
print("- è‡ªåŠ¨å¤„ç†åˆ†ç»„å’Œé¢œè‰²æ˜ å°„")
print("- æ›´çµæ´»çš„å¸ƒå±€æ§åˆ¶")
```
